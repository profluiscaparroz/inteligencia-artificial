## 1. O que são Algoritmos Genéticos?

Os **Algoritmos Genéticos (AGs)** são métodos de busca e otimização **metaheurísticos** inspirados na seleção natural darwiniana. Fundamentados no princípio da **sobrevivência do mais apto** (*survival of the fittest*), esses algoritmos mantêm uma **população de soluções candidatas** e as fazem evoluir ao longo de gerações por meio de operadores genéticos.

**John Holland** formalizou a teoria dos AGs em 1975, estabelecendo a base matemática para a evolução artificial. Desde então, os AGs tornaram-se uma das ferramentas mais poderosas para resolver problemas de otimização onde métodos exatos são impraticáveis devido à complexidade combinatorial.

**Ideia central:** Soluções de alta qualidade tendem a produzir descendentes que também são de alta qualidade. Combinando características de boas soluções e introduzindo variações aleatórias, o algoritmo explora o espaço de busca de forma eficiente.

### Características fundamentais:
- Trabalham com uma **população** de soluções (não apenas uma)
- Utilizam operadores **estocásticos** (não determinísticos)
- Não requerem informação de **gradiente** da função objetivo
- São naturalmente **paralelizáveis**
- Aplicáveis a qualquer problema que possa ser **codificado** como cromossomo

## 3. Terminologia Biológica vs. Computacional

Os AGs utilizam uma terminologia fortemente inspirada na biologia. A tabela abaixo apresenta o mapeamento entre os conceitos biológicos e suas contrapartes computacionais:

| Termo Biológico | Equivalente Computacional | Descrição |
|----------------|--------------------------|----------|
| **Cromossomo** | Solução candidata (indivíduo) | Uma única solução para o problema |
| **Gene** | Parâmetro ou variável da solução | Um componente da solução |
| **Alelo** | Valor de um gene | O valor assumido por um parâmetro |
| **Genótipo** | Representação interna da solução (ex: vetor de bits) | Como a solução é codificada |
| **Fenótipo** | Solução decodificada (ex: rota do caixeiro) | O que a solução significa no problema |
| **Fitness** | Qualidade da solução (função objetivo) | Quão boa é a solução |
| **População** | Conjunto de soluções candidatas | Todas as soluções da geração atual |
| **Geração** | Uma iteração do algoritmo | Um ciclo completo de evolução |
| **Seleção** | Escolha dos melhores para reprodução | Quais soluções geram descendentes |
| **Crossover** | Combinação de dois pais para gerar filhos | Recombinação de soluções |
| **Mutação** | Pequena alteração aleatória na solução | Introdução de diversidade genética |

> ⚠️ **Nota importante:** A distinção entre **genótipo** e **fenótipo** é crucial. O AG opera no espaço do genótipo, mas a função de fitness avalia o fenótipo.

## 4. O Ciclo do Algoritmo Genético

O ciclo de um AG pode ser descrito em etapas sequenciais que se repetem até a condição de parada ser satisfeita:

1. **Inicialização:** Gerar uma população aleatória de N indivíduos (soluções candidatas)
2. **Avaliação de Fitness:** Calcular a qualidade de cada indivíduo usando a função objetivo
3. **Seleção:** Escolher os pais para reprodução com base no fitness (melhores têm maior probabilidade)
4. **Crossover:** Combinar dois pais para gerar filhos que herdam características de ambos
5. **Mutação:** Aplicar pequenas alterações aleatórias nos filhos para manter diversidade
6. **Nova Geração:** Substituir a população antiga pela nova geração de filhos
7. **Critério de Parada:** Verificar se número máximo de gerações foi atingido ou se o fitness é satisfatório
8. **Repetir** a partir do passo 2 até a condição de parada

```
┌─────────────────────────────────────────────────────────┐
│              CICLO DO ALGORITMO GENÉTICO                │
│                                                         │
│   ┌──────────────────┐                                  │
│   │  Inicialização   │  Gerar população aleatória       │
│   └────────┬─────────┘                                  │
│            │                                            │
│            ▼                                            │
│   ┌──────────────────┐                                  │
│   │    Avaliação     │  Calcular fitness de cada        │
│   │    de Fitness    │  indivíduo                       │
│   └────────┬─────────┘                                  │
│            │                                            │
│            ▼                                            │
│   ┌──────────────────┐    Critério     ┌─────────────┐  │
│   │    Critério de   │───  Satisfeito ─►   FIM / Best│  │

│            │ Não                                        │

│   ┌──────────────────┐                                  │

```

## 5. Operadores Genéticos em Detalhes

### 5.1 Seleção

A seleção determina quais indivíduos têm o privilégio de se reproduzir, influenciando diretamente a **pressão seletiva** do algoritmo.

**Roleta (Roulette Wheel Selection):**
- A probabilidade de seleção de cada indivíduo é proporcional ao seu fitness
- Implementação: seleção cumulativa (CDF)
- Vantagem: indivíduos com alto fitness dominam rapidamente
- Desvantagem: pode causar convergência prematura

**Torneio (Tournament Selection):**
- Escolher `k` indivíduos aleatoriamente e selecionar o melhor
- Controle de pressão seletiva pelo parâmetro `k`
- Mais robusto que a roleta em termos de diversidade

**Elitismo:**
- Manter os `N` melhores indivíduos da geração anterior diretamente
- Garante que a melhor solução nunca se perde
- Geralmente combinado com outros métodos de seleção

---

### 5.2 Crossover

O crossover recombina material genético de dois pais para produzir filhos.

**Ponto único:** Divide o cromossomo em um ponto e troca as partes
```
Pai1: [1 0 1 1 | 0 0 1]      Filho1: [1 0 1 1 | 1 0 0]
Pai2: [0 1 0 0 | 1 0 0]  →   Filho2: [0 1 0 0 | 0 0 1]
```

**Dois pontos:** Dois pontos de corte, troca o segmento interno
```
Pai1: [1 0 | 1 1 0 | 0 1]    Filho1: [1 0 | 0 1 1 | 0 1]
Pai2: [0 1 | 0 1 1 | 1 0]  → Filho2: [0 1 | 1 1 0 | 1 0]
```

**Uniforme:** Cada gene tem 50% de chance de vir de cada pai

**PMX (Partially Mapped Crossover):** Para problemas de permutação (TSP)
- Preserva a validade da permutação
- Mapeia genes conflitantes entre as cópias dos pais

---

### 5.3 Mutação

A mutação introduz diversidade genética para evitar convergência prematura.

| Tipo | Representação | Descrição |
|------|--------------|----------|
| **Bit-flip** | Binária | Inverter um bit aleatório |
| **Swap** | Permutação | Trocar dois genes de posição |
| **Inversão** | Permutação | Inverter um segmento do cromossomo |
| **Real-valued** | Real | Adicionar ruído gaussiano: `x' = x + N(0, σ)` |

> ⚠️ **Taxa de mutação:** Tipicamente baixa (0.001 a 0.1). Muito alta → busca aleatória. Muito baixa → convergência prematura.

````python
import numpy as np
import matplotlib.pyplot as plt
import random
from copy import deepcopy

# Configurar seed para reprodutibilidade
np.random.seed(42)
random.seed(42)

class AlgoritmoGeneticoTSP:
    """
    Algoritmo Genético para o Problema do Caixeiro Viajante (TSP).
    Representação: permutação de cidades.
    """
    
    def __init__(self, cidades, tamanho_populacao=50, geracoes=200,
                 taxa_crossover=0.8, taxa_mutacao=0.02, elitismo=2):
        self.cidades = np.array(cidades)
        self.n_cidades = len(cidades)
        self.tamanho_pop = tamanho_populacao
        self.geracoes = geracoes
        self.taxa_crossover = taxa_crossover
        self.taxa_mutacao = taxa_mutacao
        self.elitismo = elitismo
        self.melhor_fitness_hist = []
        self.media_fitness_hist = []
    
    def distancia_total(self, rota):
        """Calcula distância total de uma rota (volta ao início)."""
        dist = 0
        for i in range(len(rota)):
            c1 = self.cidades[rota[i]]
            c2 = self.cidades[rota[(i + 1) % len(rota)]]
            dist += np.sqrt((c1[0]-c2[0])**2 + (c1[1]-c2[1])**2)
        return dist
    
    def fitness(self, rota):
        """Fitness = 1 / distância (maior = melhor)."""
        return 1 / self.distancia_total(rota)
    
    def inicializar_populacao(self):
        pop = []
        for _ in range(self.tamanho_pop):
            rota = list(range(self.n_cidades))
            random.shuffle(rota)
            pop.append(rota)
        return pop
    
    def selecao_torneio(self, populacao, fitness_pop, k=3):
        """Seleção por torneio: escolhe k candidatos, retorna o melhor."""
        candidatos = random.sample(range(len(populacao)), k)
        melhor = max(candidatos, key=lambda i: fitness_pop[i])
        return populacao[melhor][:]
    
    def crossover_pmx(self, pai1, pai2):
        """PMX Crossover para permutações."""
        size = len(pai1)
        filho = [-1] * size
        p1, p2 = sorted(random.sample(range(size), 2))
        filho[p1:p2+1] = pai1[p1:p2+1]
        for i in range(p1, p2+1):
            if pai2[i] not in filho:
                j = i
                while p1 <= j <= p2:
                    j = pai2.index(pai1[j])
                filho[j] = pai2[i]
        for i in range(size):
            if filho[i] == -1:
                filho[i] = pai2[i]
        return filho
    
    def mutacao_swap(self, rota):
        """Mutação por troca de dois genes."""
        rota = rota[:]
        if random.random() < self.taxa_mutacao:
            i, j = random.sample(range(len(rota)), 2)
            rota[i], rota[j] = rota[j], rota[i]
        return rota
    
    def run(self):
        populacao = self.inicializar_populacao()
        
        for geracao in range(self.geracoes):
            fitness_pop = [self.fitness(ind) for ind in populacao]
            
            # Estatísticas
            melhor_idx = np.argmax(fitness_pop)
            self.melhor_fitness_hist.append(1 / fitness_pop[melhor_idx])
            self.media_fitness_hist.append(np.mean([1/f for f in fitness_pop]))
            
            # Elitismo: preservar melhores
            idx_ordenados = np.argsort(fitness_pop)[::-1]
            nova_pop = [populacao[i][:] for i in idx_ordenados[:self.elitismo]]
            
            # Reprodução
            while len(nova_pop) < self.tamanho_pop:
                pai1 = self.selecao_torneio(populacao, fitness_pop)
                pai2 = self.selecao_torneio(populacao, fitness_pop)
                if random.random() < self.taxa_crossover:
                    filho = self.crossover_pmx(pai1, pai2)
                else:
                    filho = pai1[:]
                filho = self.mutacao_swap(filho)
                nova_pop.append(filho)
            
            populacao = nova_pop
        
        # Retornar melhor solução
        fitness_final = [self.fitness(ind) for ind in populacao]
        melhor = populacao[np.argmax(fitness_final)]
        return melhor, self.distancia_total(melhor)

# Criar problema TSP com 15 cidades aleatórias
cidades = [(random.uniform(0,100), random.uniform(0,100)) for _ in range(15)]

print("=== Algoritmo Genético - Problema do Caixeiro Viajante ===")
print(f"Número de cidades: {len(cidades)}")

ag = AlgoritmoGeneticoTSP(
    cidades=cidades, tamanho_populacao=50, geracoes=300,
    taxa_crossover=0.8, taxa_mutacao=0.05, elitismo=2
)
melhor_rota, melhor_dist = ag.run()

print(f"Melhor rota encontrada: {melhor_rota}")
print(f"Distância total: {melhor_dist:.2f}")

# Visualização
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gráfico 1: Evolução do fitness
axes[0].plot(ag.melhor_fitness_hist, 'b-', label='Melhor Distância', linewidth=2)
axes[0].plot(ag.media_fitness_hist, 'r--', label='Distância Média', alpha=0.7)
axes[0].set_title('Evolução do AG - TSP')
axes[0].set_xlabel('Geração')
axes[0].set_ylabel('Distância Total')
axes[0].legend()
axes[0].grid(True)

# Gráfico 2: Melhor rota
cidades_arr = np.array(cidades)
rota_fechada = melhor_rota + [melhor_rota[0]]
axes[1].plot(cidades_arr[rota_fechada, 0], cidades_arr[rota_fechada, 1], 'b-o', markersize=8)
for i, (x, y) in enumerate(cidades):
    axes[1].annotate(str(i), (x, y), textcoords="offset points", xytext=(5,5))
axes[1].set_title(f'Melhor Rota (Distância: {melhor_dist:.2f})')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].grid(True)

plt.tight_layout()
plt.savefig('ag_tsp.png', dpi=80, bbox_inches='tight')
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import random

def funcao_objetivo(x):
    """
    Função de teste: f(x) = x * sin(10*pi*x) + 2
    Tem múltiplos máximos locais no intervalo [-1, 2]
    Máximo global próximo de x = 1.85
    """
    return x * np.sin(10 * np.pi * x) + 2.0

class AGContinuo:
    """AG para otimização de função contínua com representação real."""
    
    def __init__(self, lim_inf=-1, lim_sup=2, pop_size=50,
                 geracoes=100, taxa_mut=0.1, sigma=0.1):
        self.lim_inf = lim_inf
        self.lim_sup = lim_sup
        self.pop_size = pop_size
        self.geracoes = geracoes
        self.taxa_mut = taxa_mut
        self.sigma = sigma
        self.historico = []
    
    def inicializar(self):
        return np.random.uniform(self.lim_inf, self.lim_sup, self.pop_size)
    
    def selecao_roleta(self, populacao, fitness):
        fitness_pos = fitness - fitness.min() + 1e-6
        probs = fitness_pos / fitness_pos.sum()
        idx = np.random.choice(len(populacao), p=probs)
        return populacao[idx]
    
    def crossover_aritmetico(self, pai1, pai2):
        alpha = np.random.random()
        return alpha * pai1 + (1 - alpha) * pai2
    
    def mutacao_gaussiana(self, individuo):
        if np.random.random() < self.taxa_mut:
            individuo += np.random.normal(0, self.sigma)
            individuo = np.clip(individuo, self.lim_inf, self.lim_sup)
        return individuo
    
    def run(self):
        pop = self.inicializar()
        
        for _ in range(self.geracoes):
            fitness = np.array([funcao_objetivo(x) for x in pop])
            self.historico.append(fitness.max())
            
            nova_pop = [pop[np.argmax(fitness)]]  # elitismo
            while len(nova_pop) < self.pop_size:
                pai1 = self.selecao_roleta(pop, fitness)
                pai2 = self.selecao_roleta(pop, fitness)
                filho = self.crossover_aritmetico(pai1, pai2)
                filho = self.mutacao_gaussiana(filho)
                nova_pop.append(filho)
            pop = np.array(nova_pop)
        
        fitness_final = np.array([funcao_objetivo(x) for x in pop])
        melhor_x = pop[np.argmax(fitness_final)]
        return melhor_x, funcao_objetivo(melhor_x)

print("=== AG para Otimização de Função Contínua ===")
ag_cont = AGContinuo(pop_size=50, geracoes=100, taxa_mut=0.1, sigma=0.05)
melhor_x, melhor_y = ag_cont.run()
print(f"Melhor x encontrado: {melhor_x:.6f}")
print(f"f(x) máximo: {melhor_y:.6f}")

# Visualização
x_vals = np.linspace(-1, 2, 1000)
y_vals = funcao_objetivo(x_vals)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x)')
axes[0].axvline(melhor_x, color='r', linestyle='--', label=f'Melhor x={melhor_x:.3f}')
axes[0].scatter([melhor_x], [melhor_y], color='red', zorder=5, s=100)
axes[0].set_title('Função Objetivo e Máximo Encontrado')
axes[0].set_xlabel('x')
axes[0].set_ylabel('f(x)')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(ag_cont.historico, 'g-', linewidth=2)
axes[1].set_title('Evolução do Melhor Fitness')
axes[1].set_xlabel('Geração')
axes[1].set_ylabel('f(x) Máximo')
axes[1].grid(True)

plt.tight_layout()
plt.savefig('ag_continuo.png', dpi=80, bbox_inches='tight')
plt.show()
````

## 6. Comparação: AG vs. Outros Algoritmos de Otimização

| Algoritmo | Tipo | Gradiente | Ótimo Global | Paralelismo | Complexidade | Quando Usar |
|-----------|------|-----------|-------------|-------------|-------------|-------------|
| **AG** | Metaheurística populacional | ❌ Não | Provável (não garantido) | ✅ Alto | O(G × N × F) | Problemas complexos, multimodais, discretos |
| **Simulated Annealing** | Metaheurística baseada em trajetória | ❌ Não | Provável (não garantido) | ❌ Baixo | O(I × F) | Espaços contínuos, uma solução por vez |
| **Busca Tabu** | Metaheurística baseada em memória | ❌ Não | Provável (não garantido) | ❌ Baixo | O(I × V) | Otimização combinatória com vizinhança definida |
| **Gradiente Descendente** | Exato / local | ✅ Sim | ❌ Mínimo local | ❌ Baixo | O(I × D) | Funções diferenciáveis, convexas |
| **Força Bruta** | Exato | ❌ Não | ✅ Garantido | ✅ Alto | O(N!) | Problemas pequenos com solução exata necessária |
| **PSO** | Metaheurística por enxame | ❌ Não | Provável (não garantido) | ✅ Médio | O(G × N × D) | Otimização contínua, rápida convergência |

> **Legenda:** G = gerações, N = tamanho da população, F = custo do fitness, I = iterações, V = tamanho da vizinhança, D = dimensões

### Quando escolher o AG?
- O espaço de busca é **grande e complexo** (combinatorial, multimodal)
- A função objetivo é **caixa-preta** (sem gradiente disponível)
- Diversidade de soluções é importante (exploração vs. explotação)
- Existem **restrições complexas** que dificultam métodos exatos

## 7. Aplicações dos Algoritmos Genéticos

### 🏗️ Engenharia
- **Design estrutural:** otimização de estruturas aeronáuticas e civil minimizando peso e maximizando resistência
- **Design de antenas:** NASA utilizou AGs para projetar antenas com geometrias não convencionais mas altamente eficientes
- **Design de circuitos:** síntese automática de circuitos analógicos e digitais

### 🚚 Logística
- **Roteamento de veículos (VRP):** otimizar rotas de entrega minimizando distância e respeitando janelas de tempo
- **Escalonamento de tarefas:** alocação de recursos em projetos e manufatura
- **Otimização de supply chain:** minimização de custos em cadeias de suprimentos complexas

### 🤖 Machine Learning
- **Seleção de features:** identificar subconjunto ótimo de variáveis para modelos preditivos
- **Otimização de hiperparâmetros:** buscar configurações ideais para redes neurais e outros modelos
- **Neuroevolução (NEAT):** evoluir simultaneamente pesos e topologia de redes neurais

- **Otimização de portfólio:** maximizar retorno esperado sujeito a restrições de risco (extensão de Markowitz)

### 🧬 Bioinformática

- **Estrutura de proteínas:** predição e design de dobramento de proteínas (protein folding)

### 🤖 Robótica

- **Planejamento de trajetórias:** otimizar caminhos em ambientes complexos

### 🎮 Jogos
- **Evolução de estratégias:** desenvolver agentes competitivos em jogos de tabuleiro e videogames

## 8. Vantagens e Desvantagens

### ✅ Vantagens

1. **Não requer informação de gradiente** (*black-box optimization*): pode ser aplicado a qualquer função objetivo, mesmo que descontínua, não diferenciável ou com ruído

2. **Robusto a mínimos/máximos locais**: a diversidade populacional permite explorar múltiplas regiões do espaço de busca simultaneamente, escapando de ótimos locais

3. **Paralelizável naturalmente**: cada indivíduo pode ser avaliado independentemente, facilitando implementação em GPU, clusters e computação distribuída

4. **Aplicável a qualquer tipo de representação**: binária, inteira, real, permutação, árvore — qualquer estrutura que possa ser codificada como cromossomo

5. **Otimização multiobjetivo**: extensões como NSGA-II e MOEA/D permitem otimizar múltiplos objetivos conflitantes encontrando a fronteira de Pareto

---

### ❌ Desvantagens

1. **Não garante o ótimo global**: sendo uma metaheurística, pode convergir para soluções subótimas, especialmente em espaços de alta dimensionalidade

2. **Muitos hiperparâmetros para ajustar**: tamanho da população, taxa de crossover, taxa de mutação, elitismo, critério de parada — todos influenciam significativamente o desempenho

3. **Convergência prematura**: se a diversidade da população cair muito cedo, o AG pode ficar preso em um ótimo local. Técnicas como reinicialização e niching mitigam esse problema

4. **Custo computacional cresce com o tamanho da população**: avaliar fitness de centenas ou milhares de indivíduos por geração pode ser proibitivo para funções custosas

5. **Difícil de escalar para espaços muito grandes**: em problemas com milhares de variáveis (alta dimensionalidade), o desempenho dos AGs clássicos degrada consideravelmente

---

### 📊 Resumo dos Parâmetros Típicos

| Parâmetro | Valor Típico | Impacto |
|-----------|-------------|--------|
| Tamanho da população | 50 – 200 | Diversidade vs. custo computacional |
| Taxa de crossover | 0.6 – 0.9 | Exploração de novas soluções |
| Taxa de mutação | 0.001 – 0.1 | Diversidade vs. estabilidade |
| Elitismo | 1 – 5% | Preservação das melhores soluções |
| Torneio (k) | 2 – 5 | Pressão seletiva |
