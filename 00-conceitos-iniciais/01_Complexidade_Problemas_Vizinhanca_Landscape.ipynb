## 2. Complexidade de Problemas

A **teoria da complexidade computacional** classifica os problemas de acordo com os recursos (tempo, espaço de memória) necessários para resolvê-los. Em IA, essa teoria é crucial porque muitos problemas de interesse prático são computacionalmente difíceis.

### 2.1 Classes P, NP, NP-Completo e NP-Difícil

#### Classe P — *Polynomial Time*

São os problemas que podem ser **resolvidos em tempo polinomial** por um algoritmo determinístico.

- Formalmente: existe um algoritmo com complexidade O(nᵏ) para alguma constante k.
- Exemplos: ordenação (O(n log n)), busca em lista ordenada (O(log n)), caminho mínimo (O(V² ou E log V)).

> **Intuição:** Problemas em P são considerados "tratáveis" — crescem de forma gerenciável com o tamanho da entrada.

#### Classe NP — *Nondeterministic Polynomial Time*

São os problemas para os quais uma **solução candidata pode ser verificada em tempo polinomial**, mas não necessariamente encontrada em tempo polinomial.

- Exemplos: Problema do Caixeiro Viajante (TSP), Satisfatibilidade Booleana (SAT), coloração de grafos.

> **Atenção:** NP ≠ "não-polinomial". NP significa que podemos **verificar** uma solução rapidamente, não que podemos **encontrá-la** rapidamente.

#### Classe NP-Completo

São os problemas que estão **simultaneamente em NP** e são tão difíceis quanto qualquer outro problema em NP (todo problema em NP pode ser reduzido a ele em tempo polinomial).

- Se qualquer problema NP-Completo pudesse ser resolvido em tempo polinomial, todos os problemas em NP também poderiam.
- Exemplos: SAT (o primeiro NP-completo provado — Teorema de Cook-Levin, 1971), TSP de decisão, problema da mochila (versão de decisão), coloração de grafos com k cores.

#### Classe NP-Difícil (*NP-Hard*)

São problemas **pelo menos tão difíceis quanto os NP-Completos**, mas que não precisam estar em NP (a solução pode não ser verificável em tempo polinomial).

- Exemplos: TSP de otimização (encontrar o menor caminho, não apenas decidir se existe um caminho menor que k), Halting Problem, otimização combinatória geral.

```
                    ┌─────────────────────────┐
                    │       NP-Difícil         │
                    │  ┌───────────────────┐  │
                    │  │   NP-Completo     │  │
                    │  │  ┌───────────┐   │  │
                    │  │  │    NP     │   │  │
                    │  │  │  ┌─────┐  │   │  │
                    │  │  │  │  P  │  │   │  │
                    │  │  │  └─────┘  │   │  │
                    │  │  └───────────┘   │  │
                    │  └───────────────────┘  │
                    └─────────────────────────┘
                    (Assumindo P ≠ NP — questão aberta)
```

#### A Questão P vs. NP

A pergunta **"P = NP?"** é o problema em aberto mais famoso da ciência da computação (e um dos Problemas do Milênio do Clay Mathematics Institute, com prêmio de US$ 1 milhão).

- A maioria dos pesquisadores acredita que **P ≠ NP**, mas isso nunca foi provado.
- Se P = NP: todos os problemas NP-Completos poderiam ser resolvidos eficientemente → revolução na criptografia, IA, biologia computacional.

````python
import numpy as np
import matplotlib.pyplot as plt
import math

# Comparação visual de crescimento de funções de complexidade
n = np.arange(1, 21)

complexidades = {
    'O(log n)':      np.log2(n),
    'O(n)':          n,
    'O(n log n)':    n * np.log2(n),
    'O(n²)':         n**2,
    'O(n³)':         n**3,
    'O(2ⁿ)':         2**n.astype(float),
    'O(n!)':         np.array([float(math.factorial(int(i))) if i <= 15 else float('inf') for i in n]),
}

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Escala linear
ax = axes[0]
for nome, vals in list(complexidades.items())[:5]:
    vals_clip = np.where(vals > 1e4, np.nan, vals)
    ax.plot(n, vals_clip, marker='o', markersize=3, label=nome)
ax.set_title('Crescimento de Complexidade (Escala Linear)', fontsize=12)
ax.set_xlabel('Tamanho da entrada (n)')
ax.set_ylabel('Operações')
ax.legend()
ax.set_ylim(0, 500)
ax.grid(True, alpha=0.3)

# Escala logarítmica — todas as complexidades
ax2 = axes[1]
for nome, vals in complexidades.items():
    vals_pos = np.where(vals <= 0, np.nan, vals)
    vals_clip = np.where(vals_pos > 1e15, np.nan, vals_pos)
    ax2.semilogy(n, vals_clip, marker='o', markersize=3, label=nome)
ax2.set_title('Crescimento de Complexidade (Escala Log)', fontsize=12)
ax2.set_xlabel('Tamanho da entrada (n)')
ax2.set_ylabel('Operações (escala log)')
ax2.legend(fontsize=8)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Tabela: operações para n = 5, 10, 20, 30
print(f"{'Complexidade':<12} | {'n=5':>12} | {'n=10':>12} | {'n=20':>15} | {'n=30':>20}")
print("-" * 80)
valores = {
    'O(log n)':  [np.log2(n) for n in [5,10,20,30]],
    'O(n)':      [5, 10, 20, 30],
    'O(n log n)':[5*np.log2(5), 10*np.log2(10), 20*np.log2(20), 30*np.log2(30)],
    'O(n²)':     [25, 100, 400, 900],
    'O(2ⁿ)':     [32, 1024, 1048576, 1073741824],
    'O(n!)':     [120, 3628800, '~2.4×10¹⁸', '~2.65×10³²'],
}
for nome, vals in valores.items():
    linha = f"{nome:<12} | "
    for v in vals:
        if isinstance(v, float):
            linha += f"{v:>12.2f} | "
        elif isinstance(v, int):
            linha += f"{v:>12,} | "
        else:
            linha += f"{v:>15} | "
    print(linha)
````

### 2.2 Explosão Combinatória

A **explosão combinatória** é o fenômeno pelo qual o número de possibilidades cresce de forma incontrolável com o tamanho do problema.

#### Exemplos Concretos

**Problema do Caixeiro Viajante (TSP):**
- 5 cidades → 12 rotas possíveis (4!/2)
- 10 cidades → 181.440 rotas
- 20 cidades → ~6×10¹⁶ rotas
- 50 cidades → ~3×10⁶⁴ rotas (mais que átomos no universo observável!)

**Jogo de Xadrez:**
- Número de posições possíveis: ~10⁴³ (número de Shannon)
- Número de partidas possíveis: ~10¹²⁰ (número de Chebyshev)
- Para comparar: o universo tem ~10⁸⁰ átomos

**Consequência prática para IA:**  
Para problemas NP-difíceis, **não existe algoritmo exato eficiente** (assumindo P ≠ NP). É preciso recorrer a:
- **Heurísticas:** regras práticas que guiam a busca sem garantia de ótimo global
- **Metaheurísticas:** estratégias gerais de busca (busca tabu, algoritmos genéticos, PSO)
- **Aproximações:** algoritmos que garantem uma solução dentro de um fator do ótimo
- **Casos especiais:** estruturas do problema que permitem solução eficiente

### 2.3 Problemas Clássicos de IA e sua Complexidade

| Problema | Classe | Melhor Algoritmo Exato | Abordagem Prática |
|----------|--------|----------------------|-------------------|
| Busca em grafo (caminho) | P | O(V + E) — BFS/DFS | — |
| Caminho mínimo (Dijkstra) | P | O(E log V) | — |
| Satisfatibilidade (SAT) | NP-Completo | Exponencial | Solvers DPLL, CDCL |
| Caixeiro Viajante (TSP) | NP-Difícil | O(n² 2ⁿ) | Heurísticas, AG, ACO |
| Coloração de Grafos | NP-Completo | Exponencial | Busca local, AG |
| Problema da Mochila | NP-Difícil | Pseudo-polinomial (PD) | AG, greedy |
| Planejamento de rotas | NP-Difícil | — | A*, heurísticas |
| Xadrez (solução perfeita) | EXPTIME-Completo | Impraticável | Alpha-Beta, MCTS |

> **Mensagem-chave:** A maioria dos problemas interessantes de otimização em IA é NP-difícil. Por isso, a área desenvolveu toda uma família de algoritmos inteligentes que *aceitam soluções boas (mas não necessariamente ótimas)* em troca de tempo de execução viável.

## 3. Espaço de Estados

O **espaço de estados** (ou espaço de busca) é a representação abstrata de todos os estados que um sistema pode assumir, conectados pelos operadores que transformam um estado em outro.

### Definição Formal

Um problema de busca é definido pela tupla **(S, s₀, A, T, G, c)** onde:

| Componente | Símbolo | Descrição |
|------------|---------|-----------|
| **Conjunto de estados** | S | Todos os estados possíveis do problema |
| **Estado inicial** | s₀ ∈ S | Estado de partida |
| **Ações/Operadores** | A(s) | Ações aplicáveis ao estado s |
| **Função de transição** | T(s, a) → s' | Resultado de aplicar ação a ao estado s |
| **Teste de objetivo** | G(s) → {True, False} | Verifica se s é solução |
| **Função de custo** | c(s, a, s') | Custo de aplicar ação a em s chegando a s' |

### Representações do Espaço de Estados

```
Grafo de Estados (abstrato):
                    s₀
                  /    \
               a₁       a₂
               /           \
             s₁             s₂
            /  \           /  \
          a₃   a₄        a₅   a₆
          /      \      /       \
        s₃       s₄  s₅        s₆ ← objetivo
```

### Tamanho do Espaço de Estados

O tamanho do espaço de estados cresce rapidamente:

**Exemplo — Quebra-cabeça de 8 peças (3×3):**
- 9! / 2 = 181.440 estados alcançáveis
- Resolvível com BFS em segundos

**Exemplo — Quebra-cabeça de 15 peças (4×4):**
- ~10¹³ estados possíveis
- Necessita de busca heurística (A*) para ser tratável

**Exemplo — Cubo de Rubik:**
- ~4,3 × 10¹⁹ estados
- Qualquer estado é solucionável em no máximo 20 movimentos ("God's Number")
- Descoberto em 2010 após anos de computação distribuída

### Tipos de Espaço de Estados

| Tipo | Características | Exemplos |
|------|----------------|----------|
| **Discreto finito** | Estados e ações enumeráveis | Xadrez, quebra-cabeças |
| **Discreto infinito** | Estados contáveis mas sem limite | Planejamento com recursos ilimitados |
| **Contínuo** | Estados são valores reais | Robótica, controle de sistemas |
| **Parcialmente observável** | Agente não conhece o estado completo | Navegação com sensores ruidosos |
| **Estocástico** | Ações têm resultados probabilísticos | Jogos com dado, IA em ambientes reais |

---

````python
# Representação do Espaço de Estados: Quebra-cabeça de 8 peças
# Estado = tupla representando a grade 3x3 (0 = espaço vazio)

from collections import deque

def obter_vizinhos_8puzzle(estado):
    """Retorna os estados vizinhos do quebra-cabeça de 8 peças."""
    grid = list(estado)
    pos_vazio = grid.index(0)
    linha, col = divmod(pos_vazio, 3)
    vizinhos = []
    movimentos = [(-1,0,'↑'), (1,0,'↓'), (0,-1,'←'), (0,1,'→')]
    for dl, dc, nome in movimentos:
        nl, nc = linha + dl, col + dc
        if 0 <= nl < 3 and 0 <= nc < 3:
            novo_grid = grid[:]
            nova_pos = nl * 3 + nc
            novo_grid[pos_vazio], novo_grid[nova_pos] = novo_grid[nova_pos], novo_grid[pos_vazio]
            vizinhos.append((tuple(novo_grid), nome))
    return vizinhos

def formatar_estado(estado):
    s = ""
    for i, v in enumerate(estado):
        s += str(v) if v != 0 else '_'
        s += ' '
        if (i + 1) % 3 == 0:
            s += '\n'
    return s

# Contagem de estados explorados com BFS
estado_inicial = (1, 2, 3, 4, 0, 5, 7, 8, 6)
estado_objetivo = (1, 2, 3, 4, 5, 6, 7, 8, 0)

print("Estado Inicial:")
print(formatar_estado(estado_inicial))
print("Estado Objetivo:")
print(formatar_estado(estado_objetivo))

# BFS com contagem de estados
fila = deque([(estado_inicial, [])])
visitados = {estado_inicial}
estados_explorados = 0

while fila:
    estado, caminho = fila.popleft()
    estados_explorados += 1
    if estado == estado_objetivo:
        print(f"Solução encontrada em {len(caminho)} movimentos!")
        print(f"Caminho: {' → '.join(caminho)}")
        print(f"Estados explorados: {estados_explorados:,}")
        break
    for vizinho, mov in obter_vizinhos_8puzzle(estado):
        if vizinho not in visitados:
            visitados.add(vizinho)
            fila.append((vizinho, caminho + [mov]))
````

## 4. Vizinhos e Tipos de Vizinhança

O conceito de **vizinhança** é central em IA: define quais soluções, estados ou posições estão "próximas" umas das outras. A escolha da vizinhança influencia diretamente o comportamento dos algoritmos.

> **Definição formal:** Dado um espaço S e uma solução s ∈ S, a **vizinhança** N(s) é um conjunto de soluções que podem ser alcançadas a partir de s por meio de uma única modificação (operador de vizinhança).

### 4.1 Vizinhança em Grafos

Em grafos, os **vizinhos** de um nó v são os nós diretamente conectados a v por uma aresta.

```
Grafo não-direcionado:           Vizinhos de C:
    A — B — D                    N(C) = {A, B, E}
    |   |                        
    C — E — F                    Vizinhos de B:
                                 N(B) = {A, D, E}
```

**Grau do nó:** O grau de v (deg(v)) é o número de vizinhos de v — equivalente ao tamanho de N(v).

**Grafo dirigido:** Em grafos direcionados, distinguimos:
- **Vizinhos de saída** (sucessores): nós para os quais v aponta
- **Vizinhos de entrada** (predecessores): nós que apontam para v

---

### 4.2 Vizinhança de Von Neumann

A **vizinhança de Von Neumann** (proposta por John von Neumann) em uma grade 2D considera apenas os **4 vizinhos ortogonais** (norte, sul, leste, oeste) de uma célula.

```
    . X .
    X C X      C = célula central
    . X .      X = vizinhos de Von Neumann
               . = NÃO são vizinhos
```

- **Raio r = 1:** 4 vizinhos
- **Raio r = 2:** 12 vizinhos
- **Raio geral r:** 2r(r+1) vizinhos
- **Uso:** Autômatos celulares, grade urbana (movimento apenas horizontal/vertical), simulações de difusão

---

### 4.3 Vizinhança de Moore

A **vizinhança de Moore** (proposta por Edward F. Moore) considera os **8 vizinhos** incluindo as diagonais.

```
    X X X
    X C X      C = célula central
    X X X      X = vizinhos de Moore (todos os 8)
```

- **Raio r = 1:** 8 vizinhos
- **Raio r = 2:** 24 vizinhos
- **Raio geral r:** (2r+1)² − 1 vizinhos
- **Uso:** Jogo da Vida de Conway, autômatos celulares com movimento diagonal, modelos populacionais

---

### 4.4 Vizinhança em Otimização Local

Em algoritmos de **busca local** e **metaheurísticas**, a vizinhança é definida em termos de **operadores de modificação** aplicados a uma solução:

| Problema | Solução | Operador de Vizinhança |
|----------|---------|----------------------|
| TSP | Permutação de cidades | 2-opt: troca 2 arestas |
| Problema da Mochila | Vetor binário | Flip: inverte um bit |
| Escalonamento | Sequência de tarefas | Swap: troca 2 tarefas |
| Coloração de grafos | Vetor de cores | Recolor: muda cor de 1 nó |

**Tamanho da vizinhança** determina o custo de cada iteração:
- Vizinhança pequena → iteração rápida, risco de estagnação
- Vizinhança grande → iteração lenta, mais informação por passo

---

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def vizinhanca_von_neumann(linha, col, raio=1, linhas=7, colunas=7):
    """Retorna células na vizinhança de Von Neumann de raio r."""
    vizinhos = []
    for dl in range(-raio, raio+1):
        for dc in range(-raio, raio+1):
            if abs(dl) + abs(dc) <= raio and (dl, dc) != (0, 0):
                nl, nc = linha + dl, col + dc
                if 0 <= nl < linhas and 0 <= nc < colunas:
                    vizinhos.append((nl, nc))
    return vizinhos

def vizinhanca_moore(linha, col, raio=1, linhas=7, colunas=7):
    """Retorna células na vizinhança de Moore de raio r."""
    vizinhos = []
    for dl in range(-raio, raio+1):
        for dc in range(-raio, raio+1):
            if (dl, dc) != (0, 0):
                nl, nc = linha + dl, col + dc
                if 0 <= nl < linhas and 0 <= nc < colunas:
                    vizinhos.append((nl, nc))
    return vizinhos

def desenhar_vizinhanca(ax, linha, col, vizinhos, titulo, linhas=7, colunas=7):
    grade = np.zeros((linhas, colunas))
    for (l, c) in vizinhos:
        grade[l, c] = 1
    grade[linha, col] = 2
    
    colors = ['white', '#4CAF50', '#F44336']
    from matplotlib.colors import ListedColormap
    cmap = ListedColormap(colors)
    
    ax.imshow(grade, cmap=cmap, vmin=0, vmax=2)
    ax.set_title(titulo, fontsize=11, fontweight='bold')
    ax.set_xticks(range(colunas))
    ax.set_yticks(range(linhas))
    ax.grid(True, color='gray', linewidth=0.5)
    for (l, c) in vizinhos:
        ax.text(c, l, '●', ha='center', va='center', fontsize=14, color='white')
    ax.text(col, linha, 'C', ha='center', va='center', fontsize=13, 
            fontweight='bold', color='white')
    ax.set_xticklabels([])
    ax.set_yticklabels([])

centro = (3, 3)
fig, axes = plt.subplots(2, 3, figsize=(15, 9))

configs = [
    ('Von Neumann r=1', vizinhanca_von_neumann(*centro, raio=1), 0, 0),
    ('Von Neumann r=2', vizinhanca_von_neumann(*centro, raio=2), 0, 1),
    ('Moore r=1',       vizinhanca_moore(*centro, raio=1),       0, 2),
    ('Moore r=2',       vizinhanca_moore(*centro, raio=2),       1, 0),
]

for titulo, vizinhos, row, col_idx in configs:
    ax = axes[row][col_idx]
    desenhar_vizinhanca(ax, *centro, vizinhos, f'{titulo}\n({len(vizinhos)} vizinhos)')

# Tabela comparativa
ax_tab = axes[1][1]
ax_tab.axis('off')
dados = [
    ['Raio', 'Von Neumann', 'Moore'],
    ['1', '4', '8'],
    ['2', '12', '24'],
    ['3', '24', '48'],
    ['r', '2r(r+1)', '(2r+1)²−1'],
]
tabela = ax_tab.table(cellText=dados[1:], colLabels=dados[0],
                       loc='center', cellLoc='center')
tabela.auto_set_font_size(False)
tabela.set_fontsize(11)
tabela.scale(1.5, 2.0)
ax_tab.set_title('Comparação de Vizinhanças', fontsize=11, fontweight='bold')

# Vizinhança 2-opt para TSP
ax_tsp = axes[1][2]
cidades = np.array([[1,1],[2,3],[4,2],[3,0],[0,2]])
rota = [0,1,2,3,4]
xs = [cidades[i,0] for i in rota] + [cidades[rota[0],0]]
ys = [cidades[i,1] for i in rota] + [cidades[rota[0],1]]
ax_tsp.plot(xs, ys, 'b-o', linewidth=2, markersize=8, label='Rota atual')
# Mostra a troca 2-opt (inverter segmento 1-3)
rota_vizinha = [0,3,2,1,4]
xv = [cidades[i,0] for i in rota_vizinha] + [cidades[rota_vizinha[0],0]]
yv = [cidades[i,1] for i in rota_vizinha] + [cidades[rota_vizinha[0],1]]
ax_tsp.plot(xv, yv, 'r--o', linewidth=2, markersize=8, alpha=0.7, label='Vizinho 2-opt')
for i, (x, y) in enumerate(cidades):
    ax_tsp.annotate(f'C{i}', (x, y), textcoords='offset points', xytext=(8,5))
ax_tsp.set_title('Vizinhança 2-opt (TSP)', fontsize=11, fontweight='bold')
ax_tsp.legend(fontsize=8)
ax_tsp.grid(True, alpha=0.3)
ax_tsp.set_xlim(-0.5, 5)

plt.tight_layout()
plt.show()

print("Von Neumann r=1:", vizinhanca_von_neumann(*centro, raio=1))
print(f"Total: {len(vizinhanca_von_neumann(*centro, raio=1))} vizinhos")
print()
print("Moore r=1:", vizinhanca_moore(*centro, raio=1))
print(f"Total: {len(vizinhanca_moore(*centro, raio=1))} vizinhos")
````

## 5. Landscape — Paisagem de Aptidão

O conceito de **landscape** (paisagem de aptidão ou fitness landscape) foi proposto por Sewall Wright em 1932, no contexto da genética evolutiva, e tornou-se uma metáfora poderosa para entender qualquer problema de otimização.

### 5.1 Conceito e Definição

> **Landscape de Aptidão:** Um espaço de busca equipado com uma função de aptidão (fitness) que associa a cada solução um valor numérico — formando uma "superfície" topográfica onde picos representam soluções ótimas.

Formalmente, um landscape é a tripla **(S, N, f)** onde:
- **S** = espaço de soluções
- **N : S → 2^S** = função de vizinhança
- **f : S → ℝ** = função de aptidão (ou função objetivo)

**Nota importante:** O landscape depende tanto de `f` quanto de `N`. A mesma função objetivo pode ter landscapes completamente diferentes com vizinhanças distintas.

---

### 5.2 Anatomia de um Landscape

```
     Aptidão
        ↑
  ____  |      Ótimo    _____
 /    \ |      Global  /     \
/  Ótimo\_____/       /  Platô \____
  Local        Vale    Crista
                                  → Espaço de Soluções
```

| Característica | Descrição | Implicação para Algoritmos |
|----------------|-----------|--------------------------|
| **Ótimo Global** | Solução com maior aptidão em todo S | Meta final — difícil de garantir em NP-difícil |
| **Ótimo Local** | Melhor solução na vizinhança | Algoritmos gulosos ficam presos aqui |
| **Vale** | Região de baixa aptidão entre ótimos | Barreira que separa bacias de atração |
| **Platô** | Região plana com mesma aptidão | Perda de direção no gradiente |
| **Crista** | Caminho estreito de alta aptidão | Difícil de percorrer por algoritmos aleatórios |
| **Bacia de Atração** | Conjunto de pontos que convergem ao mesmo ótimo local | Determina "zona de influência" de cada ótimo |

---

### 5.3 Tipos de Landscape

#### Landscape Unimodal (Suave)
- Apenas **um ótimo global**, sem ótimos locais
- Algoritmos de subida de encosta (*hill climbing*) são suficientes
- Exemplo: minimização de f(x) = x² (parabolóide)

#### Landscape Multimodal (Rugoso)
- **Múltiplos ótimos locais** — mais realista em problemas práticos
- Requer estratégias de diversificação para escapar de ótimos locais
- Exemplo: função de Rastrigin, TSP com muitas cidades

#### Landscape Neutro (*Flat Landscape*)
- Grandes regiões com **aptidão idêntica** (platôs extensos)
- Deriva neutra: o algoritmo anda sem direção clara
- Comum em espaços de permutação com funções objetivo pouco discriminativas

#### Landscape Enganoso (*Deceptive Landscape*)
- O gradiente local aponta **na direção errada** — afastando-se do ótimo global
- Comum em funções com "falsas trilhas"
- Representa o maior desafio para algoritmos baseados em gradiente local

```
Landscape Rugoso:        Landscape Enganoso:
  *                         *
 * *  *                    * *     Ótimo global
*   **  *   G              *   *  aqui, mas o
        * ** (G = global)         gradiente aponta
  ótimos locais                   para longe!
```

---

### 5.4 Ótimo Local vs. Ótimo Global

**Ótimo Local:** Uma solução s* é um ótimo local se:
```
f(s*) ≥ f(s)  para todo s ∈ N(s*)
```

**Ótimo Global:** Uma solução s* é o ótimo global se:
```
f(s*) ≥ f(s)  para todo s ∈ S
```

| Algoritmo | Garante Ótimo Global? | Como lida com ótimos locais? |
|-----------|----------------------|------------------------------|
| Hill Climbing | ❌ | Fica preso |
| Simulated Annealing | ✅* (assintoticamente) | Aceita pioras com probabilidade e^(−Δ/T) |
| Busca Tabu | ❌ (mas escapa) | Lista tabu proíbe revisitas |
| Algoritmos Genéticos | ❌ (probabilístico) | Crossover + mutação exploram diversas regiões |
| Branch and Bound | ✅ | Poda sistemática do espaço |

> \* Simulated Annealing garante ótimo global apenas no limite teórico com resfriamento infinitamente lento.

---

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definição de funções de landscape clássicas
def rastrigin(X, Y):
    """Landscape multimodal — muitos ótimos locais. Ótimo global em (0,0)."""
    A = 10
    return A*2 + (X**2 - A*np.cos(2*np.pi*X)) + (Y**2 - A*np.cos(2*np.pi*Y))

def paraboloide(X, Y):
    """Landscape unimodal (suave). Ótimo global em (0,0)."""
    return X**2 + Y**2

def ackley(X, Y):
    """Landscape com um ótimo global cercado de muitos ótimos locais."""
    return (-20 * np.exp(-0.2 * np.sqrt(0.5*(X**2 + Y**2)))
            - np.exp(0.5*(np.cos(2*np.pi*X) + np.cos(2*np.pi*Y)))
            + np.e + 20)

x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

fig, axes = plt.subplots(1, 3, figsize=(18, 5), subplot_kw={'projection': '3d'})

landscapes = [
    (paraboloide(X, Y), 'Landscape Unimodal\n(Parabolóide)', '#2196F3'),
    (rastrigin(X, Y),   'Landscape Multimodal\n(Rastrigin)',  '#F44336'),
    (ackley(X, Y),      'Landscape com Ótimo\nGlobal + Locais (Ackley)', '#4CAF50'),
]

for ax, (Z, titulo, cor) in zip(axes, landscapes):
    surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.85, linewidth=0)
    ax.set_title(titulo, fontsize=10, fontweight='bold')
    ax.set_xlabel('x₁')
    ax.set_ylabel('x₂')
    ax.set_zlabel('f(x₁,x₂)')
    ax.view_init(elev=30, azim=45)

plt.tight_layout()
plt.show()

# Visualização 2D: demonstrando ótimo local vs global
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1D multimodal
x1d = np.linspace(-5, 5, 1000)
f1d = np.sin(2*x1d) + np.sin(3*x1d) + 0.5*np.cos(x1d) - 0.1*x1d**2 + 2

ax = axes[0]
ax.plot(x1d, f1d, 'b-', linewidth=2)
otimo_global_idx = np.argmax(f1d)
otimos_locais = []
for i in range(1, len(f1d)-1):
    if f1d[i] > f1d[i-1] and f1d[i] > f1d[i+1]:
        otimos_locais.append(i)

# Ótimo global
ax.plot(x1d[otimo_global_idx], f1d[otimo_global_idx], 'r*', markersize=18,
        label=f'Ótimo Global: f={f1d[otimo_global_idx]:.2f}', zorder=5)
# Ótimos locais
for idx in otimos_locais[:4]:
    if idx != otimo_global_idx and abs(idx - otimo_global_idx) > 10:
        ax.plot(x1d[idx], f1d[idx], 'go', markersize=10, zorder=4)

ax.plot([], [], 'go', markersize=10, label='Ótimos Locais')
ax.fill_between(x1d, f1d, min(f1d), alpha=0.1, color='blue')
ax.axhline(y=f1d[otimo_global_idx], color='red', linestyle='--', alpha=0.4)
ax.set_title('Landscape 1D: Ótimo Global vs. Ótimos Locais', fontsize=11)
ax.set_xlabel('Espaço de Soluções')
ax.set_ylabel('Aptidão f(x)')
ax.legend()
ax.grid(True, alpha=0.3)

# Trajetória de Hill Climbing preso em ótimo local
ax2 = axes[1]
ax2.plot(x1d, f1d, 'b-', linewidth=2, label='Landscape')

# Simula Hill Climbing partindo de x=-2
x_atual = -2.0
trajetoria = [x_atual]
for _ in range(20):
    melhor = x_atual
    for delta in [-0.3, -0.1, 0.1, 0.3]:
        x_novo = x_atual + delta
        if -5 <= x_novo <= 5:
            idx_novo = int((x_novo + 5) / 10 * 999)
            idx_atual = int((x_atual + 5) / 10 * 999)
            if f1d[idx_novo] > f1d[idx_atual]:
                melhor = x_novo
    if melhor == x_atual:
        break
    x_atual = melhor
    trajetoria.append(x_atual)

traj_y = [f1d[int((xi + 5)/10*999)] for xi in trajetoria]
ax2.plot(trajetoria, traj_y, 'r-o', linewidth=2, markersize=7,
         label=f'Hill Climbing (preso em ótimo local!)', zorder=4)
ax2.plot(trajetoria[0], traj_y[0], 'ms', markersize=12, label='Início', zorder=5)
ax2.plot(trajetoria[-1], traj_y[-1], 'r*', markersize=15, label='Ótimo Local Encontrado', zorder=5)
ax2.plot(x1d[otimo_global_idx], f1d[otimo_global_idx], 'g^', markersize=14,
         label='Ótimo Global (não encontrado)', zorder=5)
ax2.set_title('Hill Climbing Preso em Ótimo Local', fontsize=11)
ax2.set_xlabel('Espaço de Soluções')
ax2.set_ylabel('Aptidão f(x)')
ax2.legend(fontsize=8)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

## 6. Path (Caminho) e Link

Algoritmos de busca exploram o espaço de estados navegando por **caminhos** — sequências de estados conectados por **links** (arestas/operadores). Compreender esses conceitos é fundamental para analisar e projetar algoritmos.

### 6.1 Conceitos Fundamentais de Grafos

**Grafo G = (V, E):**
- **Vértices (V):** Estados, posições, nós
- **Arestas/Links (E):** Transições, conexões, operadores

**Propriedades das Arestas:**
- **Peso/Custo:** Valor numérico associado (distância, tempo, energia)
- **Direção:** Grafo dirigido (dígrafo) vs. não-dirigido
- **Capacidade:** Em redes de fluxo

### 6.2 Tipos de Caminho

**Caminho (Path):** Sequência de vértices v₁, v₂, ..., vₖ tal que existe aresta (vᵢ, vᵢ₊₁) para todo i.

| Tipo | Definição | Exemplo de Uso |
|------|-----------|----------------|
| **Caminho Simples** | Nenhum vértice se repete | BFS, DFS |
| **Trilha** | Nenhuma aresta se repete | Problema de Euler |
| **Ciclo** | Caminho que retorna ao início | Detecção de ciclos, TSP |
| **Caminho Ótimo** | Menor custo total | Dijkstra, A* |
| **Caminho Hamiltoniano** | Visita cada vértice exatamente uma vez | TSP, planejamento |
| **Caminho Euleriano** | Percorre cada aresta exatamente uma vez | Problema das Pontes de Königsberg |

**Comprimento de um Caminho:**
- **Não-ponderado:** Número de arestas
- **Ponderado:** Soma dos pesos das arestas

```
Grafo ponderado:               Caminhos de A a D:
                               
    A —2— B —3— D              A→B→D: custo = 2+3 = 5
    |         /                A→C→D: custo = 4+1 = 5
    4        1                 A→B→C→D: custo = 2+1+1 = 4 ← ótimo
    |       /                  
    C —1— /
```

### 6.3 Custo de Caminho e Funções Objetivo

Em problemas de otimização, associamos um **custo** a cada caminho:

```
Custo(π) = Σ c(sᵢ, aᵢ, sᵢ₊₁)
```

Onde:
- π = caminho (s₀, a₀, s₁, a₁, ..., sₙ)
- c(sᵢ, aᵢ, sᵢ₊₁) = custo do passo i

**Funções objetivo comuns:**
- **Minimizar distância total** (TSP, roteamento)
- **Minimizar tempo** (escalonamento)
- **Maximizar aptidão** (algoritmos genéticos)
- **Minimizar custo + heurística** (A*: f(n) = g(n) + h(n))

### 6.4 Path Relinking (Reencadeamento de Caminhos)

O **Path Relinking** é uma técnica de busca proposta por Glover (1996) que intensifica a busca **explorando trajetórias entre soluções de elite**.

**Ideia central:**
1. Mantém um conjunto de **soluções de elite** (altamente qualificadas)
2. Dado um par de soluções elite (s₀, sₜ), gera um **caminho** no espaço de soluções de s₀ a sₜ
3. Avalia todas as soluções intermediárias no caminho
4. Adiciona as melhores ao conjunto de elite

```
Espaço de Soluções:

    s₀ ——→ s₁ ——→ s₂ ——→ s₃ ——→ sₜ
    elite              melhor    elite
    (origem)          intermediário   (destino)
                      ← adicionado ao elite
```

**Diferença de Path Relinking para outras buscas:**
| Técnica | Estrutura de Guia | Objetivo |
|---------|------------------|----------|
| Hill Climbing | Gradiente local | Subir na vizinhança |
| Busca Tabu | Lista de tabus | Evitar ciclos |
| Path Relinking | Par de soluções elite | Explorar entre soluções boas |

**Uso prático:** Combinado com GRASP (Greedy Randomized Adaptive Search Procedure), Busca Tabu e Algoritmos Genéticos para intensificação pós-otimização.

---

````python
import heapq
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

# Grafo ponderado de exemplo (cidades fictícias)
grafo = {
    'A': {'B': 2, 'C': 4, 'D': 8},
    'B': {'A': 2, 'C': 1, 'E': 5},
    'C': {'A': 4, 'B': 1, 'D': 1, 'F': 3},
    'D': {'A': 8, 'C': 1, 'G': 2},
    'E': {'B': 5, 'F': 2, 'G': 6},
    'F': {'C': 3, 'E': 2, 'G': 1},
    'G': {'D': 2, 'E': 6, 'F': 1},
}

def dijkstra(grafo, inicio, fim):
    """Algoritmo de Dijkstra para caminho mínimo."""
    dist = {v: float('inf') for v in grafo}
    dist[inicio] = 0
    anterior = {v: None for v in grafo}
    fila = [(0, inicio)]
    
    while fila:
        d_atual, u = heapq.heappop(fila)
        if d_atual > dist[u]:
            continue
        for v, peso in grafo[u].items():
            nova_dist = dist[u] + peso
            if nova_dist < dist[v]:
                dist[v] = nova_dist
                anterior[v] = u
                heapq.heappush(fila, (nova_dist, v))
    
    # Reconstrói caminho
    caminho = []
    atual = fim
    while atual is not None:
        caminho.append(atual)
        atual = anterior[atual]
    caminho.reverse()
    return caminho, dist[fim]

# Encontra caminho mínimo de A a G
caminho, custo = dijkstra(grafo, 'A', 'G')
print(f"Caminho mínimo de A a G: {' → '.join(caminho)}")
print(f"Custo total: {custo}")

# Todos os caminhos simples de A a G (para comparação)
def todos_caminhos(grafo, inicio, fim, caminho_atual=None, visitados=None):
    if caminho_atual is None:
        caminho_atual = [inicio]
    if visitados is None:
        visitados = {inicio}
    if inicio == fim:
        return [caminho_atual[:]]
    resultados = []
    for vizinho in grafo[inicio]:
        if vizinho not in visitados:
            visitados.add(vizinho)
            caminho_atual.append(vizinho)
            resultados.extend(todos_caminhos(grafo, vizinho, fim, caminho_atual, visitados))
            caminho_atual.pop()
            visitados.remove(vizinho)
    return resultados

caminhos = todos_caminhos(grafo, 'A', 'G')
custos = []
for c in caminhos:
    custo_c = sum(grafo[c[i]][c[i+1]] for i in range(len(c)-1))
    custos.append((custo_c, c))
custos.sort()

print(f"\nTodos os {len(caminhos)} caminhos simples de A a G (ordenados por custo):")
for custo_c, c in custos[:6]:
    print(f"  {'→'.join(c):<25} custo = {custo_c}")
if len(custos) > 6:
    print(f"  ... ({len(custos)-6} outros)")

# Visualização
fig, ax = plt.subplots(1, 1, figsize=(10, 7))
G = nx.Graph()
pos = {'A': (0,2), 'B': (1,3), 'C': (2,2), 'D': (3,3), 
       'E': (2,4), 'F': (3,2), 'G': (4,3)}

for u, vizinhos in grafo.items():
    for v, peso in vizinhos.items():
        if u < v:
            G.add_edge(u, v, weight=peso)

arestas_caminho = [(caminho[i], caminho[i+1]) for i in range(len(caminho)-1)]
arestas_outras = [e for e in G.edges() if e not in arestas_caminho and 
                   (e[1],e[0]) not in arestas_caminho]

nx.draw_networkx_nodes(G, pos, ax=ax, node_color='lightblue', node_size=700)
nx.draw_networkx_nodes(G, pos, nodelist=[caminho[0]], ax=ax, 
                        node_color='green', node_size=700)
nx.draw_networkx_nodes(G, pos, nodelist=[caminho[-1]], ax=ax, 
                        node_color='red', node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=arestas_outras, ax=ax, 
                        edge_color='gray', width=1.5)
nx.draw_networkx_edges(G, pos, edgelist=arestas_caminho, ax=ax,
                        edge_color='red', width=4, arrows=True)
nx.draw_networkx_labels(G, pos, ax=ax, font_size=12, font_weight='bold')
labels_peso = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, labels_peso, ax=ax, font_size=10)

ax.set_title(f'Caminho Mínimo: {" → ".join(caminho)} (custo = {custos[0][0]})', 
             fontsize=12, fontweight='bold')
ax.axis('off')
plt.tight_layout()
plt.show()
````

## 7. Exploração vs. Explotação

Um dos dilemas mais fundamentais em IA — especialmente em aprendizado por reforço, metaheurísticas e algoritmos evolutivos — é o **trade-off exploração-explotação** (*exploration-exploitation trade-off*).

### Definições

**Exploração (*Exploration*):**
- Buscar em regiões **não visitadas** do espaço de soluções
- Objetivo: descobrir novas soluções potencialmente melhores
- Risco: gastar tempo em regiões ruins

**Explotação (*Exploitation*):**
- Aprofundar a busca em regiões **já conhecidas como boas**
- Objetivo: refinar soluções já encontradas
- Risco: convergir prematuramente para um ótimo local

### O Dilema

> *"Você deve continuar explorando novos restaurantes ou ir sempre ao seu favorito?"*

| Estratégia Extrema | Consequência |
|-------------------|--------------|
| Apenas exploração | Nunca converge — anda aleatoriamente pelo espaço |
| Apenas explotação | Converge prematuramente para o primeiro ótimo local |
| **Equilíbrio dinâmico** | **Explora no início, explota no final — comportamento ideal** |

### Como Diferentes Algoritmos Controlam o Trade-off

| Algoritmo | Mecanismo de Controle |
|-----------|----------------------|
| **Simulated Annealing** | Temperatura T: alta T → exploração; baixa T → explotação |
| **Algoritmos Genéticos** | Taxa de mutação: alta → exploração; elitismo → explotação |
| **Busca Tabu** | Tamanho da lista tabu e diversificação |
| **PSO** | Peso de inércia w e coeficientes cognitivo/social |
| **ε-greedy (RL)** | ε alto → exploração; ε baixo → explotação |
| **UCB (RL)** | Upper Confidence Bound equilibra automaticamente |

### Convergência Prematura

**Convergência prematura** ocorre quando um algoritmo estagna em um ótimo local antes de explorar adequadamente o espaço de soluções. É a consequência direta de explotação excessiva.

**Sintomas:**
- Diversidade da população cai rapidamente (em AGs)
- Todas as partículas convergem para o mesmo ponto (PSO)
- Busca fica em loop em mesma região (busca local)

**Estratégias de prevenção:**
- **Reinicialização:** Reiniciar parcialmente a busca em novas regiões
- **Perturbação:** Aplicar mudanças aleatórias grandes (*kick moves*)
- **Manutenção de diversidade:** Pressão de diversidade explícita
- **Multi-start:** Executar múltiplas buscas independentes

---

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

def landscape_1d(x):
    """Landscape multimodal para demonstração."""
    return (np.sin(2*x) + np.sin(3.5*x) + np.cos(1.5*x) 
            - 0.05*x**2 + np.sin(0.5*x + 1))

x = np.linspace(-5, 5, 1000)
f = landscape_1d(x)

def simulated_annealing(f_func, x_range=(-5,5), T_inicial=5.0, T_final=0.01, 
                         alpha=0.98, n_iter=500, semente=None):
    """Simulated Annealing com rastreamento de temperatura."""
    if semente:
        np.random.seed(semente)
    x_atual = np.random.uniform(*x_range)
    f_atual = f_func(x_atual)
    melhor_x, melhor_f = x_atual, f_atual
    T = T_inicial
    historico = [(x_atual, f_atual, T)]
    
    for _ in range(n_iter):
        # Gera vizinho
        x_novo = x_atual + np.random.normal(0, T * 0.5)
        x_novo = np.clip(x_novo, *x_range)
        f_novo = f_func(x_novo)
        delta = f_novo - f_atual
        
        # Aceita melhora ou aceita piora com probabilidade exp(Δ/T)
        if delta > 0 or np.random.random() < np.exp(delta / T):
            x_atual, f_atual = x_novo, f_novo
            if f_atual > melhor_f:
                melhor_x, melhor_f = x_atual, f_atual
        
        T = max(T * alpha, T_final)
        historico.append((x_atual, f_atual, T))
    
    return melhor_x, melhor_f, historico

melhor_x, melhor_f, historico = simulated_annealing(
    landscape_1d, semente=42, n_iter=800
)

x_hist = [h[0] for h in historico]
f_hist = [h[1] for h in historico]
T_hist = [h[2] for h in historico]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Landscape com trajetória
ax1 = axes[0][0]
ax1.plot(x, f, 'b-', linewidth=2, label='Landscape', zorder=1)
scatter = ax1.scatter(x_hist[::10], f_hist[::10], 
                       c=range(len(x_hist[::10])), cmap='plasma',
                       s=30, zorder=3, alpha=0.7, label='Trajetória SA')
ax1.plot(melhor_x, melhor_f, 'r*', markersize=20, label=f'Melhor (f={melhor_f:.3f})', zorder=4)
plt.colorbar(scatter, ax=ax1, label='Iteração')
ax1.set_title('Trajetória do Simulated Annealing', fontsize=11)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.legend(fontsize=8)
ax1.grid(True, alpha=0.3)

# Evolução da temperatura
ax2 = axes[0][1]
ax2.plot(T_hist, 'r-', linewidth=2)
ax2.fill_between(range(len(T_hist)), T_hist, alpha=0.3, color='red')
ax2.axhline(y=1.0, color='gray', linestyle='--', label='T=1 (transição exp./expl.)')
ax2.set_title('Temperatura ao Longo das Iterações', fontsize=11)
ax2.set_xlabel('Iteração')
ax2.set_ylabel('Temperatura T')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_yscale('log')

# Comparação: alta exploração vs alta explotação
_, _, hist_expl = simulated_annealing(landscape_1d, T_inicial=10.0, alpha=0.999, 
                                       n_iter=800, semente=10)  # Muita exploração
_, _, hist_expt = simulated_annealing(landscape_1d, T_inicial=0.1, alpha=0.99, 
                                       n_iter=800, semente=10)  # Muita explotação

ax3 = axes[1][0]
ax3.plot(x, f, 'b-', linewidth=2, label='Landscape')
x_e1 = [h[0] for h in hist_expl[::5]]
f_e1 = [h[1] for h in hist_expl[::5]]
x_e2 = [h[0] for h in hist_expt[::5]]
f_e2 = [h[1] for h in hist_expt[::5]]
ax3.scatter(x_e1, f_e1, c='green', s=20, alpha=0.5, label='Alta Exploração (T grande)', zorder=3)
ax3.scatter(x_e2, f_e2, c='orange', s=20, alpha=0.5, label='Alta Explotação (T pequeno)', zorder=3)
ax3.set_title('Exploração vs. Explotação', fontsize=11)
ax3.set_xlabel('x')
ax3.set_ylabel('f(x)')
ax3.legend(fontsize=8)
ax3.grid(True, alpha=0.3)

# Evolução do melhor valor encontrado
f_melhor_SA = [max(f_hist[:i+1]) for i in range(len(f_hist))]
ax4 = axes[1][1]
ax4.plot(f_melhor_SA, 'g-', linewidth=2, label='SA (equilíbrio)')
ax4.plot([max([h[1] for h in hist_expl[:i+1]]) for i in range(len(hist_expl))],
         'b--', linewidth=2, alpha=0.8, label='Alta Exploração')
ax4.plot([max([h[1] for h in hist_expt[:i+1]]) for i in range(len(hist_expt))],
         'r:', linewidth=2, label='Alta Explotação')
ax4.set_title('Evolução do Melhor Valor', fontsize=11)
ax4.set_xlabel('Iteração')
ax4.set_ylabel('Melhor f(x) encontrado')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.suptitle('Exploração vs. Explotação no Simulated Annealing', 
             fontsize=13, fontweight='bold', y=1.01)
plt.show()

print(f"Resultado SA: x = {melhor_x:.4f}, f(x) = {melhor_f:.4f}")
print(f"Ótimo real: x ≈ {x[np.argmax(f)]:.4f}, f(x) ≈ {np.max(f):.4f}")
````

## 8. Metaheurísticas — Visão Geral

**Metaheurísticas** são estratégias de busca de **alto nível** que guiam heurísticas subordinadas para explorar eficientemente o espaço de soluções de problemas difíceis. São os algoritmos "inteligentes" que lidam com a explosão combinatória.

> *"Meta"* vem do grego e significa "além" ou "acima de". Metaheurísticas estão "acima" de heurísticas específicas — são princípios gerais que podem ser aplicados a classes amplas de problemas.

### Taxonomia das Metaheurísticas

```
                        Metaheurísticas
                       /               \
               Baseadas em          Baseadas em
               Trajetória           População
               /    |    \           /        \
          Hill  Busca Simulated  Algoritmos  Enxame de
         Climb  Tabu  Annealing  Genéticos  Partículas
                              \             /
                           Computação
                           Evolutiva
```

### Principais Metaheurísticas

| Metaheurística | Inspiração | Tipo | Pontos Fortes |
|----------------|-----------|------|---------------|
| **Hill Climbing** | Subida de encosta | Trajetória | Simples, rápido |
| **Simulated Annealing** | Resfriamento de metais | Trajetória | Escapa ótimos locais |
| **Busca Tabu** | Memória de curto prazo | Trajetória | Evita ciclos, intensificação |
| **GRASP** | Construção + melhoria | Trajetória | Combina guloso + aleatório |
| **Algoritmos Genéticos** | Evolução darwiniana | Populacional | Diversidade, paralelismo |
| **Evolução Diferencial** | Variação genética | Populacional | Contínuo, robusto |
| **PSO** | Enxame de pássaros | Populacional | Simples, eficaz em contínuo |
| **Colônia de Formigas (ACO)** | Feromônio | Populacional | Grafos, TSP |
| **Algoritmos Meméticos** | AG + busca local | Híbrida | Melhor de dois mundos |

### Critérios de Parada

Um aspecto crítico de toda metaheurística é **quando parar**. Critérios comuns:

1. **Número máximo de iterações:** Mais simples, mas pode ser ineficiente
2. **Sem melhoria por k iterações:** Detecta estagnação
3. **Tempo de CPU limite:** Controle prático
4. **Qualidade da solução:** Pára ao atingir valor-alvo conhecido
5. **Convergência da população:** Para AGs, quando diversidade cai muito

### Intensificação e Diversificação

Dois princípios complementares que toda metaheurística precisa equilibrar:

**Intensificação (*Intensification*):**
- Concentrar a busca em regiões promissoras
- Explorar em profundidade vizinhanças de soluções boas
- Análoga à explotação

**Diversificação (*Diversification*):**
- Garantir a exploração de regiões diversas do espaço de busca
- Evitar convergência prematura
- Análoga à exploração

> **Regra prática:** Uma boa metaheurística equilibra intensificação e diversificação ao longo do tempo — iniciando com mais diversificação e finalizando com mais intensificação.

### Ciclo de Projeto de um Algoritmo de Busca

Ao projetar ou selecionar um algoritmo para um problema de IA, o engenheiro deve responder:

1. **Qual é a complexidade do problema?**  
   → P: use algoritmo exato. NP-difícil: use metaheurística.

2. **Como é o espaço de estados?**  
   → Discreto? Contínuo? Qual o tamanho?

3. **Qual vizinhança usar?**  
   → Vizinhança pequena (rápida por iteração) ou grande (mais informativa)?

4. **Como é o landscape com esta vizinhança?**  
   → Rugoso? Neutro? Enganoso? → Escolhe estratégia de exploração.

5. **Como representar e explorar caminhos?**  
   → Preservar caminho completo (para reconstrição) ou apenas estado atual?

6. **Qual o equilíbrio exploração-explotação?**  
   → Quanto tempo gastar explorando vs. refinando?

### Tabela de Decisão: Quando Usar Cada Abordagem

| Característica do Problema | Abordagem Recomendada |
|---------------------------|----------------------|
| Landscape unimodal suave | Hill climbing, gradiente descendente |
| Landscape multimodal, solução única | Simulated annealing, busca tabu |
| Landscape multimodal, população de soluções | Algoritmos genéticos, PSO |
| Estrutura de grafo, caminho mínimo | Dijkstra, A* |
| Alta dimensionalidade contínua | Evolução diferencial, CMA-ES |
| Sequências/permutações | Colônia de formigas, 2-opt + busca tabu |
| Tempo de otimização muito limitado | GRASP, hill climbing multi-start |

````python
# Exercício 1b — TSP: tempo de busca exaustiva
import math

print("=" * 60)
print("Exercício 1b — TSP: Inviabilidade da Busca Exaustiva")
print("=" * 60)

velocidade = 1e9  # rotas/segundo

for n in [10, 15, 20, 25, 30]:
    if n <= 20:
        rotas = math.factorial(n - 1) // 2
        tempo_s = rotas / velocidade
        if tempo_s < 60:
            tempo_str = f"{tempo_s:.3f} segundos"
        elif tempo_s < 3600:
            tempo_str = f"{tempo_s/60:.1f} minutos"
        elif tempo_s < 86400:
            tempo_str = f"{tempo_s/3600:.1f} horas"
        elif tempo_s < 3.15e7:
            tempo_str = f"{tempo_s/86400:.1f} dias"
        elif tempo_s < 3.15e9:
            tempo_str = f"{tempo_s/3.15e7:.1f} anos"
        else:
            tempo_str = f"{tempo_s/3.15e7:.2e} anos"
        print(f"n={n:2d}: {rotas:>20,} rotas → {tempo_str}")
    else:
        # Estimativa em log
        log_rotas = sum(math.log10(i) for i in range(2, n))
        print(f"n={n:2d}: ~10^{log_rotas:.1f} rotas → estimativa: muito além da idade do universo")

print()
print("Idade do universo: ~13.8 bilhões de anos ≈ 4.35 × 10¹⁷ segundos")
print("Número de átomos no universo observável: ~10⁸⁰")
````
