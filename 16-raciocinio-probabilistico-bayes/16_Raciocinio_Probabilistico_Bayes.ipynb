## 1. O que é Raciocínio Probabilístico Bayesiano?

O **Raciocínio Probabilístico de Bayes** é uma abordagem para modelar e atualizar crenças sob incerteza com base em evidências. Ele é fundamentado no **Teorema de Bayes**, que expressa como atualizar a probabilidade de uma hipótese à luz de novas evidências observadas.

Em vez de buscar certeza absoluta (como na lógica clássica), o raciocínio bayesiano trabalha com **graus de crença** representados como probabilidades, atualizando-os de forma coerente à medida que novas informações chegam.

### O Teorema de Bayes

$$P(H \mid E) = \frac{P(E \mid H) \cdot P(H)}{P(E)}$$

| Termo | Nome | Significado |
|-------|------|-------------|
| $P(H)$ | **Prior** (a priori) | Probabilidade da hipótese **antes** de observar a evidência |
| $P(E \mid H)$ | **Verossimilhança** | Probabilidade de observar a evidência **dado que** a hipótese é verdadeira |
| $P(E)$ | **Evidência** | Probabilidade marginal da evidência (normalizador) |
| $P(H \mid E)$ | **Posterior** (a posteriori) | Probabilidade da hipótese **após** observar a evidência |

### Intuição: O Médico e o Diagnóstico

Um médico sabe que:
- 1% da população tem uma doença rara: $P(D) = 0.01$
- Um exame positivo ocorre em 90% dos doentes: $P(+|D) = 0.90$
- O exame dá falso positivo em 5% dos saudáveis: $P(+|\neg D) = 0.05$

Se um paciente testa positivo, qual a probabilidade de ele ter a doença?

$$P(D|+) = \frac{P(+|D) \cdot P(D)}{P(+|D) \cdot P(D) + P(+|\neg D) \cdot P(\neg D)} = \frac{0.90 \times 0.01}{0.90 \times 0.01 + 0.05 \times 0.99} \approx 0.154$$

**Resultado surpreendente:** apenas ~15,4% de chance de ter a doença mesmo com teste positivo! Isso é o **paradoxo do falso positivo** — a baixa prevalência domina o resultado.

## 3. Redes Bayesianas

Uma **Rede Bayesiana** (Bayesian Network, BN) é um grafo acíclico dirigido (DAG) onde:
- **Nós** representam variáveis aleatórias
- **Arestas** representam dependências causais ou correlações
- **Tabelas de Probabilidade Condicional (CPT)** quantificam as dependências

### Propriedade de Markov Local

Cada variável é independente de seus **não-descendentes**, dado seus **pais** na rede:

$$P(X_i \mid \text{todos os outros}) = P(X_i \mid \text{pais}(X_i))$$

### Fatoração da Distribuição Conjunta

$$P(X_1, X_2, \ldots, X_n) = \prod_{i=1}^{n} P(X_i \mid \text{pais}(X_i))$$

### Exemplo: Rede de Diagnóstico Médico

```
    [Fumar]    [Poluição]
       ↓            ↓
    [Câncer] ←───────┘
       ↓         ↓
  [Dispneia]  [Raio-X]
```

Cada nó tem uma CPT que define $P(X \mid \text{pais}(X))$.

## 4. Classificador Naive Bayes

O **Naive Bayes** aplica o Teorema de Bayes com a suposição **ingênua** de independência condicional entre as features dado a classe:

$$P(C \mid x_1, \ldots, x_n) \propto P(C) \cdot \prod_{i=1}^{n} P(x_i \mid C)$$

A classe predita é:
$$\hat{y} = \arg\max_C P(C) \cdot \prod_{i=1}^{n} P(x_i \mid C)$$

### Variantes do Naive Bayes

| Variante | Tipo de Feature | Distribuição assumida |
|----------|----------------|----------------------|
| **Gaussian NB** | Contínua | $P(x_i \mid C) \sim \mathcal{N}(\mu_C, \sigma_C^2)$ |
| **Multinomial NB** | Contagens (texto) | Distribuição multinomial |
| **Bernoulli NB** | Binária (presença/ausência) | Distribuição de Bernoulli |
| **Complement NB** | Texto desbalanceado | Complemento das classes |

## 5. Implementação do Zero — Teorema de Bayes e Naive Bayes

````python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

# ─────────────────────────────────────────────────────────────
# PARTE 1: Teorema de Bayes — Diagnóstico Médico
# ─────────────────────────────────────────────────────────────
def teorema_de_bayes(prior, verossimilhanca, verossimilhanca_negativa):
    """
    Aplica o Teorema de Bayes:
    P(H|E) = P(E|H) * P(H) / [P(E|H)*P(H) + P(E|~H)*P(~H)]
    """
    p_h = prior
    p_nao_h = 1 - prior
    p_e_dado_h = verossimilhanca
    p_e_dado_nao_h = verossimilhanca_negativa

    p_e = p_e_dado_h * p_h + p_e_dado_nao_h * p_nao_h
    posterior = (p_e_dado_h * p_h) / p_e
    return posterior, p_e

# Exemplo: Diagnóstico de doença rara
print("=" * 60)
print("  Diagnóstico Médico — Teorema de Bayes")
print("=" * 60)

prevalencia = 0.01      # 1% da população tem a doença
sensibilidade = 0.90    # P(+|Doente) = 90%
falso_positivo = 0.05   # P(+|Saudável) = 5%

posterior, p_positivo = teorema_de_bayes(prevalencia, sensibilidade, falso_positivo)

print(f"Prior P(D)       = {prevalencia:.3f} ({prevalencia*100:.1f}%)")
print(f"Verossimilhança  P(+|D)  = {sensibilidade:.3f}")
print(f"Falso positivo   P(+|~D) = {falso_positivo:.3f}")
print(f"Evidência        P(+)    = {p_positivo:.4f}")
print(f"Posterior        P(D|+)  = {posterior:.4f} ({posterior*100:.2f}%)")

# Visualizar efeito da prevalência no posterior
prevalencias = np.linspace(0.001, 0.5, 300)
posteriores = [teorema_de_bayes(p, sensibilidade, falso_positivo)[0] for p in prevalencias]

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(prevalencias * 100, [p * 100 for p in posteriores], 'b-', linewidth=2.5)
plt.axvline(prevalencia * 100, color='red', linestyle='--', alpha=0.7, label=f'Prevalência = {prevalencia*100}%')
plt.axhline(posterior * 100, color='green', linestyle='--', alpha=0.7, label=f'Posterior = {posterior*100:.1f}%')
plt.xlabel('Prevalência da Doença (%)')
plt.ylabel('P(Doença | Teste Positivo) (%)')
plt.title('Efeito da Prevalência no Valor Preditivo Positivo')
plt.legend()
plt.grid(True, alpha=0.3)

# Gráfico de pizza: decomposição dos casos positivos
plt.subplot(1, 2, 2)
verdadeiros_positivos = sensibilidade * prevalencia
falsos_positivos = falso_positivo * (1 - prevalencia)
plt.pie(
    [verdadeiros_positivos, falsos_positivos],
    labels=[f'Verdadeiro Positivo\n({verdadeiros_positivos:.3f})',
            f'Falso Positivo\n({falsos_positivos:.3f})'],
    colors=['steelblue', 'coral'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Composição dos Testes Positivos\n(prevalência = 1%)')

plt.suptitle('Paradoxo do Falso Positivo — Teorema de Bayes', fontsize=13)
plt.tight_layout()
plt.show()

print("\n⚠ Conclusão: Com prevalência de 1%, apenas",
      f"{posterior*100:.1f}% dos testes positivos são verdadeiros positivos!")
print("✅ Análise do Teorema de Bayes concluída.")
````

````python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import math

# ─────────────────────────────────────────────────────────────
# PARTE 2: Classificador Naive Bayes Gaussiano do zero
# ─────────────────────────────────────────────────────────────
class NaiveBayesGaussiano:
    """
    Classificador Naive Bayes Gaussiano implementado do zero.
    Assume que P(xi|C) ~ Normal(media_C, desvio_C).
    """

    def __init__(self):
        self.classes = None
        self.prior = {}          # P(C)
        self.media = {}          # média de cada feature por classe
        self.variancia = {}      # variância de cada feature por classe

    def fit(self, X, y):
        """Treina o modelo a partir dos dados de treinamento."""
        n_amostras, n_features = X.shape
        self.classes = np.unique(y)

        for c in self.classes:
            X_c = X[y == c]
            self.prior[c] = len(X_c) / n_amostras
            self.media[c] = X_c.mean(axis=0)
            self.variancia[c] = X_c.var(axis=0) + 1e-9  # suavização

    def _log_verossimilhanca_gaussiana(self, x, media, variancia):
        """Log-verossimilhança gaussiana: log P(x|média, variância)."""
        return -0.5 * np.sum(
            np.log(2 * np.pi * variancia) + ((x - media) ** 2) / variancia
        )

    def predict_proba(self, X):
        """Retorna probabilidades a posteriori para cada amostra."""
        probas = []
        for x in X:
            scores = {}
            for c in self.classes:
                log_prior = math.log(self.prior[c])
                log_veros = self._log_verossimilhanca_gaussiana(
                    x, self.media[c], self.variancia[c]
                )
                scores[c] = log_prior + log_veros
            # Converter para probabilidades (softmax em log-space)
            max_score = max(scores.values())
            exp_scores = {c: math.exp(s - max_score) for c, s in scores.items()}
            total = sum(exp_scores.values())
            probas.append({c: v / total for c, v in exp_scores.items()})
        return probas

    def predict(self, X):
        """Prediz a classe mais provável para cada amostra."""
        probas = self.predict_proba(X)
        return np.array([max(p, key=p.get) for p in probas])

    def score(self, X, y):
        """Calcula a acurácia no conjunto X, y."""
        return np.mean(self.predict(X) == y)


# ─────────────────────────────────────────────────────────────
# Testar no dataset Iris (gerado manualmente)
# ─────────────────────────────────────────────────────────────
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X, y = iris.data, iris.target

X_treino, X_teste, y_treino, y_teste = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Treinar Naive Bayes do zero
nb = NaiveBayesGaussiano()
nb.fit(X_treino, y_treino)
acc_scratch = nb.score(X_teste, y_teste)

print("=" * 55)
print("  Naive Bayes Gaussiano — Dataset Iris")
print("=" * 55)
print(f"  Implementação própria — Acurácia: {acc_scratch*100:.2f}%")

# Mostrar probabilidades a posteriori para as primeiras 5 amostras
print("\nProbabilidades a posteriori (primeiras 5 amostras de teste):")
probas = nb.predict_proba(X_teste[:5])
nomes_classes = iris.target_names
print(f"  {'Setosa':>10} {'Versicolor':>12} {'Virginica':>11} | Predito  | Real")
print("  " + "-" * 60)
for i, (p, real) in enumerate(zip(probas, y_teste[:5])):
    predito = max(p, key=p.get)
    print(f"  {p[0]:>10.4f} {p[1]:>12.4f} {p[2]:>11.4f} | "
          f"{nomes_classes[predito]:>8} | {nomes_classes[real]}")

# Visualizar fronteiras de decisão (features 0 e 2)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Feature 0 vs Feature 2
ax = axes[0]
feat_x, feat_y = 0, 2
x_min, x_max = X[:, feat_x].min() - 0.5, X[:, feat_x].max() + 0.5
y_min, y_max = X[:, feat_y].min() - 0.5, X[:, feat_y].max() + 0.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 150), np.linspace(y_min, y_max, 150))

# Criar dados de grade usando apenas features 0 e 2 (demais = média)
medias_globais = X_treino.mean(axis=0)
grade_full = np.tile(medias_globais, (xx.ravel().shape[0], 1))
grade_full[:, feat_x] = xx.ravel()
grade_full[:, feat_y] = yy.ravel()

Z = nb.predict(grade_full).reshape(xx.shape)
ax.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
cores = ['navy', 'darkorange', 'green']
for c, cor in enumerate(cores):
    mask = y_teste == c
    ax.scatter(X_teste[mask, feat_x], X_teste[mask, feat_y],
               c=cor, label=nomes_classes[c], edgecolors='black', s=50)
ax.set_xlabel(iris.feature_names[feat_x])
ax.set_ylabel(iris.feature_names[feat_y])
ax.set_title('Fronteira de Decisão — Naive Bayes\n(features 0 e 2)')
ax.legend()
ax.grid(True, alpha=0.3)

# Distribuições gaussianas aprendidas (feature 2)
ax = axes[1]
feat_plot = 2
x_vals = np.linspace(X[:, feat_plot].min() - 0.5, X[:, feat_plot].max() + 0.5, 300)
for c, cor in enumerate(cores):
    mu = nb.media[c][feat_plot]
    sigma2 = nb.variancia[c][feat_plot]
    gaussiana = (1 / np.sqrt(2 * np.pi * sigma2)) * np.exp(-0.5 * (x_vals - mu)**2 / sigma2)
    gaussiana = gaussiana * nb.prior[c]
    ax.plot(x_vals, gaussiana, color=cor, linewidth=2.5, label=f'{nomes_classes[c]} (μ={mu:.2f})')
    # Histograma dos dados de treino
    ax.hist(X_treino[y_treino == c, feat_plot], bins=12, color=cor,
            alpha=0.2, density=True, weights=np.ones(sum(y_treino == c)) * nb.prior[c])
ax.set_xlabel(iris.feature_names[feat_plot])
ax.set_ylabel('P(feature | classe) × P(classe)')
ax.set_title('Distribuições Gaussianas Aprendidas\n(weighted likelihood)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.suptitle('Naive Bayes Gaussiano — Iris Dataset', fontsize=13)
plt.tight_layout()
plt.show()
print("\n✅ Naive Bayes do zero concluído.")
````

## 6. Implementação com Bibliotecas — scikit-learn

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.datasets import load_iris, fetch_20newsgroups
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────────
# 1. Gaussian NB (sklearn) vs Implementação própria
# ─────────────────────────────────────────────────────────────
iris = load_iris()
X, y = iris.data, iris.target
X_treino, X_teste, y_treino, y_teste = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

gnb_sklearn = GaussianNB()
gnb_sklearn.fit(X_treino, y_treino)
acc_sklearn = gnb_sklearn.score(X_teste, y_teste)

nb_proprio = NaiveBayesGaussiano()
nb_proprio.fit(X_treino, y_treino)
acc_proprio = nb_proprio.score(X_teste, y_teste)

print("=" * 55)
print("  Comparação: Implementação própria vs sklearn")
print("=" * 55)
print(f"  Própria (GaussianNB): {acc_proprio*100:.2f}%")
print(f"  sklearn  (GaussianNB): {acc_sklearn*100:.2f}%")

# ─────────────────────────────────────────────────────────────
# 2. Naive Bayes para Classificação de Texto (Spam / Not Spam)
# ─────────────────────────────────────────────────────────────
# Dataset simulado de e-mails
emails_spam = [
    "ganhe dinheiro fácil agora clique aqui",
    "oferta imperdível ganhe prêmio grátis",
    "você foi selecionado para ganhar dinheiro",
    "clique aqui para ganhar prêmio incrível",
    "promoção especial desconto imperdível hoje",
    "ganhe dinheiro trabalhando em casa fácil",
    "oferta relâmpago clique ganhe agora grátis",
    "você ganhou bilhete premiado clique aqui",
    "investimento garantido lucro fácil rápido",
    "aproveite promoção exclusiva dinheiro grátis",
]

emails_ham = [
    "reunião amanhã às 10h sala de conferências",
    "prezado colega segue em anexo o relatório",
    "confirmo presença no evento de sexta feira",
    "o projeto foi aprovado pelo comitê técnico",
    "bom dia favor revisar o documento enviado",
    "encaminho o orçamento conforme solicitado",
    "convite para apresentação do trabalho final",
    "favor confirmar recebimento desta mensagem",
    "segue cronograma de atividades do trimestre",
    "lembrete da reunião de equipe amanhã cedo",
]

textos = emails_spam + emails_ham
rotulos = [1] * len(emails_spam) + [0] * len(emails_ham)

# Pipeline: TF-IDF + Multinomial NB
pipeline_spam = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', MultinomialNB(alpha=1.0))
])

pipeline_spam.fit(textos, rotulos)

# Testar novos e-mails
novos_emails = [
    "ganhe dinheiro agora clique aqui grátis",
    "reunião de equipe segunda feira às 9h",
    "promoção imperdível oferta especial hoje",
    "prezado favor confirmar recebimento do documento",
]

predicoes = pipeline_spam.predict(novos_emails)
probas_spam = pipeline_spam.predict_proba(novos_emails)

print("\n" + "=" * 55)
print("  Filtro de Spam — Naive Bayes Multinomial")
print("=" * 55)
print(f"{'E-mail':>40} | {'Classe':>8} | {'P(spam)':>8}")
print("-" * 65)
for email, pred, proba in zip(novos_emails, predicoes, probas_spam):
    classe = 'SPAM' if pred == 1 else 'HAM '
    print(f"  {email[:38]:>38} | {classe:>8} | {proba[1]:>7.3f}")

# ─────────────────────────────────────────────────────────────
# 3. Comparação de variantes do Naive Bayes
# ─────────────────────────────────────────────────────────────
print("\n" + "=" * 55)
print("  Cross-Validation — Variantes do Naive Bayes (Iris)")
print("=" * 55)

modelos = [
    ('GaussianNB', GaussianNB()),
    ('BernoulliNB', BernoulliNB()),
]

resultados_cv = {}
for nome, modelo in modelos:
    scores = cross_val_score(modelo, X, y, cv=10, scoring='accuracy')
    resultados_cv[nome] = scores
    print(f"  {nome:15s}: {scores.mean()*100:.2f}% ± {scores.std()*100:.2f}%")

# Visualização: comparação das variantes
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Box plot de CV scores
ax = axes[0]
dados_box = [resultados_cv[n] * 100 for n in resultados_cv]
nomes_box = list(resultados_cv.keys())
bp = ax.boxplot(dados_box, labels=nomes_box, patch_artist=True)
cores_box = ['steelblue', 'coral', 'green', 'gold']
for patch, cor in zip(bp['boxes'], cores_box):
    patch.set_facecolor(cor)
    patch.set_alpha(0.7)
ax.set_title('Variantes do Naive Bayes\n(10-fold CV, Iris)')
ax.set_ylabel('Acurácia (%)')
ax.grid(True, axis='y', alpha=0.3)

# Matriz de confusão do GaussianNB
ax = axes[1]
gnb_sklearn.fit(X_treino, y_treino)
y_pred = gnb_sklearn.predict(X_teste)
cm = confusion_matrix(y_teste, y_pred)
im = ax.imshow(cm, cmap='Blues')
plt.colorbar(im, ax=ax)
ax.set_xticks(range(3))
ax.set_yticks(range(3))
ax.set_xticklabels(iris.target_names)
ax.set_yticklabels(iris.target_names)
ax.set_xlabel('Predito')
ax.set_ylabel('Real')
ax.set_title('Matriz de Confusão\nGaussian NB (Iris)')
for i in range(3):
    for j in range(3):
        ax.text(j, i, str(cm[i, j]), ha='center', va='center',
                color='white' if cm[i, j] > cm.max() / 2 else 'black', fontsize=14)

plt.suptitle('Naive Bayes com scikit-learn', fontsize=13)
plt.tight_layout()
plt.show()
print("\n✅ Implementação com bibliotecas concluída.")
````

## 7. Redes Bayesianas — Inferência Manual

````python
import numpy as np
import matplotlib.pyplot as plt

# ─────────────────────────────────────────────────────────────
# Rede Bayesiana: Alarme de Carro (Pearl, 1988)
# Variáveis: Assalto (A), Terremoto (T), Alarme (Al),
#            João liga (J), Maria liga (M)
# ─────────────────────────────────────────────────────────────

# Probabilidades a priori (priors)
P_A = 0.001    # P(Assalto = True)
P_T = 0.002    # P(Terremoto = True)

# CPT do Alarme dado Assalto e Terremoto
# P(Al=T | A, T): tabela indexada por (A, T)
P_Al_dado_AT = {
    (True, True):   0.95,
    (True, False):  0.94,
    (False, True):  0.29,
    (False, False): 0.001,
}

# CPT de João dado Alarme
P_J_dado_Al = {True: 0.90, False: 0.05}

# CPT de Maria dado Alarme
P_M_dado_Al = {True: 0.70, False: 0.01}

def inferencia_enumeracao(evidencias):
    """
    Inferência por enumeração completa.
    Calcula P(Assalto=True | evidências).
    """
    def prob_conjunta(a, t, al, j_obs, m_obs):
        """Calcula P(A=a, T=t, Al=al, J=j_obs, M=m_obs)."""
        p = (P_A if a else 1 - P_A)
        p *= (P_T if t else 1 - P_T)
        p_al = P_Al_dado_AT[(a, t)]
        p *= (p_al if al else 1 - p_al)
        if j_obs is not None:
            p_j = P_J_dado_Al[al]
            p *= (p_j if j_obs else 1 - p_j)
        if m_obs is not None:
            p_m = P_M_dado_Al[al]
            p *= (p_m if m_obs else 1 - p_m)
        return p

    j_obs = evidencias.get('J', None)
    m_obs = evidencias.get('M', None)

    # Marginalizar sobre T e Al
    p_a_true = 0.0
    p_a_false = 0.0
    for t in [True, False]:
        for al in [True, False]:
            p_a_true  += prob_conjunta(True,  t, al, j_obs, m_obs)
            p_a_false += prob_conjunta(False, t, al, j_obs, m_obs)

    # Normalizar
    total = p_a_true + p_a_false
    return p_a_true / total, p_a_false / total

# Consultas de inferência
consultas = [
    ({}, "P(Assalto) sem evidências"),
    ({'J': True}, "P(Assalto | João ligou)"),
    ({'M': True}, "P(Assalto | Maria ligou)"),
    ({'J': True, 'M': True}, "P(Assalto | João E Maria ligaram)"),
    ({'J': False, 'M': False}, "P(Assalto | ninguém ligou)"),
]

print("=" * 60)
print("  Rede Bayesiana do Alarme — Inferência por Enumeração")
print("=" * 60)
resultados_rb = []
for evid, descricao in consultas:
    p_true, p_false = inferencia_enumeracao(evid)
    resultados_rb.append((descricao, p_true))
    print(f"  {descricao}")
    print(f"    → P(Assalto=True)  = {p_true:.6f} ({p_true*100:.4f}%)")
    print(f"    → P(Assalto=False) = {p_false:.6f}")
    print()

# Visualizar
fig, ax = plt.subplots(figsize=(12, 5))
labels = [r[0].replace('P(Assalto | ', '').replace(')', '').replace('P(Assalto) sem evidências', 'Sem evidência')
          for r in resultados_rb]
valores = [r[1] * 100 for r in resultados_rb]
cores = ['lightblue' if v < 1 else 'steelblue' if v < 10 else 'navy' for v in valores]
bars = ax.barh(labels, valores, color=cores, edgecolor='black', alpha=0.8)
for bar, val in zip(bars, valores):
    ax.text(val + 0.002, bar.get_y() + bar.get_height() / 2,
            f'{val:.4f}%', va='center', fontsize=10)
ax.set_xlabel('P(Assalto = True) (%)')
ax.set_title('Inferência Bayesiana — Rede do Alarme de Carro\n(atualização de crença com evidências)')
ax.grid(True, axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
print("✅ Inferência na Rede Bayesiana concluída.")
````

## 8. Comparação: Abordagem Bayesiana vs. Frequentista

| Aspecto | Frequentista | Bayesiana |
|---------|-------------|----------|
| **Definição de probabilidade** | Frequência relativa em experimentos repetidos | Grau de crença (subjetiva ou objetiva) |
| **Parâmetros** | Fixos, desconhecidos | Variáveis aleatórias com distribuições |
| **Conhecimento prévio** | Não incorporado formalmente | Incorporado via distribuição prior |
| **Incerteza sobre parâmetros** | Intervalos de confiança | Intervalos de credibilidade |
| **Atualização** | Novos dados → novo modelo | Prior + dados → posterior |
| **Tamanho de amostra pequeno** | Problemas com inferência | Prior ajuda a regularizar |
| **Exemplo** | p-valor, teste t | P(H\|E) via Teorema de Bayes |

## 10. Vantagens e Desvantagens

| Vantagens | Desvantagens |
|-----------|-------------|
| ✅ Princípio matemático coerente para raciocínio sob incerteza | ❌ Escolha do prior pode ser subjetiva e influente |
| ✅ Atualização incremental de crenças | ❌ Inferência exata é NP-difícil em redes complexas |
| ✅ Naive Bayes é simples, rápido e surpreendentemente eficaz | ❌ Suposição de independência do Naive Bayes raramente é verdadeira |
| ✅ Redes Bayesianas modelam dependências causais explicitamente | ❌ Estrutura da rede bayesiana é difícil de aprender automaticamente |
| ✅ Funciona bem com poucos dados (prior regulariza) | ❌ Inferência em grandes redes requer métodos aproximados (MCMC, VI) |
| ✅ Interpretabilidade: probabilidades têm significado claro | ❌ Naive Bayes produz probabilidades mal calibradas |
