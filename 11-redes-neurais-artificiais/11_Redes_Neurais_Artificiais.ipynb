## 1. O que são Redes Neurais Artificiais?

As **Redes Neurais Artificiais (RNA)** são modelos computacionais inspirados no sistema nervoso humano. São sistemas compostos por neurônios artificiais interconectados que aprendem a partir de exemplos, ajustando seus parâmetros internos para realizar tarefas como classificação, regressão, reconhecimento de padrões e geração de conteúdo.

### Definição

Uma RNA é um sistema de neurônios artificiais interconectados que **aprende a partir de exemplos**. Diferentemente de sistemas baseados em regras, as RNAs descobrem padrões nos dados de forma automática, sem que o programador precise especificar explicitamente como resolver o problema.

### Linha do Tempo Histórica

| Ano  | Marco |
|------|-------|
| 1943 | **McCulloch & Pitts** — primeiro neurônio artificial matemático |
| 1957 | **Perceptron de Rosenblatt** — primeiro modelo capaz de aprender |
| 1969 | **Minsky & Papert** — demonstram limitações do Perceptron (problema XOR) |
| 1986 | **Rumelhart, Hinton & Williams** — algoritmo Backpropagation |

| 2012 | **AlexNet** — revolução em visão computacional com GPUs |
| 2017 | **Transformer** — "Attention is All You Need", base do BERT e GPT |

### RNA e Deep Learning

As RNAs são a **base do Deep Learning**. Quando empilhamos muitas camadas de neurônios artificiais, obtemos redes profundas capazes de aprender representações hierárquicas dos dados — desde bordas simples até conceitos abstratos como rostos, objetos e linguagem.

## 2. O Neurônio Biológico vs. Neurônio Artificial

### Neurônio Biológico

O neurônio biológico é a unidade fundamental do sistema nervoso. Seu funcionamento pode ser descrito em três etapas:

1. **Dendritos** — recebem sinais elétricos de outros neurônios
2. **Soma (corpo celular)** — integra todos os sinais recebidos
3. **Axônio** — transmite o sinal de saída para outros neurônios via sinapses

### Mapeamento para o Neurônio Artificial

| Componente Biológico | Componente Artificial |
|----------------------|----------------------|
| Dendritos | Entradas: x1, x2, ..., xn |
| Força sináptica | Pesos: w1, w2, ..., wn |
| Soma (corpo celular) | Soma ponderada + função de ativação |
| Axônio | Saída: y |

### Fórmula Matemática

A saída de um neurônio artificial é calculada como:

```
y = f( w1*x1 + w2*x2 + ... + wn*xn + b )
  = f( Σ(wi * xi) + b )
```

Onde:
- **xi** são as entradas
- **wi** são os pesos (força das conexões)
- **b** é o bias (limiar de ativação)
- **f(·)** é a função de ativação
- **y** é a saída do neurônio

### Aprendizado por Ajuste de Pesos

Durante o treinamento, os **pesos são aprendidos automaticamente** a partir dos dados. O algoritmo de aprendizado ajusta os pesos iterativamente para minimizar o erro entre a saída prevista e a saída desejada. É exatamente essa capacidade de ajustar os pesos que distingue as RNAs de sistemas de regras fixas.

## 3. Funções de Ativação

A função de ativação introduz **não-linearidade** na rede, permitindo que ela aprenda padrões complexos. Sem funções de ativação não-lineares, uma rede com múltiplas camadas se reduziria a uma simples transformação linear.

### Degrau (Step)

```
f(x) = 1, se x >= 0
f(x) = 0, se x < 0
```

- Saída **binária** (0 ou 1)
- Usada no **Perceptron original** de Rosenblatt
- Limitação: derivada nula em quase todo ponto, impossibilitando Backpropagation

### Sigmoide (Logística)

```
f(x) = 1 / (1 + e^(-x))
```

- Saída no intervalo **(0, 1)**
- Usada na **camada de saída** para classificação binária
- Problema: **vanishing gradient** em valores extremos de x

### Tanh (Tangente Hiperbólica)

```
f(x) = (e^x - e^(-x)) / (e^x + e^(-x))
```

- Saída no intervalo **(-1, 1)**
- **Centralizada em zero** (zero-centered), o que facilita o aprendizado
- Gradientes mais fortes que sigmoide, mas ainda sofre de vanishing gradient

### ReLU (Rectified Linear Unit)

```
f(x) = max(0, x)
```

- A função de ativação **mais utilizada nas camadas ocultas**
- Computacionalmente eficiente
- **Atenua o problema do vanishing gradient** (gradiente constante para x > 0)
- Problema: "neurônios mortos" (dying ReLU) quando x < 0

### Softmax

```
f(xi) = e^xi / Σ(e^xj)  para j = 1, ..., K
```

- Usada na **camada de saída para classificação multi-classe**
- Converte um vetor de scores em uma **distribuição de probabilidade**
- A soma de todas as saídas é sempre igual a 1

## 4. Arquiteturas de Redes Neurais

### Perceptron Simples

- Composto por **um único neurônio**
- Realiza **classificação binária linear**
- Funciona apenas para problemas **linearmente separáveis**
- Falha em problemas como XOR

### MLP — Multi-Layer Perceptron

- Possui **camada de entrada**, uma ou mais **camadas ocultas** e **camada de saída**
- Cada neurônio está **totalmente conectado** (fully connected) à camada seguinte
- Treinado com **Backpropagation** e gradiente descendente
- Capaz de aprender funções não-lineares complexas

### CNN — Convolutional Neural Network

- Possui **camadas convolucionais** que detectam padrões locais (bordas, texturas)
- **Camadas de pooling** reduzem a dimensionalidade espacial
- **Camadas densas** realizam a classificação final
- Estado da arte em **visão computacional** (imagens, vídeos)

### RNN / LSTM — Redes Recorrentes

- Possuem **conexões recorrentes** que permitem processar sequências
- **LSTM (Long Short-Term Memory)** resolve o problema de dependências de longo prazo
- Aplicadas em **séries temporais, NLP, reconhecimento de fala**

### Transformers

- Baseados no **mecanismo de atenção (attention)**
- Processam sequências em paralelo (ao contrário das RNNs)
- Base dos modelos **BERT** (Google) e **GPT** (OpenAI)
- **Estado da arte** em NLP e cada vez mais em visão (Vision Transformer)

## 5. O Algoritmo Backpropagation

O **Backpropagation** (retropropagação do erro) é o algoritmo fundamental para treinar redes neurais com múltiplas camadas. Combina a **regra da cadeia** do cálculo diferencial com o **gradiente descendente**.

### Passo 1: Forward Pass (Passagem Direta)

- Os dados fluem da **camada de entrada até a saída**
- Cada neurônio computa sua soma ponderada e aplica a função de ativação
- Ao final, a rede produz uma **predição ŷ**

### Passo 2: Função de Perda (Loss Function)

- Mede o **erro entre a predição ŷ e o valor real y**
- **MSE** (Mean Squared Error) para regressão: `L = (1/n) * Σ(yi - ŷi)²`
- **Cross-Entropy** para classificação: `L = -Σ yi * log(ŷi)`

### Passo 3: Backward Pass (Passagem Reversa)

- Usa a **regra da cadeia** para calcular o gradiente da perda em relação a cada peso
- Propaga o erro da camada de saída até a camada de entrada
- Para cada peso w: `dL/dw = dL/dŷ * dŷ/dz * dz/dw`

### Passo 4: Atualização dos Pesos (Gradiente Descendente)

```
w = w - taxa_aprendizado * (dL/dw)
```

- A **taxa de aprendizado (learning rate)** controla o tamanho do passo
- Muito alta: oscila e diverge; muito baixa: converge lentamente

### Problema do Vanishing Gradient

- Com funções sigmoide ou tanh, os gradientes ficam **muito pequenos** nas camadas iniciais
- O aprendizado das camadas iniciais torna-se extremamente lento
- **ReLU** atenua esse problema pois seu gradiente é constante (1) para x > 0
- Outras soluções: **Batch Normalization**, **inicialização cuidadosa dos pesos**, **ResNets**

````python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    """
    Implementação do Perceptron de Rosenblatt (1957) do zero.
    Classificador binário linear para problemas linearmente separáveis.
    """
    
    def __init__(self, taxa_aprendizado=0.1, epocas=100):
        self.taxa_aprendizado = taxa_aprendizado
        self.epocas = epocas
        self.pesos = None
        self.bias = None
        self.historico_erros = []
    
    def funcao_degrau(self, x):
        return np.where(x >= 0, 1, 0)
    
    def fit(self, X, y):
        n_amostras, n_features = X.shape
        self.pesos = np.zeros(n_features)
        self.bias = 0
        
        for epoca in range(self.epocas):
            erros = 0
            for xi, yi in zip(X, y):
                saida = self.funcao_degrau(np.dot(xi, self.pesos) + self.bias)
                erro = yi - saida
                self.pesos += self.taxa_aprendizado * erro * xi
                self.bias += self.taxa_aprendizado * erro
                erros += int(erro != 0)
            self.historico_erros.append(erros)
        return self
    
    def predict(self, X):
        saida_linear = np.dot(X, self.pesos) + self.bias
        return self.funcao_degrau(saida_linear)

# Dados de treinamento: porta AND
print("=== Testando Perceptron - Porta AND ===")
X_and = np.array([[0,0],[0,1],[1,0],[1,1]])
y_and = np.array([0, 0, 0, 1])

perceptron_and = Perceptron(taxa_aprendizado=0.1, epocas=20)
perceptron_and.fit(X_and, y_and)

print("\nPorta AND:")
for xi, yi in zip(X_and, y_and):
    pred = perceptron_and.predict(xi.reshape(1, -1))
    print(f"  Entrada: {xi} | Esperado: {yi} | Predito: {pred[0]}")

print(f"\nPesos finais: {perceptron_and.pesos}")
print(f"Bias final: {perceptron_and.bias}")

# Dados de treinamento: porta OR
print("\n=== Testando Perceptron - Porta OR ===")
X_or = np.array([[0,0],[0,1],[1,0],[1,1]])
y_or = np.array([0, 1, 1, 1])

perceptron_or = Perceptron(taxa_aprendizado=0.1, epocas=20)
perceptron_or.fit(X_or, y_or)

print("\nPorta OR:")
for xi, yi in zip(X_or, y_or):
    pred = perceptron_or.predict(xi.reshape(1, -1))
    print(f"  Entrada: {xi} | Esperado: {yi} | Predito: {pred[0]}")

# Visualização dos erros
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(perceptron_and.historico_erros, 'b-o', markersize=4)
plt.title('Convergência - Porta AND')
plt.xlabel('Época')
plt.ylabel('Número de Erros')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(perceptron_or.historico_erros, 'r-o', markersize=4)
plt.title('Convergência - Porta OR')
plt.xlabel('Época')
plt.ylabel('Número de Erros')
plt.grid(True)

plt.tight_layout()
salvar_figuras = False  # Mude para True para salvar figuras em arquivo
if salvar_figuras:
    plt.savefig('perceptron_convergencia.png', dpi=80, bbox_inches='tight')
plt.show()
print("\nLimitação: O Perceptron NÃO consegue resolver XOR (não linearmente separável)")
````

````python
import numpy as np

class RedeNeuralMLP:
    """
    Rede Neural MLP (Multi-Layer Perceptron) implementada do zero.
    Suporta múltiplas camadas ocultas com ativação sigmoide.
    Usa Backpropagation para treinamento.
    """
    
    def __init__(self, camadas, taxa_aprendizado=0.1, epocas=1000, seed=42):
        """
        camadas: lista com número de neurônios por camada
                 Ex: [2, 4, 1] = 2 entradas, 4 neurônios ocultos, 1 saída
        """
        self.camadas = camadas
        self.taxa_aprendizado = taxa_aprendizado
        self.epocas = epocas
        self.pesos = []
        self.biases = []
        self.historico_loss = []
        self._rng = np.random.default_rng(seed)
        self._inicializar_pesos()
    
    def _inicializar_pesos(self):
        for i in range(len(self.camadas) - 1):
            W = self._rng.standard_normal((self.camadas[i], self.camadas[i+1])) * 0.1
            b = np.zeros((1, self.camadas[i+1]))
            self.pesos.append(W)
            self.biases.append(b)
    
    def sigmoide(self, x):
        return 1 / (1 + np.exp(-np.clip(x, -500, 500)))
    
    def sigmoide_derivada(self, x):
        s = self.sigmoide(x)
        return s * (1 - s)
    
    def forward_pass(self, X):
        self.ativacoes = [X]
        self.entradas_z = []
        entrada = X
        for W, b in zip(self.pesos, self.biases):
            z = np.dot(entrada, W) + b
            self.entradas_z.append(z)
            a = self.sigmoide(z)
            self.ativacoes.append(a)
            entrada = a
        return entrada
    
    def backward_pass(self, X, y):
        n = X.shape[0]
        deltas = []
        # Erro na camada de saída
        erro_saida = self.ativacoes[-1] - y
        delta = erro_saida * self.sigmoide_derivada(self.entradas_z[-1])
        deltas.append(delta)
        # Propagar erro para camadas anteriores
        for i in range(len(self.pesos) - 2, -1, -1):
            delta = np.dot(deltas[-1], self.pesos[i+1].T) * self.sigmoide_derivada(self.entradas_z[i])
            deltas.append(delta)
        deltas.reverse()
        # Atualizar pesos
        for i in range(len(self.pesos)):
            self.pesos[i] -= self.taxa_aprendizado * np.dot(self.ativacoes[i].T, deltas[i]) / n
            self.biases[i] -= self.taxa_aprendizado * np.mean(deltas[i], axis=0, keepdims=True)
    
    def treinar(self, X, y):
        for epoca in range(self.epocas):
            saida = self.forward_pass(X)
            self.backward_pass(X, y)
            if epoca % 100 == 0:
                loss = np.mean((saida - y) ** 2)
                self.historico_loss.append(loss)
        return self
    
    def prever(self, X):
        saida = self.forward_pass(X)
        return (saida >= 0.5).astype(int)

# Problema XOR - não linearmente separável, requer MLP
print("=== Testando MLP - Problema XOR ===")
X_xor = np.array([[0,0],[0,1],[1,0],[1,1]])
y_xor = np.array([[0],[1],[1],[0]])

mlp = RedeNeuralMLP(camadas=[2, 4, 1], taxa_aprendizado=0.5, epocas=5000)
mlp.treinar(X_xor, y_xor)

print("\nResultados XOR:")
for xi, yi in zip(X_xor, y_xor):
    pred = mlp.prever(xi.reshape(1, -1))
    saida_raw = mlp.forward_pass(xi.reshape(1, -1))
    print(f"  Entrada: {xi} | Esperado: {yi[0]} | Predito: {pred[0][0]} | Saída raw: {saida_raw[0][0]:.4f}")

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.plot(range(0, mlp.epocas, 100), mlp.historico_loss, 'g-')
plt.title('Curva de Aprendizado - MLP no XOR')
plt.xlabel('Época')
plt.ylabel('Loss (MSE)')
plt.grid(True)
plt.tight_layout()
salvar_figuras = False  # Mude para True para salvar figuras em arquivo
if salvar_figuras:
    plt.savefig('mlp_loss.png', dpi=80, bbox_inches='tight')
plt.show()
print(f"\nÚltimo loss registrado (a cada 100 épocas): {mlp.historico_loss[-1]:.6f}")
````

````python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import matplotlib.pyplot as plt

print("=== Comparação: MLP do Zero vs. sklearn MLPClassifier ===\n")

# Dataset: make_moons (não linearmente separável)
X, y = make_moons(n_samples=300, noise=0.3, random_state=42)
X = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# sklearn MLPClassifier
clf = MLPClassifier(
    hidden_layer_sizes=(10, 5),
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42,
    learning_rate_init=0.01
)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print("sklearn MLPClassifier:")
print(f"  Acurácia: {accuracy_score(y_test, y_pred):.4f}")
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))

# Visualização
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, (X_plot, y_plot, title) in zip(axes, [
    (X_train, y_train, "Conjunto de Treino"),
    (X_test, y_test, "Conjunto de Teste")
]):
    xx, yy = np.meshgrid(np.linspace(X[:,0].min()-0.5, X[:,0].max()+0.5, 100),
                          np.linspace(X[:,1].min()-0.5, X[:,1].max()+0.5, 100))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='RdYlBu')
    ax.scatter(X_plot[:,0], X_plot[:,1], c=y_plot, cmap='RdYlBu', edgecolors='k', s=30)
    ax.set_title(title)
    ax.grid(True, alpha=0.3)

plt.suptitle('MLPClassifier - Fronteira de Decisão (Make Moons)', fontsize=14)
plt.tight_layout()
salvar_figuras = False  # Mude para True para salvar figuras em arquivo
if salvar_figuras:
    plt.savefig('mlp_sklearn.png', dpi=80, bbox_inches='tight')
plt.show()

print("\nArquitetura utilizada: 2 entradas -> 10 neurônios -> 5 neurônios -> 1 saída")
print(f"Iterações realizadas: {clf.n_iter_}")
print(f"Loss final: {clf.loss_:.6f}")
````

## 8. Vantagens e Desvantagens

### ✅ Vantagens

- **Aprendizado automático de representações complexas:** As RNAs descobrem automaticamente quais características dos dados são relevantes, sem necessidade de engenharia manual de features.
- **Excelente desempenho em dados não estruturados:** Imagens, textos e áudio são tratados de forma nativa pelas arquiteturas adequadas (CNN, Transformer, RNN).
- **Escalabilidade:** O desempenho melhora consistentemente com mais dados e maior capacidade computacional (mais GPUs/TPUs).
- **Versatilidade:** A mesma arquitetura base pode ser adaptada para resolver diferentes problemas com ajuste de hiperparâmetros e dados.
- **Estado da arte:** Superam os demais métodos em tarefas como reconhecimento de imagens, tradução automática e geração de linguagem natural.

### ❌ Desvantagens

- **Caixa-preta (Black Box):** Difícil interpretabilidade — é desafiador explicar por que a rede tomou determinada decisão, o que é crítico em domínios como medicina e direito.
- **Grande quantidade de dados rotulados:** Para atingir alto desempenho, geralmente são necessários milhares ou milhões de exemplos etiquetados.
- **Alto custo computacional de treinamento:** Modelos grandes (como GPT-4) exigem data centers inteiros e consomem enormes quantidades de energia.
- **Hiperparâmetros difíceis de ajustar:** Número de camadas, neurônios, taxa de aprendizado, regularização — o ajuste ótimo requer experiência e experimentação extensiva.
- **Propenso a overfitting:** Sem técnicas de regularização (Dropout, L2, Data Augmentation), a rede pode memorizar os dados de treino e generalizar mal.
- **Tempo de treinamento elevado:** Modelos de linguagem modernos levam dias ou semanas para treinar, mesmo com hardware de ponta.
