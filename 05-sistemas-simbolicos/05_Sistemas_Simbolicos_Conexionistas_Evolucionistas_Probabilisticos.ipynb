## 1. Introdução às Abordagens da IA

A Inteligência Artificial, desde sua fundação na Conferência de Dartmouth em 1956, não seguiu um caminho único de desenvolvimento. Ao longo das décadas, pesquisadores propuseram diferentes formas de representar e processar o conhecimento, inspirando-se em distintas fontes: a lógica matemática, o cérebro humano, a evolução biológica e a teoria das probabilidades. Essas diferentes inspirações deram origem a **quatro grandes paradigmas** (ou escolas de pensamento) da IA, cada um com suas próprias ferramentas, técnicas e filosofias.

Segundo **Russell & Norvig (2021)**, em *Inteligência Artificial: Uma Abordagem Moderna*, a evolução da IA pode ser compreendida pela alternância de predominância entre essas abordagens. Nos primeiros anos (décadas de 1950–70), a abordagem **simbólica** dominou, com foco na manipulação de símbolos e regras lógicas. A partir da década de 1980, a abordagem **conexionista** ganhou força com o ressurgimento das redes neurais. Paralelamente, os **algoritmos evolucionistas** se consolidaram como métodos poderosos de otimização, e os **métodos probabilísticos** se tornaram fundamentais para lidar com a incerteza inerente ao mundo real.

**Nilsson (1998)**, em *Artificial Intelligence: A New Synthesis*, destaca que cada paradigma oferece uma perspectiva complementar sobre a inteligência. Não se trata de abordagens concorrentes, mas sim de ferramentas distintas que, quando combinadas, permitem construir sistemas de IA mais robustos e capazes. A compreensão dessas quatro escolas é essencial para qualquer estudante de Ciência da Computação que deseje atuar na área de Inteligência Artificial.

Os quatro paradigmas podem ser resumidos da seguinte forma:

| Paradigma | Inspiração | Representação Principal | Exemplo de Técnica |
|---|---|---|---|
| **Simbólico** | Lógica e raciocínio humano | Símbolos e regras | Sistemas Especialistas |
| **Conexionista** | Cérebro e neurônios biológicos | Redes de conexões com pesos | Redes Neurais Artificiais |
| **Evolucionista** | Evolução e seleção natural | Populações de soluções candidatas | Algoritmos Genéticos |
| **Probabilístico** | Estatística e teoria das probabilidades | Distribuições de probabilidade | Redes Bayesianas |

Nas seções a seguir, exploraremos cada um desses paradigmas em profundidade, com fundamentação teórica e exemplos práticos em Python.

## 2. Sistemas Simbólicos (IA Simbólica)

### 2.1 Fundamentos Teóricos

A **IA Simbólica** (também conhecida como **GOFAI — Good Old-Fashioned Artificial Intelligence**, termo cunhado por John Haugeland em 1985) é a abordagem mais antiga e historicamente mais influente da Inteligência Artificial. Sua premissa fundamental é que a **inteligência pode ser modelada pela manipulação de símbolos** — representações abstratas de conceitos, objetos e relações do mundo real — por meio de regras formais bem definidas.

Essa abordagem tem suas raízes na **lógica matemática** de Aristóteles, passando pelos trabalhos de George Boole (álgebra booleana), Gottlob Frege (lógica de predicados), Alfred North Whitehead e Bertrand Russell (*Principia Mathematica*) e Alan Turing (máquinas de Turing). A ideia central é que o raciocínio humano pode ser formalizado como um sistema de manipulação simbólica: dadas premissas (fatos) e regras de inferência, é possível derivar conclusões de forma mecânica e determinística.

Os **Sistemas Especialistas** são o exemplo mais emblemático da IA Simbólica. Desenvolvidos principalmente nas décadas de 1970 e 1980, esses sistemas utilizam uma **base de conhecimento** (conjunto de fatos e regras fornecidos por especialistas humanos) e um **motor de inferência** (mecanismo que aplica as regras aos fatos para derivar novas conclusões). Exemplos clássicos incluem o **MYCIN** (diagnóstico médico), o **DENDRAL** (análise de compostos químicos) e o **XCON/R1** (configuração de computadores). A representação do conhecimento pode utilizar **lógica proposicional**, **lógica de predicados**, **frames**, **redes semânticas** ou **regras de produção** (do tipo SE...ENTÃO).

A IA Simbólica apresenta importantes **vantagens**: o raciocínio é **explicável e auditável** (é possível rastrear a cadeia de inferências que levou a uma conclusão), o conhecimento é **estruturado e modular** (novas regras podem ser adicionadas sem reformular todo o sistema) e há **garantias formais** sobre a correção das inferências (quando o sistema lógico é consistente e completo). Porém, também possui **limitações significativas**: dificuldade em lidar com **conhecimento incerto ou incompleto**, necessidade de **elicitação manual** do conhecimento (o chamado "gargalo da aquisição de conhecimento"), fragilidade diante de situações não previstas pelas regras e dificuldade em aprender com a experiência.

### 2.2 Exemplo Prático: Sistema de Raciocínio Simbólico

Vamos implementar um sistema simbólico simples que utiliza uma base de fatos e regras de produção (SE...ENTÃO) para realizar inferências. O sistema simulará um pequeno **diagnóstico de problemas em um computador**.

```python
# === Sistema Simbólico: Motor de Inferência com Encadeamento para Frente ===

class SistemaSimbolico:
    """Sistema especialista simples com encadeamento para frente (forward chaining)."""

    def __init__(self):
        self.fatos = set()
        self.regras = []
        self.historico = []

    def adicionar_fato(self, fato):
        """Adiciona um fato à base de conhecimento."""
        self.fatos.add(fato)

    def adicionar_regra(self, condicoes, conclusao, explicacao):
        """Adiciona uma regra SE...ENTÃO ao sistema."""
        self.regras.append({
            'condicoes': condicoes,
            'conclusao': conclusao,
            'explicacao': explicacao
        })

    def inferir(self):
        """Executa o motor de inferência com encadeamento para frente."""
        novas_inferencias = True
        while novas_inferencias:
            novas_inferencias = False
            for regra in self.regras:
                if all(c in self.fatos for c in regra['condicoes']):
                    if regra['conclusao'] not in self.fatos:
                        self.fatos.add(regra['conclusao'])
                        self.historico.append(regra)
                        novas_inferencias = True

    def explicar(self):
        """Mostra a cadeia de raciocínio utilizada."""
        print("=" * 60)
        print("CADEIA DE RACIOCÍNIO:")
        print("=" * 60)
        for i, regra in enumerate(self.historico, 1):
            conds = ' E '.join(regra['condicoes'])
            print(f"\nPasso {i}: SE ({conds})")
            print(f"         ENTÃO: {regra['conclusao']}")
            print(f"         Motivo: {regra['explicacao']}")
        print("\n" + "=" * 60)
        print(f"FATOS FINAIS: {sorted(self.fatos)}")
        print("=" * 60)


# --- Criando o sistema de diagnóstico ---
sistema = SistemaSimbolico()

# Adicionando regras de diagnóstico de computador
sistema.adicionar_regra(
    ['nao_liga'],
    'verificar_energia',
    'Se o computador não liga, deve-se verificar a fonte de energia.'
)
sistema.adicionar_regra(
    ['verificar_energia', 'energia_ok'],
    'problema_placa_mae',
    'Se a energia está OK mas não liga, o problema pode ser na placa-mãe.'
)
sistema.adicionar_regra(
    ['liga', 'tela_azul'],
    'problema_driver',
    'Tela azul frequentemente indica problema de driver ou memória.'
)
sistema.adicionar_regra(
    ['liga', 'lento'],
    'verificar_disco',
    'Lentidão pode indicar problema no disco rígido.'
)
sistema.adicionar_regra(
    ['verificar_disco', 'disco_ok'],
    'verificar_malware',
    'Se o disco está OK, verificar presença de malware.'
)
sistema.adicionar_regra(
    ['verificar_malware', 'malware_encontrado'],
    'diagnostico_malware',
    'Malware encontrado é a causa da lentidão.'
)

# Simulando sintomas reportados pelo usuário
print("SINTOMAS INFORMADOS: computador liga, está lento, disco OK, malware encontrado")
print()

sistema.adicionar_fato('liga')
sistema.adicionar_fato('lento')
sistema.adicionar_fato('disco_ok')
sistema.adicionar_fato('malware_encontrado')

# Executando inferência
sistema.inferir()
sistema.explicar()
```

**Observe** como o sistema segue uma cadeia lógica de raciocínio: a partir dos fatos iniciais (sintomas), as regras são disparadas em sequência, produzindo novos fatos até chegar ao diagnóstico final. Essa **transparência no raciocínio** é uma das maiores vantagens da IA Simbólica — cada conclusão pode ser explicada e justificada.

## 3. Sistemas Conexionistas

### 3.1 Fundamentos Teóricos

A abordagem **conexionista** da Inteligência Artificial inspira-se diretamente no funcionamento do **cérebro humano**, mais especificamente na forma como os neurônios biológicos se conectam e se comunicam por meio de sinapses. A ideia central é que a inteligência emerge não de regras explícitas, mas sim de **padrões de ativação em redes de unidades simples interconectadas**. Cada unidade (neurônio artificial) realiza uma computação elementar, mas o comportamento coletivo da rede é capaz de resolver problemas complexos.

O marco inicial do conexionismo foi o trabalho de **McCulloch e Pitts (1943)**, que propuseram o primeiro modelo matemático de um neurônio artificial. Em 1958, **Frank Rosenblatt** apresentou o **Perceptron**, a primeira rede neural capaz de aprender a partir de exemplos. O Perceptron simples calcula uma soma ponderada de suas entradas, aplica uma **função de ativação** (tipicamente a função degrau) e produz uma saída binária. O aprendizado ocorre pelo **ajuste dos pesos** das conexões, utilizando uma regra de atualização baseada no erro entre a saída desejada e a saída produzida.

Em 1969, **Minsky e Papert** demonstraram em seu livro *Perceptrons* que o Perceptron simples (de uma única camada) não é capaz de resolver problemas que não são **linearmente separáveis**, como a função XOR. Essa limitação causou o primeiro grande declínio no interesse por redes neurais. O campo só foi revitalizado na década de 1980, com o desenvolvimento do algoritmo **backpropagation** (retropropagação do erro) por **Rumelhart, Hinton e Williams (1986)**, que permitiu o treinamento de redes com múltiplas camadas (Perceptron Multicamadas — MLP).

As **vantagens** dos sistemas conexionistas incluem: capacidade de **aprender a partir de dados** (sem necessidade de programação explícita de regras), tolerância a **ruído e dados incompletos**, capacidade de **generalização** (aplicar o conhecimento aprendido a situações novas) e habilidade para resolver problemas não-lineares complexos. As **limitações** incluem: dificuldade de **interpretação** (o conhecimento fica distribuído nos pesos da rede, sendo difícil de explicar — o problema da "caixa preta"), necessidade de **grandes volumes de dados** para treinamento, alto **custo computacional** e sensibilidade à escolha de hiperparâmetros (taxa de aprendizado, arquitetura da rede, etc.).

### 3.2 Exemplo Prático: Perceptron Simples do Zero

Vamos implementar um **Perceptron** do zero, sem utilizar bibliotecas de Machine Learning. Usaremos apenas NumPy para operações vetoriais. O Perceptron será treinado para aprender a operação lógica **AND** e **OR**.

```python
import numpy as np

class Perceptron:
    """Implementação de um Perceptron simples (neurônio artificial)."""

    def __init__(self, n_entradas, taxa_aprendizado=0.1):
        self.pesos = np.zeros(n_entradas)
        self.bias = 0.0
        self.taxa_aprendizado = taxa_aprendizado
        self.historico_erros = []

    def funcao_ativacao(self, soma):
        """Função degrau (step function)."""
        return 1 if soma >= 0 else 0

    def prever(self, entradas):
        """Calcula a saída do Perceptron para uma entrada."""
        soma = np.dot(entradas, self.pesos) + self.bias
        return self.funcao_ativacao(soma)

    def treinar(self, X, y, epocas=20):
        """Treina o Perceptron usando a regra de aprendizado."""
        for epoca in range(epocas):
            erros = 0
            for entradas, rotulo in zip(X, y):
                previsao = self.prever(entradas)
                erro = rotulo - previsao
                # Regra de atualização do Perceptron
                self.pesos += self.taxa_aprendizado * erro * entradas
                self.bias += self.taxa_aprendizado * erro
                erros += abs(erro)
            self.historico_erros.append(erros)
            if erros == 0:
                print(f"  Convergiu na época {epoca + 1}!")
                break


# --- Dados de treinamento: porta lógica AND ---
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])
y_or  = np.array([0, 1, 1, 1])

# Treinando para AND
print("=" * 40)
print("TREINAMENTO: Porta Lógica AND")
print("=" * 40)
p_and = Perceptron(n_entradas=2, taxa_aprendizado=0.1)
p_and.treinar(X, y_and)
print(f"Pesos finais: {p_and.pesos}, Bias: {p_and.bias}")
print("\nResultados:")
for entrada in X:
    print(f"  {entrada} -> {p_and.prever(entrada)}")

# Treinando para OR
print("\n" + "=" * 40)
print("TREINAMENTO: Porta Lógica OR")
print("=" * 40)
p_or = Perceptron(n_entradas=2, taxa_aprendizado=0.1)
p_or.treinar(X, y_or)
print(f"Pesos finais: {p_or.pesos}, Bias: {p_or.bias}")
print("\nResultados:")
for entrada in X:
    print(f"  {entrada} -> {p_or.prever(entrada)}")

# Demonstrando a limitação com XOR
print("\n" + "=" * 40)
print("TENTATIVA: Porta Lógica XOR")
print("=" * 40)
y_xor = np.array([0, 1, 1, 0])
p_xor = Perceptron(n_entradas=2, taxa_aprendizado=0.1)
p_xor.treinar(X, y_xor, epocas=20)
print(f"Pesos finais: {p_xor.pesos}, Bias: {p_xor.bias}")
print("\nResultados (INCORRETOS — XOR não é linearmente separável):")
for i, entrada in enumerate(X):
    pred = p_xor.prever(entrada)
    correto = '✓' if pred == y_xor[i] else '✗'
    print(f"  {entrada} -> {pred} (esperado: {y_xor[i]}) {correto}")
```

**Observe** que o Perceptron aprende com sucesso as funções AND e OR (que são linearmente separáveis), mas **falha** com a função XOR (que não é linearmente separável). Essa limitação, demonstrada por Minsky e Papert em 1969, é uma das mais importantes na história das redes neurais e motivou o desenvolvimento de redes com múltiplas camadas.

## 4. Sistemas Evolucionistas

### 4.1 Fundamentos Teóricos

Os **Sistemas Evolucionistas** (ou **Computação Evolutiva**) constituem uma família de algoritmos de otimização e busca inspirados na **teoria da evolução de Charles Darwin** e nos mecanismos da **genética**. A ideia central é simular o processo de seleção natural: uma **população de soluções candidatas** evolui ao longo de gerações, onde as soluções mais aptas (que melhor resolvem o problema) têm maior probabilidade de sobreviver e se reproduzir, transmitindo suas características para a próxima geração.

Os primeiros trabalhos em computação evolutiva surgiram na década de 1960, com três vertentes principais desenvolvidas de forma independente: as **Estratégias Evolutivas** de Rechenberg e Schwefel (Alemanha), a **Programação Evolutiva** de Lawrence Fogel (EUA) e os **Algoritmos Genéticos** de John Holland (EUA). Holland, em seu livro *Adaptation in Natural and Artificial Systems* (1975), formalizou os princípios dos Algoritmos Genéticos (AGs) e introduziu o **Teorema dos Esquemas**, que fornece uma base teórica para compreender por que os AGs funcionam.

Um **Algoritmo Genético** típico opera da seguinte forma: (1) **Inicialização** — gera-se uma população aleatória de indivíduos, onde cada indivíduo representa uma solução candidata codificada como um "cromossomo" (tipicamente uma string binária ou um vetor de valores); (2) **Avaliação** — cada indivíduo é avaliado por uma **função de aptidão** (fitness) que mede a qualidade da solução; (3) **Seleção** — indivíduos mais aptos são selecionados com maior probabilidade para reprodução; (4) **Cruzamento (crossover)** — pares de indivíduos trocam partes de seus cromossomos para gerar filhos; (5) **Mutação** — pequenas alterações aleatórias são introduzidas nos cromossomos; (6) **Substituição** — a nova geração substitui a anterior. O processo se repete até que um critério de parada seja atingido.

As **vantagens** dos sistemas evolucionistas incluem: não exigem conhecimento sobre o gradiente da função objetivo (são métodos **derivative-free**), são capazes de explorar amplas regiões do espaço de busca (evitando ótimos locais), podem lidar com espaços de busca complexos e discretos, e são facilmente **paralelizáveis**. As **limitações** incluem: não há garantia de encontrar o **ótimo global**, o desempenho depende fortemente da escolha dos **operadores genéticos** e seus parâmetros (tamanho da população, taxas de cruzamento e mutação), e podem ser **computacionalmente custosos** para problemas de grande escala.

### 4.2 Exemplo Prático: Algoritmo Genético para Otimização de Função

Vamos implementar um **Algoritmo Genético** simples para encontrar o valor de **x** que maximiza a função **f(x) = x · sin(10π · x) + 2**, no intervalo [-1, 2]. Essa função possui muitos máximos locais, sendo um bom teste para métodos de otimização global.

```python
import numpy as np
import random

# --- Algoritmo Genético para Otimização de Função ---

def funcao_objetivo(x):
    """Função a ser maximizada: f(x) = x * sin(10 * pi * x) + 2."""
    return x * np.sin(10 * np.pi * x) + 2

def criar_populacao(tamanho, limite_inf, limite_sup):
    """Cria uma população inicial aleatória."""
    return [random.uniform(limite_inf, limite_sup) for _ in range(tamanho)]

def avaliar_populacao(populacao):
    """Avalia a aptidão de cada indivíduo."""
    return [funcao_objetivo(ind) for ind in populacao]

def selecao_torneio(populacao, aptidoes, k=3):
    """Seleciona um indivíduo por torneio."""
    indices = random.sample(range(len(populacao)), k)
    melhor = max(indices, key=lambda i: aptidoes[i])
    return populacao[melhor]

def cruzamento(pai1, pai2, taxa_cruzamento=0.8):
    """Cruzamento aritmético entre dois pais."""
    if random.random() < taxa_cruzamento:
        alpha = random.random()
        filho1 = alpha * pai1 + (1 - alpha) * pai2
        filho2 = (1 - alpha) * pai1 + alpha * pai2
        return filho1, filho2
    return pai1, pai2

def mutacao(individuo, taxa_mutacao=0.1, desvio=0.1, limite_inf=-1, limite_sup=2):
    """Mutação por perturbação gaussiana."""
    if random.random() < taxa_mutacao:
        individuo += random.gauss(0, desvio)
        individuo = max(limite_inf, min(limite_sup, individuo))
    return individuo

def algoritmo_genetico(tam_pop=50, geracoes=100, limite_inf=-1, limite_sup=2):
    """Executa o Algoritmo Genético."""
    populacao = criar_populacao(tam_pop, limite_inf, limite_sup)
    melhor_historico = []

    for geracao in range(geracoes):
        aptidoes = avaliar_populacao(populacao)
        melhor_idx = max(range(tam_pop), key=lambda i: aptidoes[i])
        melhor_historico.append((populacao[melhor_idx], aptidoes[melhor_idx]))

        nova_populacao = [populacao[melhor_idx]]  # Elitismo

        while len(nova_populacao) < tam_pop:
            pai1 = selecao_torneio(populacao, aptidoes)
            pai2 = selecao_torneio(populacao, aptidoes)
            filho1, filho2 = cruzamento(pai1, pai2)
            filho1 = mutacao(filho1, limite_inf=limite_inf, limite_sup=limite_sup)
            filho2 = mutacao(filho2, limite_inf=limite_inf, limite_sup=limite_sup)
            nova_populacao.extend([filho1, filho2])

        populacao = nova_populacao[:tam_pop]

    return melhor_historico


# Executando o AG
random.seed(42)
np.random.seed(42)

historico = algoritmo_genetico(tam_pop=50, geracoes=80)

melhor_x, melhor_fitness = max(historico, key=lambda h: h[1])
print("=" * 50)
print("RESULTADO DO ALGORITMO GENÉTICO")
print("=" * 50)
print(f"Função: f(x) = x * sin(10π * x) + 2")
print(f"Intervalo de busca: [-1, 2]")
print(f"\nMelhor x encontrado: {melhor_x:.6f}")
print(f"Melhor f(x): {melhor_fitness:.6f}")
print(f"\nEvolução do fitness ao longo das gerações:")
for g in [0, 9, 19, 39, 59, 79]:
    if g < len(historico):
        print(f"  Geração {g+1:3d}: x = {historico[g][0]:8.5f}, f(x) = {historico[g][1]:.5f}")
```

**Observe** como a população evolui ao longo das gerações, convergindo progressivamente para a região do máximo global da função. O uso de **elitismo** (manter o melhor indivíduo), **seleção por torneio**, **cruzamento aritmético** e **mutação gaussiana** garante um bom equilíbrio entre **exploração** (buscar novas regiões) e **explotação** (refinar soluções promissoras).

## 5. Sistemas Probabilísticos

### 5.1 Fundamentos Teóricos

Os **Sistemas Probabilísticos** na Inteligência Artificial fundamentam-se na **Teoria das Probabilidades** e na **Estatística** para representar e raciocinar sobre **incerteza**. O mundo real é inerentemente incerto: sensores produzem medidas ruidosas, informações podem ser incompletas, e as relações entre variáveis frequentemente não são determinísticas. A IA Probabilística oferece um arcabouço matemático rigoroso para lidar com esse tipo de incerteza.

O alicerce dos sistemas probabilísticos é o **Teorema de Bayes**, formulado pelo reverendo **Thomas Bayes** no século XVIII e formalizado por **Pierre-Simon Laplace**. O teorema permite atualizar a probabilidade de uma hipótese à luz de novas evidências:

$$P(H|E) = \frac{P(E|H) \cdot P(H)}{P(E)}$$

Onde: **P(H|E)** é a probabilidade posterior (da hipótese H dado a evidência E), **P(E|H)** é a verossimilhança (probabilidade de observar E se H for verdadeira), **P(H)** é a probabilidade a priori da hipótese, e **P(E)** é a probabilidade marginal da evidência.

As **Redes Bayesianas**, propostas por **Judea Pearl** na década de 1980, são grafos acíclicos dirigidos que representam dependências probabilísticas entre variáveis. Cada nó representa uma variável aleatória, e as arestas representam influências diretas entre variáveis. Associada a cada nó está uma tabela de probabilidade condicional. As Redes Bayesianas permitem realizar inferência probabilística eficiente, calculando a probabilidade de qualquer variável dado um conjunto de observações. O **classificador Naive Bayes** é um caso simplificado em que se assume independência condicional entre as variáveis de entrada, dado a classe.

As **vantagens** dos sistemas probabilísticos incluem: tratamento formal e rigoroso da **incerteza**, capacidade de **atualizar crenças** à medida que novas evidências são obtidas, fundamentação matemática sólida e desempenho robusto mesmo com **dados incompletos ou ruidosos**. As **limitações** incluem: a necessidade de estimar ou definir **distribuições de probabilidade a priori** (que podem ser subjetivas), o custo computacional da **inferência exata** em redes complexas (que é NP-difícil em geral), e a dificuldade em modelar **dependências temporais** complexas sem extensões específicas (como os Modelos Ocultos de Markov ou Filtros de Partículas).

### 5.2 Exemplo Prático: Classificador Naive Bayes do Zero

Vamos implementar um **classificador Naive Bayes** do zero para classificar se um e-mail é **spam** ou **não-spam** com base na presença de determinadas palavras-chave. Utilizaremos o Teorema de Bayes com a hipótese de independência condicional entre os atributos.

```python
from collections import defaultdict

class NaiveBayesClassificador:
    """Classificador Naive Bayes implementado do zero."""

    def __init__(self):
        self.prob_classe = {}          # P(classe)
        self.prob_atributo_classe = {} # P(atributo | classe)
        self.classes = set()

    def treinar(self, dados, rotulos):
        """Treina o classificador com os dados de treinamento."""
        n_total = len(rotulos)
        self.classes = set(rotulos)

        # Calcular P(classe)
        contagem_classes = defaultdict(int)
        for r in rotulos:
            contagem_classes[r] += 1
        self.prob_classe = {c: contagem_classes[c] / n_total for c in self.classes}

        # Calcular P(atributo=valor | classe) com suavização de Laplace
        self.prob_atributo_classe = {}
        n_atributos = len(dados[0])

        # Pré-calcular vocabulário por atributo (uma vez, fora do loop de classes)
        self.vocab_por_atributo = {}
        for j in range(n_atributos):
            self.vocab_por_atributo[j] = set(d[j] for d in dados)

        # Armazenar n_classe para uso na classificação
        self.n_classe = {}

        for classe in self.classes:
            self.prob_atributo_classe[classe] = {}
            indices_classe = [i for i, r in enumerate(rotulos) if r == classe]
            n_classe = len(indices_classe)
            self.n_classe[classe] = n_classe

            for j in range(n_atributos):
                valores = [dados[i][j] for i in indices_classe]
                contagem_valores = defaultdict(int)
                for v in valores:
                    contagem_valores[v] += 1

                # Suavização de Laplace (add-1 smoothing)
                V = len(self.vocab_por_atributo[j])
                self.prob_atributo_classe[classe][j] = {}
                for v in self.vocab_por_atributo[j]:
                    self.prob_atributo_classe[classe][j][v] = (
                        (contagem_valores[v] + 1) / (n_classe + V)
                    )

    def classificar(self, amostra):
        """Classifica uma nova amostra."""
        probabilidades = {}
        for classe in self.classes:
            prob = self.prob_classe[classe]
            for j, valor in enumerate(amostra):
                if valor in self.prob_atributo_classe[classe][j]:
                    prob *= self.prob_atributo_classe[classe][j][valor]
                else:
                    # Suavização de Laplace para valores não vistos:
                    # (0 + 1) / (n_classe + V)
                    V = len(self.vocab_por_atributo[j])
                    prob *= 1 / (self.n_classe[classe] + V)
            probabilidades[classe] = prob

        # Normalizar
        soma = sum(probabilidades.values())
        prob_normalizada = {c: p / soma for c, p in probabilidades.items()}
        return prob_normalizada

# --- Dados de treinamento: classificação de e-mails ---
# Atributos: [tem_oferta, tem_gratis, tem_urgente, tem_reuniao, tem_relatorio]
# 1 = palavra presente, 0 = ausente

dados_treino = [
    [1, 1, 1, 0, 0],  # spam
    [1, 1, 0, 0, 0],  # spam
    [1, 0, 1, 0, 0],  # spam
    [0, 1, 1, 0, 0],  # spam
    [1, 1, 1, 0, 1],  # spam
    [0, 0, 0, 1, 1],  # não-spam
    [0, 0, 0, 1, 0],  # não-spam
    [0, 0, 1, 1, 1],  # não-spam
    [0, 0, 0, 0, 1],  # não-spam
    [1, 0, 0, 1, 1],  # não-spam
]
rotulos_treino = ['spam', 'spam', 'spam', 'spam', 'spam',
                  'não-spam', 'não-spam', 'não-spam', 'não-spam', 'não-spam']

atributos_nomes = ['tem_oferta', 'tem_gratis', 'tem_urgente', 'tem_reuniao', 'tem_relatorio']

# Treinando o classificador
nb = NaiveBayesClassificador()
nb.treinar(dados_treino, rotulos_treino)

print("=" * 60)
print("CLASSIFICADOR NAIVE BAYES — Detecção de Spam")
print("=" * 60)
print(f"\nProbabilidades a priori: {nb.prob_classe}")

# Classificando novos e-mails
testes = [
    ([1, 1, 1, 0, 0], "E-mail com oferta, grátis e urgente"),
    ([0, 0, 0, 1, 1], "E-mail com reunião e relatório"),
    ([1, 0, 0, 1, 0], "E-mail com oferta e reunião"),
    ([0, 1, 1, 0, 0], "E-mail com grátis e urgente"),
]

print("\n" + "-" * 60)
print("CLASSIFICAÇÃO DE NOVOS E-MAILS:")
print("-" * 60)
for amostra, descricao in testes:
    resultado = nb.classificar(amostra)
    classe_pred = max(resultado, key=resultado.get)
    print(f"\n{descricao}")
    print(f"  Atributos: {dict(zip(atributos_nomes, amostra))}")
    for classe, prob in sorted(resultado.items()):
        print(f"  P({classe}) = {prob:.4f}")
    print(f"  → Classificação: {classe_pred.upper()}")
```

**Observe** como o classificador utiliza o Teorema de Bayes para calcular a probabilidade de cada classe dado os atributos observados. A **suavização de Laplace** é utilizada para evitar que probabilidades zero (quando um valor de atributo não aparece nos dados de treinamento de uma classe) anulem toda a probabilidade. Apesar da hipótese simplificadora de independência condicional, o Naive Bayes é surpreendentemente eficaz em muitas aplicações práticas, como filtragem de spam e classificação de textos.

## 6. Comparação entre as Abordagens

A tabela a seguir apresenta uma comparação abrangente entre as quatro abordagens da IA, considerando diversas dimensões de análise:

| Característica | Simbólica | Conexionista | Evolucionista | Probabilística |
|---|---|---|---|---|
| **Inspiração** | Lógica e raciocínio formal | Cérebro humano (neurônios) | Evolução biológica | Teoria das probabilidades |
| **Representação do conhecimento** | Símbolos, regras, lógica | Pesos em redes neurais | Cromossomos (populações) | Distribuições de probabilidade |
| **Aprendizado** | Programado por especialistas | A partir de dados (exemplos) | Por evolução de populações | A partir de dados + priors |
| **Transparência** | Alta (explicável) | Baixa (caixa preta) | Média | Média-alta |
| **Tratamento de incerteza** | Fraco | Moderado (implícito) | Não é o foco | Excelente |
| **Generalização** | Limitada às regras | Boa para padrões similares | Boa em espaços de busca | Boa com dados suficientes |
| **Escalabilidade** | Limitada | Alta (com hardware) | Moderada | Moderada |
| **Dados necessários** | Conhecimento de especialistas | Grandes volumes de dados | Função de aptidão | Dados + distribuições a priori |
| **Exemplos clássicos** | MYCIN, Prolog, CLIPS | MLP, CNN, RNN, Transformers | AG, Programação Genética | Redes Bayesianas, HMMs |
| **Aplicações típicas** | Sistemas especialistas, planejamento | Visão computacional, NLP | Otimização, design | Diagnóstico, filtro de spam |
| **Período de predominância** | 1956–1980 | 1986–presente | 1975–presente | 1990–presente |

### Pontos-chave da comparação:

- **IA Simbólica** é a mais transparente e explicável, mas depende de especialistas humanos para codificar o conhecimento e tem dificuldade com incerteza.
- **IA Conexionista** é a mais poderosa para reconhecimento de padrões e aprendizado a partir de dados, mas é uma "caixa preta" e exige muitos dados.
- **IA Evolucionista** é excelente para problemas de otimização sem gradiente, mas pode ser lenta e não garante o ótimo global.
- **IA Probabilística** é a mais adequada para raciocinar sob incerteza, mas pode ser computacionalmente cara e requer escolha cuidadosa de priors.

## 7. Abordagens Híbridas

### A Convergência dos Paradigmas

Na IA contemporânea, a tendência predominante é a **combinação** dos diferentes paradigmas em **sistemas híbridos** que aproveitam os pontos fortes de cada abordagem e compensam suas limitações. A dicotomia entre simbólico e conexionista, que dominou os debates nas décadas de 1980 e 1990, deu lugar a uma visão mais integradora.

Alguns exemplos notáveis de abordagens híbridas incluem:

- **Sistemas Neuro-Simbólicos:** Combinam redes neurais (para aprendizado e reconhecimento de padrões) com raciocínio simbólico (para inferência lógica e explicabilidade). Exemplos incluem o **Neural Theorem Prover** e trabalhos recentes do grupo de **Yoshua Bengio** e **Gary Marcus** sobre integração neuro-simbólica.

- **Neuroevolução:** Utiliza algoritmos genéticos para evoluir a **arquitetura e os pesos** de redes neurais, em vez de utilizar apenas backpropagation. Exemplos incluem o **NEAT (NeuroEvolution of Augmenting Topologies)** de Kenneth Stanley.

- **Redes Neurais Bayesianas:** Combinam redes neurais com inferência probabilística, permitindo que os pesos da rede sejam modelados como distribuições de probabilidade em vez de valores fixos, o que fornece estimativas de **incerteza** nas previsões.

- **Sistemas Especialistas Probabilísticos:** Integram regras simbólicas com raciocínio probabilístico, como nos **Sistemas Especialistas Fuzzy** e nas **Redes Bayesianas com variáveis lógicas**.

- **Grandes Modelos de Linguagem (LLMs):** Modelos como GPT e BERT são fundamentalmente conexionistas, mas recentes técnicas como **Chain-of-Thought**, **Retrieval-Augmented Generation (RAG)** e **tool-use** integram elementos de raciocínio simbólico e busca estruturada.

A busca por sistemas de IA que sejam ao mesmo tempo **poderosos** (capazes de aprender com dados), **explicáveis** (capazes de justificar suas decisões), **robustos** (capazes de lidar com incerteza) e **eficientes** (capazes de otimizar soluções) continuará impulsionando a integração desses paradigmas no futuro.
