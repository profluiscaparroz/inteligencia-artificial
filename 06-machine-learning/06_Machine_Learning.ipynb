## 1. O que é Machine Learning?

**Machine Learning (Aprendizado de Máquina)** é um subcampo da Inteligência Artificial que se concentra no desenvolvimento de algoritmos e modelos que permitem aos computadores aprender padrões a partir de dados, sem serem explicitamente programados para cada tarefa específica.

Segundo **Arthur Samuel** (1959), pioneiro da área, Machine Learning é definido como:

> *"O campo de estudo que dá aos computadores a habilidade de aprender sem serem explicitamente programados."*

Uma definição mais moderna e formal, proposta por **Tom Mitchell** (1997), estabelece:

> *"Um programa de computador aprende a partir da experiência E em relação a alguma classe de tarefas T e medida de desempenho P, se seu desempenho nas tarefas T, medido por P, melhora com a experiência E."*

Essa definição formal estabelece três componentes fundamentais:
- **Tarefa (T):** O problema a ser resolvido (classificação, regressão, agrupamento, etc.)
- **Experiência (E):** Os dados utilizados para aprendizado
- **Medida de desempenho (P):** Métrica para avaliar o quão bem o modelo aprendeu

### 1.1 Breve História do Machine Learning

A evolução do Machine Learning está intimamente ligada ao desenvolvimento da própria Inteligência Artificial:

- **1943 — McCulloch & Pitts:** Propuseram o primeiro modelo matemático de neurônio artificial, estabelecendo as bases teóricas para redes neurais.

- **1950 — Alan Turing:** No artigo *"Computing Machinery and Intelligence"*, sugeriu que máquinas poderiam aprender através de recompensas e punições, antecipando o Aprendizado por Reforço.

- **1957 — Frank Rosenblatt:** Desenvolveu o **Perceptron**, primeiro algoritmo de aprendizado de rede neural capaz de aprender padrões lineares. Demonstrado em hardware, causou grande entusiasmo inicial.

- **1959 — Arthur Samuel:** Criou o termo *"Machine Learning"* ao desenvolver um programa de damas que melhorava seu desempenho jogando contra si mesmo.

- **1969 — Minsky & Papert:** Publicaram *"Perceptrons"*, demonstrando limitações matemáticas do Perceptron (incapaz de resolver XOR), contribuindo para o primeiro "inverno da IA".

- **1986 — Rumelhart, Hinton & Williams:** Popularizaram o algoritmo de **Backpropagation**, permitindo treinar redes neurais multicamadas (MLPs), superando as limitações do Perceptron simples.

- **1990s — Support Vector Machines (SVMs):** Vladimir Vapnik e colegas desenvolveram SVMs, algoritmos baseados em teoria estatística rigorosa que dominaram aplicações práticas por décadas.

- **1997 — Random Forests:** Leo Breiman introduziu Random Forests, um método de ensemble que combina múltiplas árvores de decisão, tornando-se um dos algoritmos mais práticos e efetivos.

- **2006 — Deep Learning:** Geoffrey Hinton popularizou técnicas de aprendizado profundo (Deep Learning), usando redes neurais com muitas camadas, revolucionando visão computacional, processamento de linguagem natural e outras áreas.

- **2012 — AlexNet:** Rede neural convolucional profunda venceu a competição ImageNet com margem significativa, marcando o início da era moderna do Deep Learning.

- **Anos 2010-2020 — Explosão do ML:** Combinação de grandes volumes de dados (Big Data), poder computacional (GPUs), frameworks maduros (TensorFlow, PyTorch) e algoritmos avançados levou ML a aplicações práticas massivas: assistentes virtuais, veículos autônomos, diagnóstico médico, recomendação de conteúdo, etc.

### 1.2 Paradigmas de Aprendizado

O Machine Learning pode ser categorizado em diferentes paradigmas, dependendo da natureza dos dados e do tipo de feedback disponível durante o aprendizado:

#### 1.2.1 Aprendizado Supervisionado (*Supervised Learning*)

No aprendizado supervisionado, o algoritmo aprende a partir de exemplos rotulados — pares **(entrada, saída desejada)**. O objetivo é encontrar uma função que mapeia entradas para saídas corretas, generalizando para novos dados não vistos.

**Tarefas principais:**
- **Classificação:** Prever categoria discreta (ex: spam/não-spam, doença/saudável)
- **Regressão:** Prever valor contínuo (ex: preço de imóvel, temperatura)

**Algoritmos clássicos:** Regressão Linear/Logística, K-Nearest Neighbors (KNN), Árvores de Decisão, Random Forests, Support Vector Machines (SVM), Redes Neurais.

**Exemplo:** Classificar e-mails como spam ou não-spam, dados históricos de e-mails já rotulados.

#### 1.2.2 Aprendizado Não-Supervisionado (*Unsupervised Learning*)

Neste paradigma, o algoritmo trabalha com dados **não rotulados**, buscando descobrir estruturas, padrões ou relações ocultas nos dados.

**Tarefas principais:**
- **Agrupamento (Clustering):** Agrupar dados similares (ex: segmentação de clientes)
- **Redução de Dimensionalidade:** Simplificar dados mantendo informação essencial (ex: PCA)
- **Detecção de Anomalias:** Identificar outliers ou comportamentos atípicos

**Algoritmos clássicos:** K-Means, DBSCAN, Hierarchical Clustering, Principal Component Analysis (PCA), Autoencoders.

**Exemplo:** Agrupar clientes por padrões de compra sem categorias predefinidas.

#### 1.2.3 Aprendizado por Reforço (*Reinforcement Learning*)

O agente aprende através de **interação com um ambiente**, recebendo recompensas ou punições por suas ações. O objetivo é aprender uma política (estratégia) que maximize a recompensa acumulada ao longo do tempo.

**Componentes:** Agente, Ambiente, Estado, Ação, Recompensa, Política.

**Algoritmos clássicos:** Q-Learning, Deep Q-Networks (DQN), Policy Gradients, Actor-Critic, AlphaGo/AlphaZero.

**Exemplo:** Ensinar um robô a caminhar, jogar jogos (xadrez, Go, videogames).

#### 1.2.4 Aprendizado Semi-Supervisionado e Outros

- **Semi-Supervisionado:** Combina pequena quantidade de dados rotulados com grande quantidade de dados não rotulados.
- **Auto-Supervisionado:** Cria rótulos automaticamente a partir da estrutura dos dados (ex: predição de próxima palavra em texto).
- **Transfer Learning:** Reutiliza conhecimento aprendido em uma tarefa para resolver outra relacionada.

### 1.3 O Processo de Machine Learning

Um projeto típico de Machine Learning segue estas etapas:

1. **Definição do Problema:** Identificar a tarefa (classificação, regressão, etc.) e métricas de sucesso.

2. **Coleta e Exploração de Dados:** Reunir dados relevantes, entender suas características, distribuições e qualidade.

3. **Pré-processamento de Dados:**
   - Limpeza (tratar valores ausentes, outliers)
   - Transformação (normalização, codificação de variáveis categóricas)
   - Engenharia de features (criar novas features relevantes)

4. **Divisão dos Dados:** Separar dados em conjuntos de treino, validação e teste (ex: 70%-15%-15%).

5. **Escolha e Treinamento do Modelo:** Selecionar algoritmo apropriado e ajustar seus parâmetros usando dados de treino.

6. **Avaliação:** Medir desempenho em dados de validação/teste usando métricas apropriadas (acurácia, precisão, recall, RMSE, etc.).

7. **Ajuste de Hiperparâmetros:** Otimizar configurações do modelo (ex: taxa de aprendizado, profundidade de árvore).

8. **Implantação:** Colocar modelo em produção para fazer predições em dados reais.

9. **Monitoramento e Manutenção:** Avaliar desempenho contínuo e retreinar quando necessário (concept drift).

## 2. Regressão Linear — Fundamento do Aprendizado Supervisionado

A **Regressão Linear** é um dos algoritmos mais fundamentais em Machine Learning, usado para modelar a relação entre uma variável dependente contínua (alvo) e uma ou mais variáveis independentes (features).

### 2.1 Formulação Matemática

Para regressão linear simples (uma variável independente), buscamos uma função linear:

$$y = \theta_0 + \theta_1 x$$

Onde:
- $y$ é o valor predito
- $x$ é a feature (variável independente)
- $\theta_0$ é o intercepto (bias)
- $\theta_1$ é o coeficiente angular (peso)

Para regressão linear múltipla (múltiplas features):

$$y = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... + \theta_n x_n = \theta^T x$$

### 2.2 Função de Custo

Para encontrar os melhores parâmetros $\theta$, minimizamos o **Erro Quadrático Médio (MSE)**:

$$J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)})^2$$

Onde:
- $m$ é o número de exemplos de treino
- $h_\theta(x^{(i)})$ é a predição para o exemplo $i$
- $y^{(i)}$ é o valor real

### 2.3 Gradiente Descendente

O algoritmo de **Gradiente Descendente** atualiza iterativamente os parâmetros na direção que reduz o custo:

$$\theta_j := \theta_j - \alpha \frac{\partial J(\theta)}{\partial \theta_j}$$

Onde $\alpha$ é a taxa de aprendizado (learning rate).

Para regressão linear, a derivada parcial é:

$$\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)}$$

### 2.4 Implementação de Regressão Linear do Zero

```python
import numpy as np
import matplotlib.pyplot as plt

class RegressaoLinear:
    """
    Implementação de Regressão Linear usando Gradiente Descendente.
    
    Parâmetros:
    -----------
    learning_rate : float
        Taxa de aprendizado (alpha) para gradiente descendente
    n_iterations : int
        Número de iterações do gradiente descendente
    """
    
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta = None
        self.historico_custo = []
    
    def fit(self, X, y):
        """
        Treina o modelo de regressão linear.
        
        Parâmetros:
        -----------
        X : array-like, shape (n_samples, n_features)
            Features de treinamento
        y : array-like, shape (n_samples,)
            Valores alvo
        """
        # Adiciona coluna de 1s para o termo de intercepto (bias)
        m, n = X.shape
        X_bias = np.c_[np.ones((m, 1)), X]
        
        # Inicializa parâmetros aleatoriamente
        self.theta = np.random.randn(n + 1, 1)
        
        # Gradiente Descendente
        for i in range(self.n_iterations):
            # Predição
            y_pred = X_bias.dot(self.theta)
            
            # Erro
            erro = y_pred - y.reshape(-1, 1)
            
            # Custo (MSE)
            custo = (1/(2*m)) * np.sum(erro ** 2)
            self.historico_custo.append(custo)
            
            # Gradiente
            gradiente = (1/m) * X_bias.T.dot(erro)
            
            # Atualização dos parâmetros
            self.theta -= self.learning_rate * gradiente
            
            if i % 100 == 0:
                print(f"Iteração {i}: Custo = {custo:.4f}")
    
    def predict(self, X):
        """
        Faz predições usando o modelo treinado.
        
        Parâmetros:
        -----------
        X : array-like, shape (n_samples, n_features)
            Features para predição
            
        Retorna:
        --------
        y_pred : array, shape (n_samples,)
            Valores preditos
        """
        X_bias = np.c_[np.ones((X.shape[0], 1)), X]
        return X_bias.dot(self.theta).flatten()
    
    def score(self, X, y):
        """
        Calcula o coeficiente de determinação R².
        
        R² = 1 - (SS_res / SS_tot)
        onde SS_res é a soma dos quadrados dos resíduos e
        SS_tot é a soma total dos quadrados.
        """
        y_pred = self.predict(X)
        ss_res = np.sum((y - y_pred) ** 2)
        ss_tot = np.sum((y - np.mean(y)) ** 2)
        return 1 - (ss_res / ss_tot)

# Exemplo de uso
print("="*60)
print("REGRESSÃO LINEAR - IMPLEMENTAÇÃO DO ZERO")
print("="*60)

# Gera dados sintéticos
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X.flatten() + np.random.randn(100)

print("\nDados gerados: y = 4 + 3*x + ruído")
print(f"Shape de X: {X.shape}")
print(f"Shape de y: {y.shape}")

# Treina o modelo
modelo = RegressaoLinear(learning_rate=0.1, n_iterations=1000)
modelo.fit(X, y)

print("\n" + "-"*60)
print("RESULTADOS DO TREINAMENTO")
print("-"*60)
print(f"\nParâmetros aprendidos:")
print(f"  θ₀ (intercepto) = {modelo.theta[0][0]:.4f}")
print(f"  θ₁ (coeficiente) = {modelo.theta[1][0]:.4f}")
print(f"\nParâmetros reais: θ₀ = 4.0000, θ₁ = 3.0000")

# Avaliação
r2 = modelo.score(X, y)
print(f"\nR² Score: {r2:.4f}")

# Visualização
plt.figure(figsize=(12, 5))

# Gráfico 1: Dados e linha de regressão
plt.subplot(1, 2, 1)
plt.scatter(X, y, alpha=0.5, label='Dados')
plt.plot(X, modelo.predict(X), 'r-', linewidth=2, label='Modelo aprendido')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regressão Linear')
plt.legend()
plt.grid(True, alpha=0.3)

# Gráfico 2: Convergência do custo
plt.subplot(1, 2, 2)
plt.plot(modelo.historico_custo)
plt.xlabel('Iteração')
plt.ylabel('Custo (MSE)')
plt.title('Convergência do Gradiente Descendente')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
```

**Observações sobre a implementação:**

1. **Adição de bias:** Adicionamos uma coluna de 1s à matriz X para incorporar o termo de intercepto ($\theta_0$) diretamente na multiplicação matricial.

2. **Inicialização:** Os parâmetros são inicializados aleatoriamente. A escolha da inicialização pode afetar a convergência.

3. **Taxa de aprendizado:** Um valor muito alto pode causar divergência; muito baixo torna o treinamento lento. A escolha de 0.1 foi adequada para este problema.

4. **Convergência:** O gráfico de custo mostra que o algoritmo converge rapidamente nas primeiras iterações e depois estabiliza.

5. **R² Score:** Próximo de 1.0 indica bom ajuste. Valores negativos indicariam que o modelo é pior que simplesmente prever a média.

### 2.5 Implementação com Scikit-learn

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

print("="*60)
print("REGRESSÃO LINEAR - USANDO SCIKIT-LEARN")
print("="*60)

# Usa os mesmos dados gerados anteriormente
modelo_sklearn = LinearRegression()
modelo_sklearn.fit(X, y)

# Predições
y_pred = modelo_sklearn.predict(X)

print("\nParâmetros aprendidos (Scikit-learn):")
print(f"  θ₀ (intercepto) = {modelo_sklearn.intercept_:.4f}")
print(f"  θ₁ (coeficiente) = {modelo_sklearn.coef_[0]:.4f}")

# Métricas
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"\nMétricas de desempenho:")
print(f"  MSE (Erro Quadrático Médio): {mse:.4f}")
print(f"  R² Score: {r2:.4f}")

print("\n" + "="*60)
print("\n✓ A biblioteca Scikit-learn usa métodos de otimização mais")
print("  eficientes (equação normal ou decomposição SVD), resultando")
print("  em parâmetros praticamente idênticos aos da nossa implementação.")
```

## 3. K-Nearest Neighbors (KNN) — Aprendizado Baseado em Instâncias

O algoritmo **K-Nearest Neighbors (K-Vizinhos Mais Próximos)** é um método de aprendizado supervisionado não-paramétrico, usado tanto para classificação quanto regressão.

### 3.1 Princípio Fundamental

O KNN baseia-se na ideia de que **instâncias similares tendem a ter rótulos similares**. Para classificar um novo exemplo:

1. Calcula-se a distância entre o novo exemplo e todos os exemplos de treino
2. Selecionam-se os K exemplos mais próximos (vizinhos)
3. Para **classificação**: atribui-se a classe majoritária entre os K vizinhos
4. Para **regressão**: calcula-se a média dos valores dos K vizinhos

### 3.2 Métricas de Distância

A escolha da métrica de distância é crucial:

**Distância Euclidiana** (mais comum):
$$d(x, x') = \sqrt{\sum_{i=1}^{n} (x_i - x'_i)^2}$$

**Distância de Manhattan:**
$$d(x, x') = \sum_{i=1}^{n} |x_i - x'_i|$$

**Distância de Minkowski** (generalização):
$$d(x, x') = \left(\sum_{i=1}^{n} |x_i - x'_i|^p\right)^{1/p}$$

### 3.3 Escolha de K

- **K pequeno (ex: 1):** Sensível a ruído, fronteiras de decisão complexas, overfitting
- **K grande:** Fronteiras suaves, mais robusto a ruído, mas pode ignorar padrões locais (underfitting)
- **Regra prática:** K ímpar para classificação binária (evita empates), K ≈ √n é um bom ponto de partida
- **Validação cruzada** deve ser usada para escolher o melhor K

### 3.4 Implementação de KNN do Zero

```python
import numpy as np
from collections import Counter

class KNNClassificador:
    """
    Implementação do algoritmo K-Nearest Neighbors para classificação.
    
    Parâmetros:
    -----------
    k : int
        Número de vizinhos a considerar
    metrica : str
        Métrica de distância ('euclidiana' ou 'manhattan')
    """
    
    def __init__(self, k=3, metrica='euclidiana'):
        self.k = k
        self.metrica = metrica
        self.X_treino = None
        self.y_treino = None
    
    def fit(self, X, y):
        """
        'Treina' o modelo armazenando os dados de treino.
        KNN é um algoritmo lazy learning - não há treinamento real.
        """
        self.X_treino = X
        self.y_treino = y
    
    def _calcular_distancia(self, x1, x2):
        """
        Calcula distância entre dois pontos.
        """
        if self.metrica == 'euclidiana':
            return np.sqrt(np.sum((x1 - x2) ** 2))
        elif self.metrica == 'manhattan':
            return np.sum(np.abs(x1 - x2))
        else:
            raise ValueError(f"Métrica '{self.metrica}' não suportada")
    
    def _predizer_um(self, x):
        """
        Prediz a classe de um único exemplo.
        """
        # Calcula distâncias para todos os pontos de treino
        distancias = [self._calcular_distancia(x, x_treino) 
                     for x_treino in self.X_treino]
        
        # Obtém índices dos K vizinhos mais próximos
        indices_k_vizinhos = np.argsort(distancias)[:self.k]
        
        # Obtém rótulos dos K vizinhos
        rotulos_k_vizinhos = [self.y_treino[i] for i in indices_k_vizinhos]
        
        # Retorna classe mais comum (voto majoritário)
        classe_mais_comum = Counter(rotulos_k_vizinhos).most_common(1)[0][0]
        return classe_mais_comum
    
    def predict(self, X):
        """
        Prediz classes para múltiplos exemplos.
        """
        predicoes = [self._predizer_um(x) for x in X]
        return np.array(predicoes)
    
    def score(self, X, y):
        """
        Calcula acurácia das predições.
        """
        predicoes = self.predict(X)
        return np.mean(predicoes == y)

# Exemplo de uso com dataset Iris simulado
print("="*60)
print("K-NEAREST NEIGHBORS - IMPLEMENTAÇÃO DO ZERO")
print("="*60)

# Gera dados sintéticos de classificação (3 classes)
np.random.seed(42)
from sklearn.datasets import make_classification

X, y = make_classification(n_samples=150, n_features=2, n_informative=2,
                          n_redundant=0, n_clusters_per_class=1,
                          n_classes=3, random_state=42)

# Divide em treino e teste
from sklearn.model_selection import train_test_split
X_treino, X_teste, y_treino, y_teste = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"\nDados de treino: {X_treino.shape[0]} amostras")
print(f"Dados de teste: {X_teste.shape[0]} amostras")
print(f"Número de classes: {len(np.unique(y))}")

# Testa diferentes valores de K
print("\n" + "-"*60)
print("TESTANDO DIFERENTES VALORES DE K")
print("-"*60)

for k in [1, 3, 5, 7, 9]:
    knn = KNNClassificador(k=k, metrica='euclidiana')
    knn.fit(X_treino, y_treino)
    
    acuracia_treino = knn.score(X_treino, y_treino)
    acuracia_teste = knn.score(X_teste, y_teste)
    
    print(f"\nK = {k}:")
    print(f"  Acurácia (treino): {acuracia_treino:.4f}")
    print(f"  Acurácia (teste):  {acuracia_teste:.4f}")

# Visualização das fronteiras de decisão
def plotar_fronteiras(X, y, modelo, titulo):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    Z = modelo.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='viridis', s=50)
    plt.title(titulo)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.figure(figsize=(15, 5))

for i, k in enumerate([1, 3, 9], 1):
    plt.subplot(1, 3, i)
    knn = KNNClassificador(k=k)
    knn.fit(X_treino, y_treino)
    plotar_fronteiras(X_treino, y_treino, knn, f'KNN com K={k}')

plt.tight_layout()
plt.show()

print("\n" + "="*60)
```

**Observações sobre KNN:**

1. **Lazy Learning:** KNN não constrói modelo explícito durante treino — apenas armazena dados. Toda computação ocorre na predição.

2. **Sensibilidade a K:** K=1 produz fronteiras muito complexas (overfitting), enquanto K grande suaviza demais (underfitting).

3. **Complexidade:** Predição é O(n·d) onde n é tamanho do treino e d é número de dimensões — pode ser lento para grandes datasets.

4. **Normalização crucial:** Features com escalas diferentes dominam o cálculo de distância. Sempre normalize dados antes de usar KNN.

5. **Maldição da dimensionalidade:** Performance degrada em espaços de alta dimensão, pois conceito de "proximidade" perde significado.

### 3.5 Implementação com Scikit-learn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("="*60)
print("K-NEAREST NEIGHBORS - USANDO SCIKIT-LEARN")
print("="*60)

# Treina modelo com K=5
knn_sklearn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_sklearn.fit(X_treino, y_treino)

# Predições
y_pred = knn_sklearn.predict(X_teste)

# Métricas detalhadas
print(f"\nAcurácia: {accuracy_score(y_teste, y_pred):.4f}")

print("\nRelatório de Classificação:")
print(classification_report(y_teste, y_pred))

print("\nMatriz de Confusão:")
print(confusion_matrix(y_teste, y_pred))

print("\n" + "="*60)
print("\n✓ Scikit-learn oferece otimizações como algoritmos de busca")
print("  espacial (KD-Tree, Ball Tree) que aceleram significativamente")
print("  a busca por vizinhos em datasets grandes.")
```

## 4. Árvores de Decisão — Aprendizado Baseado em Regras

**Árvores de Decisão** são modelos de aprendizado supervisionado que realizam predições através de uma série de perguntas sobre as features, organizadas em estrutura hierárquica de árvore.

### 4.1 Estrutura e Funcionamento

Uma árvore de decisão consiste em:
- **Nó raiz:** Primeiro teste sobre features
- **Nós internos:** Testes sobre features específicas
- **Ramos:** Representam resultados dos testes
- **Folhas:** Contêm predições finais (classes ou valores)

### 4.2 Algoritmo de Construção — Divisão Recursiva

O algoritmo CART (Classification and Regression Trees) constrói a árvore recursivamente:

1. Seleciona melhor feature e ponto de corte para dividir dados
2. Divide dados em subconjuntos com base no critério escolhido
3. Repete processo recursivamente para cada subconjunto
4. Para quando critério de parada é atingido (profundidade máxima, número mínimo de amostras, pureza)

### 4.3 Critérios de Divisão

Para **classificação**, medimos impureza dos nós:

**Entropia (Information Gain):**
$$H(S) = -\sum_{i=1}^{c} p_i \log_2(p_i)$$

Onde $p_i$ é proporção da classe $i$ no conjunto $S$.

**Índice Gini:**
$$Gini(S) = 1 - \sum_{i=1}^{c} p_i^2$$

**Ganho de Informação:**
$$IG(S, A) = H(S) - \sum_{v \in valores(A)} \frac{|S_v|}{|S|} H(S_v)$$

Escolhemos a feature e corte que **maximizam o ganho** (reduzem impureza).

### 4.4 Vantagens e Desvantagens

**Vantagens:**
- Fácil interpretação e visualização
- Não requer normalização de dados
- Lida bem com features categóricas e numéricas
- Captura interações não-lineares

**Desvantagens:**
- Propensas a overfitting (especialmente árvores profundas)
- Instáveis — pequenas mudanças nos dados podem gerar árvores muito diferentes
- Viés para features com muitos valores
- Greedy — não garantem solução globalmente ótima

### 4.5 Implementação Simplificada de Árvore de Decisão

```python
import numpy as np
from collections import Counter

class No:
    """Nó da árvore de decisão."""
    def __init__(self, feature=None, threshold=None, esquerda=None, direita=None, valor=None):
        self.feature = feature        # Índice da feature para split
        self.threshold = threshold    # Valor de corte
        self.esquerda = esquerda      # Subárvore esquerda
        self.direita = direita        # Subárvore direita
        self.valor = valor            # Valor (classe) se for folha
    
    def eh_folha(self):
        return self.valor is not None

class ArvoreDecisao:
    """
    Implementação simplificada de Árvore de Decisão para classificação.
    Usa Índice Gini como critério de divisão.
    
    Parâmetros:
    -----------
    max_profundidade : int
        Profundidade máxima da árvore
    min_amostras_split : int
        Número mínimo de amostras para dividir nó
    """
    
    def __init__(self, max_profundidade=10, min_amostras_split=2):
        self.max_profundidade = max_profundidade
        self.min_amostras_split = min_amostras_split
        self.raiz = None
    
    def _gini(self, y):
        """Calcula índice Gini."""
        m = len(y)
        if m == 0:
            return 0
        contadores = Counter(y)
        impureza = 1.0
        for cont in contadores.values():
            prob = cont / m
            impureza -= prob ** 2
        return impureza
    
    def _dividir(self, X, y, feature, threshold):
        """Divide dados com base em feature e threshold."""
        mascara_esq = X[:, feature] <= threshold
        mascara_dir = ~mascara_esq
        return X[mascara_esq], X[mascara_dir], y[mascara_esq], y[mascara_dir]
    
    def _melhor_divisao(self, X, y):
        """Encontra melhor feature e threshold para dividir."""
        melhor_ganho = -1
        melhor_feature = None
        melhor_threshold = None
        
        gini_atual = self._gini(y)
        m = len(y)
        
        n_features = X.shape[1]
        
        for feature in range(n_features):
            thresholds = np.unique(X[:, feature])
            
            for threshold in thresholds:
                _, _, y_esq, y_dir = self._dividir(X, y, feature, threshold)
                
                if len(y_esq) == 0 or len(y_dir) == 0:
                    continue
                
                # Calcula Gini ponderado após divisão
                p_esq = len(y_esq) / m
                p_dir = len(y_dir) / m
                gini_pos_divisao = (p_esq * self._gini(y_esq) + 
                                   p_dir * self._gini(y_dir))
                
                ganho = gini_atual - gini_pos_divisao
                
                if ganho > melhor_ganho:
                    melhor_ganho = ganho
                    melhor_feature = feature
                    melhor_threshold = threshold
        
        return melhor_feature, melhor_threshold
    
    def _construir_arvore(self, X, y, profundidade=0):
        """Constrói árvore recursivamente."""
        n_amostras = len(y)
        n_classes = len(np.unique(y))
        
        # Critérios de parada
        if (profundidade >= self.max_profundidade or 
            n_amostras < self.min_amostras_split or 
            n_classes == 1):
            # Cria folha com classe mais comum
            valor_folha = Counter(y).most_common(1)[0][0]
            return No(valor=valor_folha)
        
        # Encontra melhor divisão
        feature, threshold = self._melhor_divisao(X, y)
        
        if feature is None:
            valor_folha = Counter(y).most_common(1)[0][0]
            return No(valor=valor_folha)
        
        # Divide dados
        X_esq, X_dir, y_esq, y_dir = self._dividir(X, y, feature, threshold)
        
        # Constrói subárvores recursivamente
        no_esquerdo = self._construir_arvore(X_esq, y_esq, profundidade + 1)
        no_direito = self._construir_arvore(X_dir, y_dir, profundidade + 1)
        
        return No(feature, threshold, no_esquerdo, no_direito)
    
    def fit(self, X, y):
        """Treina árvore de decisão."""
        self.raiz = self._construir_arvore(X, y)
    
    def _predizer_um(self, x, no):
        """Percorre árvore para fazer predição de um exemplo."""
        if no.eh_folha():
            return no.valor
        
        if x[no.feature] <= no.threshold:
            return self._predizer_um(x, no.esquerda)
        return self._predizer_um(x, no.direita)
    
    def predict(self, X):
        """Faz predições para múltiplos exemplos."""
        return np.array([self._predizer_um(x, self.raiz) for x in X])
    
    def score(self, X, y):
        """Calcula acurácia."""
        predicoes = self.predict(X)
        return np.mean(predicoes == y)

# Exemplo de uso
print("="*60)
print("ÁRVORE DE DECISÃO - IMPLEMENTAÇÃO SIMPLIFICADA")
print("="*60)

# Usa os mesmos dados de classificação anteriores
arvore = ArvoreDecisao(max_profundidade=5, min_amostras_split=2)
arvore.fit(X_treino, y_treino)

print(f"\nÁrvore treinada com profundidade máxima = 5")

acuracia_treino = arvore.score(X_treino, y_treino)
acuracia_teste = arvore.score(X_teste, y_teste)

print(f"\nAcurácia (treino): {acuracia_treino:.4f}")
print(f"Acurácia (teste):  {acuracia_teste:.4f}")

# Testa efeito da profundidade
print("\n" + "-"*60)
print("EFEITO DA PROFUNDIDADE MÁXIMA")
print("-"*60)

for prof in [2, 5, 10, 20]:
    arvore_temp = ArvoreDecisao(max_profundidade=prof)
    arvore_temp.fit(X_treino, y_treino)
    
    acc_treino = arvore_temp.score(X_treino, y_treino)
    acc_teste = arvore_temp.score(X_teste, y_teste)
    
    print(f"\nProfundidade = {prof}:")
    print(f"  Acurácia (treino): {acc_treino:.4f}")
    print(f"  Acurácia (teste):  {acc_teste:.4f}")
    
    if acc_treino > acc_teste + 0.1:
        print(f"  ⚠ Possível overfitting detectado!")

print("\n" + "="*60)
```

**Observações sobre Árvores de Decisão:**

1. **Overfitting:** Árvores profundas memorizam dados de treino, generalizando mal. Observe a diferença entre acurácia de treino e teste.

2. **Poda (Pruning):** Técnica para simplificar árvore removendo nós que contribuem pouco para performance, reduzindo overfitting.

3. **Critério de divisão:** Índice Gini é computacionalmente mais eficiente que Entropia e geralmente produz resultados similares.

4. **Interpretabilidade:** Podemos visualizar e interpretar regras de decisão facilmente — grande vantagem sobre modelos "caixa-preta".

5. **Limitações:** Árvores únicas são instáveis; métodos de ensemble (Random Forest, Gradient Boosting) combinam múltiplas árvores para melhor desempenho.

## 6. Tópicos Avançados em Machine Learning

### 6.1 Ensemble Learning

**Ensemble Learning** combina múltiplos modelos para melhorar desempenho. Principais técnicas:

**Bagging (Bootstrap Aggregating):**
- Treina modelos em amostras bootstrap (com reposição) dos dados
- Combina predições por voto (classificação) ou média (regressão)
- Exemplo: **Random Forest** — ensemble de árvores de decisão
- Reduz variância e overfitting

**Boosting:**
- Treina modelos sequencialmente, cada um corrigindo erros do anterior
- Combina modelos com pesos baseados em desempenho
- Exemplos: **AdaBoost, Gradient Boosting, XGBoost**
- Reduz viés e melhora predições em casos difíceis

**Stacking:**
- Usa predições de múltiplos modelos como features para meta-modelo
- Meta-modelo aprende como melhor combinar modelos base

### 6.2 Validação e Avaliação de Modelos

**Validação Cruzada (Cross-Validation):**
- **K-Fold CV:** Divide dados em K partes; treina em K-1, testa em 1, repete K vezes
- Fornece estimativa mais robusta de desempenho
- Variantes: Stratified K-Fold (mantém proporção de classes), Leave-One-Out

**Métricas de Classificação:**
- **Acurácia:** Proporção de predições corretas — pode enganar em classes desbalanceadas
- **Precisão:** TP / (TP + FP) — importante quando falsos positivos são custosos
- **Recall (Sensibilidade):** TP / (TP + FN) — importante quando falsos negativos são custosos
- **F1-Score:** Média harmônica de precisão e recall
- **ROC-AUC:** Área sob curva ROC — mede capacidade de discriminação

**Métricas de Regressão:**
- **MSE (Mean Squared Error):** Penaliza erros grandes
- **RMSE (Root Mean Squared Error):** MSE na mesma escala da variável alvo
- **MAE (Mean Absolute Error):** Mais robusta a outliers
- **R² (Coeficiente de Determinação):** Proporção de variância explicada

### 6.3 Regularização

Técnicas para prevenir overfitting adicionando penalidades aos parâmetros:

**Ridge (L2):** Penaliza soma dos quadrados dos parâmetros
$$J(\theta) = MSE + \alpha \sum_{j=1}^{n} \theta_j^2$$

**Lasso (L1):** Penaliza soma dos valores absolutos — pode zerar parâmetros (seleção de features)
$$J(\theta) = MSE + \alpha \sum_{j=1}^{n} |\theta_j|$$

**Elastic Net:** Combina L1 e L2

### 6.4 Feature Engineering e Seleção

**Feature Engineering:**
- Criar novas features a partir das existentes (interações, transformações)
- Codificação categórica (One-Hot, Label Encoding, Target Encoding)
- Transformações (log, raiz, polinomiais)
- Agregações temporais/espaciais

**Seleção de Features:**
- **Métodos de Filtro:** Correlação, Chi-quadrado, Informação Mútua
- **Métodos Wrapper:** Seleção forward/backward, RFE (Recursive Feature Elimination)
- **Métodos Embedded:** LASSO, importância de features em Random Forest

### 6.5 Lidando com Desbalanceamento de Classes

**Técnicas de Reamostragem:**
- **Over-sampling:** SMOTE (Synthetic Minority Over-sampling Technique)
- **Under-sampling:** Random under-sampling, Tomek Links
- **Híbridos:** SMOTEENN, SMOTETomek

**Ajustes Algorítmicos:**
- Pesos de classe (class_weight='balanced')
- Threshold adjustment
- Métodos sensíveis a custo

### 6.6 Deep Learning

**Redes Neurais Profundas** são extensões do Perceptron com múltiplas camadas ocultas:

- **MLPs (Multi-Layer Perceptrons):** Redes densamente conectadas
- **CNNs (Convolutional Neural Networks):** Especializadas em dados visuais
- **RNNs/LSTMs/Transformers:** Especializadas em dados sequenciais (texto, séries temporais)

Frameworks: **TensorFlow, PyTorch, Keras**

## 8. Considerações Éticas e Desafios

### 8.1 Vieses e Fairness

Modelos de ML podem herdar e amplificar vieses presentes nos dados de treinamento:
- **Discriminação algorítmica:** Sistemas de recrutamento, crédito ou justiça criminal discriminando grupos
- **Representatividade:** Dados não representativos levam a modelos injustos
- **Feedback loops:** Predições influenciam coleta de dados futuros, perpetuando vieses

**Mitigação:** Auditorias de fairness, datasets balanceados, métricas de equidade (demographic parity, equal opportunity).

### 8.2 Interpretabilidade vs. Performance

Trade-off fundamental:
- Modelos simples (regressão, árvores) são interpretáveis mas menos precisos
- Modelos complexos (deep learning, ensembles) são "caixas-pretas" mas mais precisos

**Explainable AI (XAI):** Técnicas para interpretar modelos complexos (SHAP, LIME, attention mechanisms).

### 8.3 Privacidade e Segurança

- **Privacidade diferencial:** Garantir que modelos não vazem informações individuais
- **Federated Learning:** Treinar modelos sem compartilhar dados raw
- **Ataques adversários:** Inputs maliciosos que enganam modelos
- **Model stealing:** Reverter modelo para extrair conhecimento proprietário

### 8.4 Impacto Ambiental

Treinamento de modelos grandes (especialmente deep learning) consome enorme energia:
- **Pegada de carbono:** Treinar modelos como GPT-3 equivale a décadas de consumo residencial
- **Sustentabilidade:** Necessidade de algoritmos mais eficientes, hardware especializado, fontes renováveis

### 8.5 Desemprego Tecnológico

Automação via ML pode substituir empregos:
- Motoristas (veículos autônomos)
- Operadores de call center (chatbots)
- Analistas (sistemas de decisão automatizados)

**Desafio social:** Requalificação profissional, políticas públicas, discussão sobre renda básica universal.
