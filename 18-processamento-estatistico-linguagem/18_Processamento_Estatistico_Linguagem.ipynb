## 1. O que é Processamento Estatístico de Linguagem Natural?

O **Processamento de Linguagem Natural** (PLN ou NLP — *Natural Language Processing*) é a área da IA dedicada a capacitar computadores a entender, interpretar e gerar texto e fala humanos. A abordagem **estatística** do PLN trata a linguagem como um fenômeno probabilístico: em vez de escrever regras gramaticais explícitas, aprende padrões diretamente de grandes volumes de texto.

### Da Abordagem Simbólica à Estatística

| Abordagem | Período | Característica | Exemplo |
|-----------|---------|---------------|----------|
| **Simbólica/Linguística** | 1950s–1980s | Regras gramaticais explícitas | Parsers, gramáticas formais |
| **Estatística** | 1990s–2010s | Probabilidades aprendidas de corpus | n-gramas, HMM, SVMs |
| **Neural/Deep** | 2010s–hoje | Representações distribuídas profundas | Word2Vec, BERT, GPT |

### Tarefas Centrais do PLN

```
Tarefas do PLN
├── Pré-processamento
│   ├── Tokenização
│   ├── Stemming / Lemmatização
│   ├── Remoção de stopwords
│   └── Normalização
├── Análise
│   ├── POS Tagging (classes gramaticais)
│   ├── NER (entidades nomeadas)

│   └── Análise de sentimentos
└── Geração e Aplicações
    ├── Classificação de texto
    ├── Tradução automática
    ├── Sumarização
    └── Sistemas de pergunta-resposta
```

## 3. Conceitos Fundamentais

### 3.1 Pré-processamento de Texto

| Etapa | Descrição | Exemplo |
|-------|-----------|----------|
| **Tokenização** | Dividir texto em tokens (palavras, subpalavras) | `"Olá mundo"` → `["Olá", "mundo"]` |
| **Lowercase** | Converter tudo para minúsculas | `"Olá"` → `"olá"` |
| **Remoção de stopwords** | Remover palavras muito comuns e sem significado | `"o", "a", "de", "em"` |
| **Stemming** | Reduzir palavra à sua raiz (heurístico) | `"correndo"` → `"corr"` |
| **Lemmatização** | Reduzir à forma canônica (morfológico) | `"correndo"` → `"correr"` |
| **Remoção de pontuação** | Eliminar caracteres especiais | `"olá!"` → `"olá"` |

### 3.2 Representações de Texto

**Bag of Words (BoW):**
- Representa texto como vetor de contagens de palavras
- Ignora ordem e contexto
- Vocabulário: $V$ palavras → vetor de dimensão $|V|$

**TF-IDF (Term Frequency – Inverse Document Frequency):**
$$\text{TF-IDF}(t, d) = \text{TF}(t, d) \times \text{IDF}(t)$$

$$\text{TF}(t, d) = \frac{\text{freq}(t, d)}{\text{total de termos em } d}$$

$$\text{IDF}(t) = \log\frac{N}{1 + df(t)}$$

Onde $N$ é o número total de documentos e $df(t)$ é o número de documentos contendo o termo $t$.

### 3.3 Modelos de N-gramas

Um **n-grama** é uma sequência de $n$ palavras consecutivas. A probabilidade de uma sequência de palavras é modelada como:

$$P(w_1, w_2, \ldots, w_m) = \prod_{i=1}^{m} P(w_i \mid w_{i-(n-1)}, \ldots, w_{i-1})$$

**Suavização de Laplace** (evitar probabilidade zero):
$$P_{Laplace}(w_i \mid w_{i-1}) = \frac{C(w_{i-1}, w_i) + 1}{C(w_{i-1}) + |V|}$$

## 4. Implementação do Zero — PLN Estatístico

````python
import re
import math
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
from typing import List, Dict

# ─────────────────────────────────────────────────────────────
# PARTE 1: Pré-processamento de texto
# ─────────────────────────────────────────────────────────────
STOPWORDS_PT = {
    'a', 'o', 'as', 'os', 'um', 'uma', 'de', 'do', 'da', 'dos', 'das',
    'em', 'no', 'na', 'nos', 'nas', 'por', 'para', 'com', 'sem', 'sob',
    'e', 'ou', 'que', 'se', 'não', 'mais', 'já', 'ser', 'ter', 'há',
    'ao', 'à', 'às', 'aos', 'this', 'is', 'the'
}

def pre_processar(texto: str, remover_stopwords: bool = True,
                  stemming_simples: bool = False) -> List[str]:
    """Pipeline de pré-processamento de texto."""
    # Lowercase
    texto = texto.lower()
    # Remover pontuação e números
    texto = re.sub(r'[^a-záéíóúâêôãõüçàèìòù\s]', ' ', texto)
    # Tokenização simples por espaços
    tokens = texto.split()
    # Remover tokens muito curtos
    tokens = [t for t in tokens if len(t) > 2]
    # Remover stopwords
    if remover_stopwords:
        tokens = [t for t in tokens if t not in STOPWORDS_PT]
    # Stemming simples (apenas para demonstração)
    if stemming_simples:
        sufixos = ['ando', 'endo', 'indo', 'ção', 'ões', 'dade', 'mente']
        novos = []
        for t in tokens:
            for suf in sufixos:
                if t.endswith(suf) and len(t) - len(suf) >= 3:
                    t = t[:-len(suf)]
                    break
            novos.append(t)
        tokens = novos
    return tokens

# Corpus de exemplo (notícias simuladas)
corpus = [
    "A inteligência artificial está revolucionando a medicina moderna e o diagnóstico de doenças",
    "Redes neurais profundas alcançam desempenho humano no reconhecimento de imagens médicas",
    "O aprendizado de máquina transforma a indústria financeira com detecção de fraudes automática",
    "Algoritmos genéticos otimizam rotas de entrega e reduzem custos logísticos em empresas",
    "Processamento de linguagem natural permite assistentes virtuais compreenderem contexto",
    "Sistemas especialistas baseados em casos auxiliam médicos no diagnóstico clínico preciso",
    "Robótica avançada com IA realiza cirurgias complexas com precisão milimétrica incrível",
    "Carros autônomos utilizam visão computacional e redes neurais para navegação segura",
    "Inteligência artificial detecta padrões em dados financeiros para prever crises econômicas",
    "Modelos de linguagem grandes como GPT transformam a forma como escrevemos e programamos",
]

rotulos = ['medicina', 'medicina', 'finanças', 'otimização',
           'nlp', 'medicina', 'robótica', 'transporte',
           'finanças', 'nlp']

# Pré-processar corpus
corpus_processado = [pre_processar(doc) for doc in corpus]

print("=" * 60)
print("  Pré-processamento de Texto")
print("=" * 60)
for i, (original, processado) in enumerate(zip(corpus[:3], corpus_processado[:3])):
    print(f"  Doc {i+1} original: {original[:70]}...")
    print(f"  Doc {i+1} processado: {processado}")
    print()

# ─────────────────────────────────────────────────────────────
# PARTE 2: Bag of Words (BoW) do zero
# ─────────────────────────────────────────────────────────────
def construir_vocabulario(corpus_tokens: List[List[str]]) -> Dict[str, int]:
    """Constrói vocabulário ordenado por frequência."""
    todas_palavras = [token for doc in corpus_tokens for token in doc]
    vocab = {palavra: idx for idx, (palavra, _) in
             enumerate(Counter(todas_palavras).most_common())}
    return vocab

def vetorizar_bow(doc_tokens: List[str], vocabulario: Dict[str, int]) -> np.ndarray:
    """Converte lista de tokens em vetor BoW."""
    vetor = np.zeros(len(vocabulario))
    for token in doc_tokens:
        if token in vocabulario:
            vetor[vocabulario[token]] += 1
    return vetor

# ─────────────────────────────────────────────────────────────
# PARTE 3: TF-IDF do zero
# ─────────────────────────────────────────────────────────────
def calcular_tf(doc_tokens: List[str]) -> Dict[str, float]:
    """Calcula TF (frequência relativa) de cada termo."""
    contagem = Counter(doc_tokens)
    total = len(doc_tokens)
    return {termo: contagem[termo] / total for termo in contagem}

def calcular_idf(corpus_tokens: List[List[str]]) -> Dict[str, float]:
    """Calcula IDF de cada termo no corpus."""
    N = len(corpus_tokens)
    df = defaultdict(int)
    for doc in corpus_tokens:
        for termo in set(doc):
            df[termo] += 1
    return {termo: math.log(N / (1 + df[termo])) for termo in df}

def tfidf_matriz(corpus_tokens: List[List[str]], vocabulario: Dict[str, int]) -> np.ndarray:
    """Constrói a matriz TF-IDF (documentos × termos)."""
    idf = calcular_idf(corpus_tokens)
    matriz = np.zeros((len(corpus_tokens), len(vocabulario)))
    for i, doc in enumerate(corpus_tokens):
        tf = calcular_tf(doc)
        for termo, tf_val in tf.items():
            if termo in vocabulario:
                j = vocabulario[termo]
                matriz[i, j] = tf_val * idf.get(termo, 0)
    return matriz

# Construir vocabulário e matrizes
vocabulario = construir_vocabulario(corpus_processado)
print(f"Vocabulário: {len(vocabulario)} termos únicos")

matriz_bow = np.array([vetorizar_bow(doc, vocabulario) for doc in corpus_processado])
matriz_tfidf = tfidf_matriz(corpus_processado, vocabulario)

print(f"Forma da matriz BoW:    {matriz_bow.shape}")
print(f"Forma da matriz TF-IDF: {matriz_tfidf.shape}")

# ─────────────────────────────────────────────────────────────
# PARTE 4: Similaridade de cosseno para busca de documentos
# ─────────────────────────────────────────────────────────────
def similaridade_cosseno(v1: np.ndarray, v2: np.ndarray) -> float:
    """Calcula a similaridade de cosseno entre dois vetores."""
    norma1 = np.linalg.norm(v1)
    norma2 = np.linalg.norm(v2)
    if norma1 == 0 or norma2 == 0:
        return 0.0
    return float(np.dot(v1, v2) / (norma1 * norma2))

def buscar_documentos(consulta: str, corpus_original: List[str],
                      matriz: np.ndarray, vocab: Dict, top_k: int = 3):
    """Busca documentos similares à consulta usando TF-IDF + cosseno."""
    consulta_tokens = pre_processar(consulta)
    vetor_consulta = vetorizar_bow(consulta_tokens, vocab)

    similaridades = [
        (i, similaridade_cosseno(vetor_consulta, matriz[i]))
        for i in range(len(corpus_original))
    ]
    similaridades.sort(key=lambda x: -x[1])

    return similaridades[:top_k]

print("\n" + "=" * 60)
print("  Busca de Documentos por Similaridade de Cosseno")
print("=" * 60)

consulta = "diagnóstico médico inteligência artificial hospital"
resultados = buscar_documentos(consulta, corpus, matriz_tfidf, vocabulario, top_k=3)

print(f"  Consulta: '{consulta}'")
print(f"  Top-3 documentos mais relevantes:")
for rank, (idx, sim) in enumerate(resultados, 1):
    print(f"    {rank}. (sim={sim:.4f}) {corpus[idx][:70]}...")

# Visualização: heatmap de similaridade entre documentos
n_docs = len(corpus)
sim_matrix = np.zeros((n_docs, n_docs))
for i in range(n_docs):
    for j in range(n_docs):
        sim_matrix[i, j] = similaridade_cosseno(matriz_tfidf[i], matriz_tfidf[j])

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

ax = axes[0]
im = ax.imshow(sim_matrix, cmap='Blues', vmin=0, vmax=1)
plt.colorbar(im, ax=ax)
ax.set_title('Matriz de Similaridade de Cosseno\n(TF-IDF entre documentos)')
labels_docs = [f'D{i+1}\n({rotulos[i]})' for i in range(n_docs)]
ax.set_xticks(range(n_docs))
ax.set_yticks(range(n_docs))
ax.set_xticklabels(labels_docs, fontsize=7)
ax.set_yticklabels(labels_docs, fontsize=7)
for i in range(n_docs):
    for j in range(n_docs):
        if i != j and sim_matrix[i, j] > 0.05:
            ax.text(j, i, f'{sim_matrix[i,j]:.2f}', ha='center', va='center',
                    fontsize=6, color='black')

# Frequência de termos por categoria
ax = axes[1]
todas_palavras = [t for doc in corpus_processado for t in doc]
freq_geral = Counter(todas_palavras).most_common(15)
palavras_plot = [p for p, _ in freq_geral]
freqs_plot = [f for _, f in freq_geral]
ax.barh(palavras_plot[::-1], freqs_plot[::-1], color='steelblue', alpha=0.8)
ax.set_xlabel('Frequência no corpus')
ax.set_title('Top-15 Termos mais Frequentes\n(após pré-processamento)')
ax.grid(True, axis='x', alpha=0.3)

plt.suptitle('PLN Estatístico — TF-IDF e Similaridade', fontsize=13)
plt.tight_layout()
plt.show()
print("\n✅ Implementação do zero concluída.")
````

## 5. Modelos de N-gramas

````python
import re
import math
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
from typing import List, Tuple

# ─────────────────────────────────────────────────────────────
# Modelo de N-gramas com suavização de Laplace
# ─────────────────────────────────────────────────────────────
class ModeloNgramas:
    """
    Modelo de linguagem baseado em n-gramas com suavização de Laplace.
    Suporta unigrama, bigrama e trigrama.
    """

    def __init__(self, n: int = 2):
        self.n = n
        self.ngramas = Counter()
        self.contextos = Counter()  # (n-1)-gramas
        self.vocabulario = set()
        self.total_tokens = 0

    def treinar(self, tokens: List[str]):
        """Treina o modelo a partir de uma lista de tokens."""
        # Adicionar tokens especiais de início/fim
        tokens_pad = ['<s>'] * (self.n - 1) + tokens + ['</s>']
        self.vocabulario = set(tokens) | {'<s>', '</s>', '<UNK>'}
        self.total_tokens = len(tokens)

        for i in range(len(tokens_pad) - self.n + 1):
            ngrama = tuple(tokens_pad[i:i + self.n])
            contexto = ngrama[:-1]
            self.ngramas[ngrama] += 1
            if self.n > 1:
                self.contextos[contexto] += 1

    def probabilidade(self, token: str, contexto: tuple) -> float:
        """P(token | contexto) com suavização de Laplace."""
        V = len(self.vocabulario)
        if self.n == 1:
            contagem = self.ngramas.get((token,), 0)
            return (contagem + 1) / (self.total_tokens + V)
        else:
            ngrama = contexto + (token,)
            num = self.ngramas.get(ngrama, 0) + 1
            den = self.contextos.get(contexto, 0) + V
            return num / den

    def perplexidade(self, tokens: List[str]) -> float:
        """Calcula a perplexidade do modelo em um conjunto de teste."""
        tokens_pad = ['<s>'] * (self.n - 1) + tokens + ['</s>']
        log_prob_total = 0.0
        N = len(tokens) + 1  # incluir </s>

        for i in range(self.n - 1, len(tokens_pad)):
            token = tokens_pad[i]
            contexto = tuple(tokens_pad[i - self.n + 1:i])
            prob = self.probabilidade(token, contexto)
            log_prob_total += math.log2(prob + 1e-12)

        return 2 ** (-log_prob_total / N)

    def gerar_texto(self, inicio: tuple = None, max_tokens: int = 15) -> str:
        """Gera texto usando o modelo de n-gramas (amostragem)."""
        import random
        if inicio is None:
            inicio = ('inteligência',) if self.n > 1 else ()

        sequencia = list(inicio)

        for _ in range(max_tokens):
            contexto = tuple(sequencia[-(self.n-1):]) if self.n > 1 else ()
            # Candidatos: tokens com probabilidade > 0
            candidatos = []
            probs = []
            palavras_vocab = list(self.vocabulario - {'<s>', '<UNK>'})
            for w in palavras_vocab:
                p = self.probabilidade(w, contexto)
                candidatos.append(w)
                probs.append(p)

            # Normalizar
            total_p = sum(probs)
            probs = [p / total_p for p in probs]

            proximo = random.choices(candidatos, weights=probs, k=1)[0]
            if proximo == '</s>':
                break
            sequencia.append(proximo)

        return ' '.join(sequencia)


# Corpus de treinamento
texto_treino = """
a inteligência artificial aprende com dados e experiências
redes neurais profundas processam imagens e textos com eficiência
algoritmos de aprendizado de máquina identificam padrões em dados
o processamento de linguagem natural entende texto humano
sistemas especialistas utilizam conhecimento para resolver problemas
a inteligência artificial transforma a medicina e a tecnologia
aprendizado por reforço treina agentes para tomar decisões
redes neurais convolucionais reconhecem objetos em imagens digitais
o aprendizado profundo revolucionou o processamento de linguagem natural
algoritmos genéticos inspirados na evolução biológica otimizam soluções
"""

tokens_treino = pre_processar(texto_treino, remover_stopwords=False)

# Treinar modelos de 1-grama, 2-grama, 3-grama
modelos = {}
for n in [1, 2, 3]:
    modelo = ModeloNgramas(n=n)
    modelo.treinar(tokens_treino)
    modelos[n] = modelo

print("=" * 60)
print("  Modelos de N-gramas — Treinamento e Avaliação")
print("=" * 60)

# Avaliar perplexidade em texto de teste
texto_teste = "inteligência artificial aprende padrões em dados médicos"
tokens_teste = pre_processar(texto_teste, remover_stopwords=False)

print(f"\nTexto de teste: '{texto_teste}'")
print(f"Tokens: {tokens_teste}")
print()
print(f"{'Modelo':>12} | {'Perplexidade':>15} | Interpretação")
print("-" * 60)
for n, modelo in modelos.items():
    pp = modelo.perplexidade(tokens_teste)
    interp = "mais específico" if n > 1 else "baseline"
    print(f"  {n}-grama    | {pp:>14.2f} | {interp}")

# Geração de texto
print("\nGeração de texto com bigrama:")
import random
random.seed(42)
for i in range(3):
    texto_gerado = modelos[2].gerar_texto(inicio=('inteligência',), max_tokens=10)
    print(f"  [{i+1}] {texto_gerado}")

# Visualização: probabilidades condicionais do bigrama
palavra_alvo = 'aprendizado'
proximo_palavras = ['de', 'profundo', 'máquina', 'por', 'artificial', 'natural']
probs_cond = [modelos[2].probabilidade(w, (palavra_alvo,)) for w in proximo_palavras]

plt.figure(figsize=(10, 4))
plt.bar(proximo_palavras, probs_cond, color='steelblue', alpha=0.8, edgecolor='black')
plt.title(f'P(palavra | "{palavra_alvo}") — Modelo de Bigrama')
plt.ylabel('Probabilidade')
plt.xlabel('Próxima palavra')
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
print("\n✅ Modelos de N-gramas concluídos.")
````

## 6. PLN com Bibliotecas — scikit-learn e NLTK

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import TruncatedSVD
import warnings
warnings.filterwarnings('ignore')

# ─────────────────────────────────────────────────────────────
# Dataset de Classificação de Texto (notícias IA)
# ─────────────────────────────────────────────────────────────
documentos = [
    # Medicina
    "IA diagnostica tumores com precisão superior aos radiologistas",
    "Redes neurais identificam câncer de pele em imagens dermatológicas",
    "Aprendizado de máquina prevê doenças cardíacas por dados clínicos",
    "Sistema especialista auxilia médicos no diagnóstico diferencial preciso",
    "Algoritmo detecta sepse em UTI com antecedência de 6 horas",
    # Finanças
    "Algoritmos de trading de alta frequência dominam bolsas mundiais",
    "Machine learning detecta fraudes em cartões de crédito em tempo real",
    "IA prevê crises financeiras analisando dados macroeconômicos globais",
    "Robo-advisors gerenciam carteiras com estratégias otimizadas por IA",
    "Redes neurais analisam sentimento do mercado em redes sociais",
    # Robótica
    "Robôs autônomos realizam tarefas complexas em ambientes industriais",
    "Braços robóticos aprendem movimentos por imitação com IA",
    "Drones autônomos entregam mercadorias usando visão computacional",
    "Robótica colaborativa trabalha ao lado de humanos com segurança",
    "Exoesqueletos controlados por IA auxiliam pacientes com paralisia",
    # NLP
    "Modelos de linguagem grandes geram texto indistinguível do humano",
    "Tradução automática neural supera qualidade de tradutores humanos",
    "Análise de sentimentos em redes sociais detecta tendências políticas",
    "Chatbots baseados em transformers respondem perguntas complexas",
    "PLN extrai informações estruturadas de documentos não estruturados",
]

categorias = (['medicina'] * 5 + ['finanças'] * 5 +
              ['robótica'] * 5 + ['nlp'] * 5)

# ─────────────────────────────────────────────────────────────
# Comparar pipelines: BoW vs TF-IDF + diferentes classifiers
# ─────────────────────────────────────────────────────────────
pipelines = [
    ('BoW + Naive Bayes',
     Pipeline([('vec', CountVectorizer()), ('clf', MultinomialNB())])),
    ('TF-IDF + Naive Bayes',
     Pipeline([('vec', TfidfVectorizer()), ('clf', MultinomialNB())])),
    ('TF-IDF + Regressão Logística',
     Pipeline([('vec', TfidfVectorizer()), ('clf', LogisticRegression(max_iter=300))])),
    ('TF-IDF + SVM Linear',
     Pipeline([('vec', TfidfVectorizer()), ('clf', LinearSVC(max_iter=300))])),
]

print("=" * 60)
print("  Classificação de Texto — Pipelines PLN (5-fold CV)")
print("=" * 60)

resultados_nlp = {}
for nome, pipeline in pipelines:
    scores = cross_val_score(pipeline, documentos, categorias, cv=4, scoring='accuracy')
    resultados_nlp[nome] = scores
    print(f"  {nome:35s}: {scores.mean()*100:.1f}% ± {scores.std()*100:.1f}%")

# ─────────────────────────────────────────────────────────────
# TF-IDF + palavras mais importantes por categoria
# ─────────────────────────────────────────────────────────────
vectorizer = TfidfVectorizer(max_features=50)
X_tfidf = vectorizer.fit_transform(documentos)
feature_names = vectorizer.get_feature_names_out()

print("\nTop-5 termos TF-IDF por categoria:")
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(categorias)

for cat in le.classes_:
    mask = np.array(categorias) == cat
    media_tfidf = X_tfidf[mask].toarray().mean(axis=0)
    top5_idx = media_tfidf.argsort()[::-1][:5]
    top5_termos = [feature_names[i] for i in top5_idx]
    print(f"  {cat:12s}: {top5_termos}")

# Visualização: projeção 2D do espaço TF-IDF (SVD/LSA)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Projeção LSA
ax = axes[0]
svd = TruncatedSVD(n_components=2, random_state=42)
X_2d = svd.fit_transform(X_tfidf)

cores_cat = {'medicina': 'steelblue', 'finanças': 'coral',
             'robótica': 'green', 'nlp': 'purple'}
for cat, cor in cores_cat.items():
    mask = np.array(categorias) == cat
    ax.scatter(X_2d[mask, 0], X_2d[mask, 1], c=cor, label=cat,
               s=80, alpha=0.8, edgecolors='black', linewidths=0.5)
for i, doc in enumerate(documentos):
    ax.annotate(str(i+1), (X_2d[i, 0], X_2d[i, 1]),
                fontsize=7, ha='center', va='bottom')
ax.set_xlabel('Componente 1 (SVD/LSA)')
ax.set_ylabel('Componente 2')
ax.set_title('Espaço TF-IDF projetado em 2D\n(LSA — Análise Semântica Latente)')
ax.legend()
ax.grid(True, alpha=0.3)

# Box plot de acurácias
ax = axes[1]
nomes = list(resultados_nlp.keys())
dados = [resultados_nlp[n] * 100 for n in nomes]
bp = ax.boxplot(dados, patch_artist=True)
cores_bp = ['lightblue', 'lightgreen', 'lightsalmon', 'lightyellow']
for patch, cor in zip(bp['boxes'], cores_bp):
    patch.set_facecolor(cor)
ax.set_xticklabels([n.replace(' + ', '\n+\n') for n in nomes], fontsize=8)
ax.set_ylabel('Acurácia (%)')
ax.set_title('Comparação de Pipelines PLN\n(4-fold CV)')
ax.grid(True, axis='y', alpha=0.3)

plt.suptitle('PLN com scikit-learn — TF-IDF e Classificação de Texto', fontsize=13)
plt.tight_layout()
plt.show()
print("\n✅ PLN com bibliotecas concluído.")
````

## 7. Comparação de Abordagens de PLN

| Abordagem | Representação | Vantagens | Limitações |
|-----------|--------------|-----------|------------|
| **Bag of Words** | Vetor de contagens | Simples, interpretável | Ignora ordem; alta dimensionalidade |
| **TF-IDF** | Pesos tf × idf | Penaliza termos comuns | Ainda ignora semântica |
| **N-gramas** | Sequências de n palavras | Captura co-ocorrências locais | Explosão do vocabulário |
| **LSA/LSI** | SVD sobre TF-IDF | Captura semântica latente | Computacionalmente caro |
| **Word2Vec/GloVe** | Embeddings densos | Semântica distribucional | Não tem contexto dinâmico |
| **BERT/GPT** | Transformers contextuais | Estado da arte em todas as tarefas | Enorme custo computacional |

## 8. Aplicações do PLN Estatístico

### Busca e Recuperação de Informação
- **Motores de busca:** TF-IDF para ranking de documentos (base do Google)
- **Sistemas de FAQ:** similaridade semântica para encontrar respostas
- **Busca jurídica:** recuperação de jurisprudência relevante

### Classificação de Documentos
- **Filtros de spam:** Naive Bayes + TF-IDF
- **Análise de sentimentos:** avaliações de produtos, posts em redes sociais
- **Categorização automática de notícias**

### Extração de Informação
- **NER (Named Entity Recognition):** identificação de pessoas, locais, datas
- **Extração de relações:** quem fez o quê a quem
- **Sumarização automática de documentos**

### Sistemas de Diálogo
- **Chatbots:** compreensão e geração de respostas
- **Assistentes virtuais:** Siri, Alexa, Google Assistant
- **Tradução automática:** Google Translate

## 9. Vantagens e Desvantagens

| Vantagens | Desvantagens |
|-----------|-------------|
| ✅ Aprende de dados sem regras manuais | ❌ Requer grandes volumes de dados |
| ✅ TF-IDF e BoW são rápidos e eficientes | ❌ BoW/TF-IDF ignoram contexto e ordem |
| ✅ N-gramas capturam co-ocorrências locais | ❌ N-gramas têm explosão combinatorial |
| ✅ Facilmente escalável a grandes corpora | ❌ Sensível a erros ortográficos e gírias |
| ✅ Integra bem com ML clássico (SVM, NB) | ❌ Dificuldade com ambiguidade semântica |
| ✅ Transparência nos pesos TF-IDF | ❌ Idioma-específico (stopwords, stemming) |
