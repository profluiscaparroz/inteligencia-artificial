## 1. Introdução à Lógica na Inteligência Artificial

A lógica é uma das bases mais antigas e importantes da Inteligência Artificial. Desde os primeiros sistemas de IA, pesquisadores perceberam que, para uma máquina "pensar" ou "raciocinar", era necessário um **sistema formal** capaz de representar conhecimento e derivar conclusões a partir dele.

### Por que a lógica é fundamental para a IA?

1. **Representação de conhecimento:** A lógica nos fornece uma linguagem precisa e sem ambiguidade para representar fatos sobre o mundo. Por exemplo, podemos representar que "Todo ser humano é mortal" de forma que um computador consiga entender e usar essa informação.

2. **Raciocínio automático:** Com regras lógicas, podemos programar máquinas para **deduzir** novas informações a partir de fatos já conhecidos. Se sabemos que "Todo ser humano é mortal" e que "Sócrates é um ser humano", a máquina pode concluir que "Sócrates é mortal".

3. **Verificação e consistência:** A lógica permite verificar se um conjunto de informações é **consistente** (não possui contradições) e se determinadas conclusões são **válidas**.

4. **Base para sistemas especialistas:** Muitos sistemas de IA clássicos — como sistemas especialistas médicos, jurídicos e de diagnóstico — são construídos sobre fundamentos lógicos.

### Tipos de lógica que estudaremos

| Tipo | Descrição | Exemplo |
|------|-----------|--------|
| **Lógica Proposicional** | Trabalha com proposições (afirmações verdadeiras ou falsas) e conectivos lógicos | "Está chovendo E faz frio" |
| **Lógica de Primeira Ordem** | Estende a lógica proposicional com predicados, variáveis e quantificadores | "Todo aluno que estuda passa na prova" |

## 2. Lógica Proposicional

### O que é uma proposição?

Uma **proposição** é uma sentença declarativa que pode ser classificada como **verdadeira (V)** ou **falsa (F)**, mas nunca ambas ao mesmo tempo.

**Exemplos de proposições:**
- "A Terra é redonda." → **Verdadeiro (V)**
- "2 + 2 = 5" → **Falso (F)**
- "O Brasil é um país da América do Sul." → **Verdadeiro (V)**

**O que NÃO é uma proposição:**
- "Que horas são?" → É uma pergunta, não tem valor verdadeiro ou falso.
- "Feche a porta!" → É uma ordem.
- "x + 1 = 3" → Depende do valor de x, então não é uma proposição simples.

### Valores Verdade

Na lógica proposicional, cada proposição tem exatamente um **valor verdade**:
- **V** (Verdadeiro) — representado em Python como `True`
- **F** (Falso) — representado em Python como `False`

### Conectivos Lógicos

Os conectivos lógicos são operadores que combinam proposições para formar proposições mais complexas. Vamos estudar cada um deles:

---

#### 2.1 Negação (NÃO / NOT) — Símbolo: ¬

A negação inverte o valor verdade de uma proposição.

| p | ¬p |
|---|----|
| V | F  |
| F | V  |

**Exemplo do dia a dia:**
- p: "Está chovendo" → ¬p: "NÃO está chovendo"
- Se "Está chovendo" é **verdadeiro**, então "NÃO está chovendo" é **falso**.

**Em Python:** `not p`

---

#### 2.2 Conjunção (E / AND) — Símbolo: ∧

A conjunção é verdadeira **somente quando ambas** as proposições são verdadeiras.

| p | q | p ∧ q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   F   |
| F | F |   F   |

**Exemplo do dia a dia:**
- "Eu vou à praia **E** está fazendo sol."
- Só é verdadeiro se **ambas** as coisas acontecerem.

**Em Python:** `p and q`

---

#### 2.3 Disjunção (OU / OR) — Símbolo: ∨

A disjunção é verdadeira quando **pelo menos uma** das proposições é verdadeira.

| p | q | p ∨ q |
|---|---|-------|
| V | V |   V   |
| V | F |   V   |
| F | V |   V   |
| F | F |   F   |

**Exemplo do dia a dia:**
- "Vou estudar Python **OU** vou estudar Java."
- É verdadeiro se eu estudar pelo menos uma das duas linguagens (ou ambas).

**Em Python:** `p or q`

---

#### 2.4 Implicação (SE... ENTÃO / IF... THEN) — Símbolo: →

A implicação `p → q` é lida como "Se p, então q". Ela é **falsa somente quando** p é verdadeiro e q é falso.

| p | q | p → q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   V   |
| F | F |   V   |

**Exemplo do dia a dia:**
- "**Se** eu estudar, **então** vou passar na prova."
- Essa afirmação só é falsa se eu estudei mas NÃO passei na prova.
- Se eu não estudei, a promessa não foi quebrada (é verdadeira por vacuidade).

**Em Python:** `(not p) or q`  
(Pois p → q é logicamente equivalente a ¬p ∨ q)

---

#### 2.5 Bicondicional (SE E SOMENTE SE) — Símbolo: ↔

O bicondicional `p ↔ q` é verdadeiro quando **ambas as proposições têm o mesmo valor verdade**.

| p | q | p ↔ q |
|---|---|-------|
| V | V |   V   |
| V | F |   F   |
| F | V |   F   |
| F | F |   V   |

**Exemplo do dia a dia:**
- "Eu vou à festa **se e somente se** meu amigo também for."
- Só é verdadeiro se ambos forem OU ambos não forem.

**Em Python:** `p == q`

## 3. Exemplo em Python: Tabela Verdade

Vamos criar um programa simples em Python que gera **tabelas verdade** para os principais conectivos lógicos. Usaremos apenas loops simples e `print` para manter o código acessível a iniciantes.

```python
# Tabela Verdade para todos os conectivos lógicos
# Usamos True para Verdadeiro (V) e False para Falso (F)

# Lista com os dois valores possíveis
valores = [True, False]

# Função auxiliar para exibir V ou F em vez de True/False
def vf(valor):
    """Converte True/False para V/F."""
    return "V" if valor else "F"

print("=" * 60)
print("TABELA VERDADE COMPLETA")
print("=" * 60)

# Cabeçalho da tabela
print(f"{'p':<5} {'q':<5} {'¬p':<5} {'p∧q':<6} {'p∨q':<6} {'p→q':<6} {'p↔q':<6}")
print("-" * 40)

# Gerando todas as combinações de p e q
for p in valores:
    for q in valores:
        # Calculando cada conectivo
        negacao_p = not p                # Negação: ¬p
        conjuncao = p and q              # Conjunção: p ∧ q
        disjuncao = p or q               # Disjunção: p ∨ q
        implicacao = (not p) or q        # Implicação: p → q (equivale a ¬p ∨ q)
        bicondicional = p == q           # Bicondicional: p ↔ q

        # Exibindo a linha da tabela
        print(f"{vf(p):<5} {vf(q):<5} {vf(negacao_p):<5} {vf(conjuncao):<6} "
              f"{vf(disjuncao):<6} {vf(implicacao):<6} {vf(bicondicional):<6}")
```

```python
# Tabela verdade para uma expressão personalizada
# Exemplo: (p ∧ q) → (p ∨ q)

print("Tabela Verdade para: (p ∧ q) → (p ∨ q)")
print("=" * 45)
print(f"{'p':<5} {'q':<5} {'p∧q':<6} {'p∨q':<6} {'(p∧q)→(p∨q)':<12}")
print("-" * 45)

for p in [True, False]:
    for q in [True, False]:
        p_e_q = p and q             # p ∧ q
        p_ou_q = p or q             # p ∨ q
        resultado = (not p_e_q) or p_ou_q  # (p∧q) → (p∨q)

        print(f"{vf(p):<5} {vf(q):<5} {vf(p_e_q):<6} {vf(p_ou_q):<6} {vf(resultado):<12}")

print("\nObserve que esta expressão é sempre verdadeira (tautologia)!")
```

### Entendendo o código acima

- Usamos dois loops `for` aninhados para gerar todas as combinações possíveis de `p` e `q` (VV, VF, FV, FF).
- Para cada combinação, calculamos o resultado de cada conectivo lógico.
- A função `vf()` converte `True`/`False` em `V`/`F` para facilitar a leitura.
- Note que a **implicação** `p → q` é implementada como `(not p) or q`, pois são logicamente equivalentes.
- Uma **tautologia** é uma expressão que é sempre verdadeira, independentemente dos valores de p e q.

## 4. Exemplo em Python: Avaliador de Expressões Lógicas

Agora vamos criar uma função que recebe os valores de `p` e `q` e avalia diversas expressões lógicas, mostrando os resultados de forma clara.

```python
def avaliar_expressoes(p, q):
    """
    Avalia as principais expressões lógicas dados os valores de p e q.

    Parâmetros:
        p (bool): valor verdade da proposição p
        q (bool): valor verdade da proposição q
    """
    print(f"Dados: p = {vf(p)}, q = {vf(q)}")
    print("-" * 40)

    # Negação
    print(f"  ¬p (NOT p)       = {vf(not p)}")
    print(f"  ¬q (NOT q)       = {vf(not q)}")

    # Conjunção
    print(f"  p ∧ q (p AND q)  = {vf(p and q)}")

    # Disjunção
    print(f"  p ∨ q (p OR q)   = {vf(p or q)}")

    # Implicação: p → q equivale a (¬p ∨ q)
    implicacao = (not p) or q
    print(f"  p → q (p → q)    = {vf(implicacao)}")

    # Bicondicional: p ↔ q é verdadeiro quando p e q têm o mesmo valor
    bicondicional = p == q
    print(f"  p ↔ q (p ↔ q)    = {vf(bicondicional)}")

    print()


# Testando com diferentes combinações
print("AVALIADOR DE EXPRESSÕES LÓGICAS")
print("=" * 40)

avaliar_expressoes(True, True)
avaliar_expressoes(True, False)
avaliar_expressoes(False, True)
avaliar_expressoes(False, False)
```

```python
# Exemplo prático: avaliando uma situação do dia a dia
#
# Cenário: Um aluno quer saber se vai passar na disciplina.
# Regras:
#   p = "O aluno tem nota >= 7"
#   q = "O aluno tem frequência >= 75%"
#   Aprovação = p AND q (precisa de AMBAS as condições)

def verificar_aprovacao(nota, frequencia):
    """
    Verifica se o aluno foi aprovado com base na nota e frequência.

    Parâmetros:
        nota (float): nota do aluno (0 a 10)
        frequencia (float): percentual de frequência (0 a 100)
    """
    p = nota >= 7           # Proposição p: nota suficiente
    q = frequencia >= 75    # Proposição q: frequência suficiente

    # Aprovação requer AMBAS as condições (conjunção)
    aprovado = p and q

    print(f"Nota: {nota} → p (nota >= 7): {vf(p)}")
    print(f"Frequência: {frequencia}% → q (freq >= 75%): {vf(q)}")
    print(f"Aprovado (p ∧ q): {vf(aprovado)}")
    print()


print("VERIFICAÇÃO DE APROVAÇÃO")
print("=" * 40)

verificar_aprovacao(8.5, 80)   # Aprovado
verificar_aprovacao(8.0, 60)   # Reprovado por frequência
verificar_aprovacao(5.0, 90)   # Reprovado por nota
verificar_aprovacao(4.0, 50)   # Reprovado por ambos
```

## 5. Lógica de Primeira Ordem (Lógica de Predicados)

A **Lógica Proposicional** é poderosa, mas tem uma limitação importante: ela não consegue expressar relações entre objetos nem generalizar sobre grupos. Por exemplo, não conseguimos representar facilmente:

- "**Todos** os gatos são mamíferos."
- "**Existe** um aluno que gosta de IA."

Para isso, usamos a **Lógica de Primeira Ordem (LPO)**, que estende a lógica proposicional com novos elementos.

### 5.1 Predicados

Um **predicado** é uma função que recebe um ou mais argumentos e retorna verdadeiro ou falso.

**Exemplos:**
- `Gato(x)` → "x é um gato"
- `Mamifero(x)` → "x é um mamífero"
- `GostaDe(x, y)` → "x gosta de y"
- `MaiorQue(x, y)` → "x é maior que y"

### 5.2 Termos: Constantes, Variáveis e Funções

- **Constantes** são objetos específicos: `Sócrates`, `Brasil`, `7`
- **Variáveis** representam objetos genéricos: `x`, `y`, `z`
- **Funções** mapeiam objetos para outros objetos: `pai(x)` = "o pai de x"

### 5.3 Quantificadores

Os quantificadores nos permitem fazer afirmações sobre **grupos** de objetos.

#### Quantificador Universal (∀ — "Para todo")

Expressa que algo vale para **todos** os elementos do domínio.

- `∀x (Gato(x) → Mamifero(x))` → "Para todo x, se x é gato, então x é mamífero"
  - Em linguagem natural: **"Todos os gatos são mamíferos."**

- `∀x (Aluno(x) → EstudaLogica(x))` → "Para todo x, se x é aluno, então x estuda lógica"
  - Em linguagem natural: **"Todos os alunos estudam lógica."**

#### Quantificador Existencial (∃ — "Existe")

Expressa que algo vale para **pelo menos um** elemento do domínio.

- `∃x (Aluno(x) ∧ GostaDeIA(x))` → "Existe um x tal que x é aluno e x gosta de IA"
  - Em linguagem natural: **"Existe um aluno que gosta de IA."**

- `∃x (Numero(x) ∧ Par(x) ∧ Primo(x))` → "Existe um número que é par e primo"
  - Em linguagem natural: **"Existe um número par que é primo."** (Sim, o 2!)

### 5.4 Resumo da Notação

| Elemento | Símbolo | Exemplo | Significado |
|----------|---------|---------|-------------|
| Predicado | P(x) | Gato(x) | x é um gato |
| Constante | a, b | Sócrates | Um objeto específico |
| Variável | x, y | x | Um objeto genérico |
| Quantificador Universal | ∀ | ∀x | Para todo x |
| Quantificador Existencial | ∃ | ∃x | Existe um x |
| Função | f(x) | pai(x) | O pai de x |

## 6. Exemplo em Python: Representação de Predicados

Vamos usar **dicionários** e **listas** do Python para representar uma pequena base de conhecimento com predicados e verificar se determinadas afirmações são verdadeiras.

```python
# Base de conhecimento sobre animais
# Cada predicado é representado como um conjunto (set) de elementos

# Predicado Gato(x): conjunto de entidades que são gatos
gato = {"Felix", "Garfield", "Tom"}

# Predicado Cachorro(x): conjunto de entidades que são cachorros
cachorro = {"Rex", "Buddy", "Pluto"}

# Predicado Mamifero(x): conjunto de entidades que são mamíferos
mamifero = {"Felix", "Garfield", "Tom", "Rex", "Buddy", "Pluto", "João"}

# Predicado Pessoa(x): conjunto de entidades que são pessoas
pessoa = {"João", "Maria", "Pedro"}

# Predicado binário GostaDe(x, y): pares (x, y) onde x gosta de y
gosta_de = {
    ("João", "Felix"),
    ("João", "programação"),
    ("Maria", "Rex"),
    ("Maria", "IA"),
    ("Pedro", "Garfield"),
    ("Pedro", "IA"),
}

print("BASE DE CONHECIMENTO")
print("=" * 40)
print(f"Gatos: {gato}")
print(f"Cachorros: {cachorro}")
print(f"Mamíferos: {mamifero}")
print(f"Pessoas: {pessoa}")
print(f"Gosta de: {gosta_de}")
```

```python
# Verificando proposições simples (predicados com constantes)

print("VERIFICAÇÃO DE PROPOSIÇÕES")
print("=" * 40)

# Gato(Felix)? → Felix é um gato?
print(f"Gato(Felix)? {vf('Felix' in gato)}")

# Gato(Rex)? → Rex é um gato?
print(f"Gato(Rex)? {vf('Rex' in gato)}")

# Mamifero(Tom)? → Tom é um mamífero?
print(f"Mamifero(Tom)? {vf('Tom' in mamifero)}")

# GostaDe(Maria, IA)? → Maria gosta de IA?
print(f"GostaDe(Maria, IA)? {vf(('Maria', 'IA') in gosta_de)}")

# GostaDe(João, IA)? → João gosta de IA?
print(f"GostaDe(João, IA)? {vf(('João', 'IA') in gosta_de)}")
```

```python
# Implementando quantificadores

def para_todo(conjunto, predicado):
    """
    Quantificador Universal (∀): verifica se o predicado
    é verdadeiro para TODOS os elementos do conjunto.
    """
    for elemento in conjunto:
        if not predicado(elemento):
            return False
    return True


def existe(conjunto, predicado):
    """
    Quantificador Existencial (∃): verifica se o predicado
    é verdadeiro para PELO MENOS UM elemento do conjunto.
    """
    for elemento in conjunto:
        if predicado(elemento):
            return True
    return False


print("QUANTIFICADORES")
print("=" * 50)

# ∀x (Gato(x) → Mamifero(x)) — "Todos os gatos são mamíferos?"
resultado = para_todo(gato, lambda x: x in mamifero)
print(f"∀x (Gato(x) → Mamifero(x)): {vf(resultado)}")
print("  → Todos os gatos são mamíferos?\n")

# ∀x (Pessoa(x) → Mamifero(x)) — "Todas as pessoas são mamíferos?"
resultado = para_todo(pessoa, lambda x: x in mamifero)
print(f"∀x (Pessoa(x) → Mamifero(x)): {vf(resultado)}")
print("  → Todas as pessoas são mamíferos?\n")

# ∃x (Pessoa(x) ∧ GostaDe(x, IA)) — "Existe uma pessoa que gosta de IA?"
resultado = existe(pessoa, lambda x: (x, "IA") in gosta_de)
print(f"∃x (Pessoa(x) ∧ GostaDe(x, IA)): {vf(resultado)}")
print("  → Existe uma pessoa que gosta de IA?\n")

# ∃x (Gato(x) ∧ GostaDe(x, IA)) — "Existe um gato que gosta de IA?"
resultado = existe(gato, lambda x: (x, "IA") in gosta_de)
print(f"∃x (Gato(x) ∧ GostaDe(x, IA)): {vf(resultado)}")
print("  → Existe um gato que gosta de IA?\n")

# ∀x (Cachorro(x) → Mamifero(x)) — "Todos os cachorros são mamíferos?"
resultado = para_todo(cachorro, lambda x: x in mamifero)
print(f"∀x (Cachorro(x) → Mamifero(x)): {vf(resultado)}")
print("  → Todos os cachorros são mamíferos?")
```

### Entendendo o código acima

- Representamos cada **predicado unário** (de um argumento) como um **conjunto** (`set`) de objetos que satisfazem o predicado.
- Representamos o **predicado binário** `GostaDe(x, y)` como um conjunto de **tuplas** `(x, y)`.
- A função `para_todo` implementa o quantificador **universal** (∀): percorre todos os elementos e verifica se o predicado é verdadeiro para cada um.
- A função `existe` implementa o quantificador **existencial** (∃): retorna `True` assim que encontra um elemento que satisfaz o predicado.
- Usamos **funções lambda** para definir as condições de forma compacta.

## 7. Exemplo em Python: Motor de Inferência Simples

Um **motor de inferência** é o componente de um sistema de IA que aplica regras lógicas a fatos conhecidos para derivar novos fatos. Vamos implementar um motor de inferência simples usando a estratégia de **encadeamento para frente** (*forward chaining*).

### Como funciona o encadeamento para frente?

1. Começamos com um conjunto de **fatos** conhecidos.
2. Verificamos todas as **regras**: se as condições (antecedente) de uma regra são satisfeitas pelos fatos, adicionamos a conclusão (consequente) aos fatos.
3. Repetimos o processo até que nenhum novo fato seja derivado.

```python
def motor_de_inferencia(fatos, regras):
    """
    Motor de inferência com encadeamento para frente (forward chaining).

    Parâmetros:
        fatos (set): conjunto de fatos iniciais (strings)
        regras (list): lista de regras, cada regra é um dicionário com:
            - 'condicoes': lista de fatos necessários (antecedente)
            - 'conclusao': fato a ser derivado (consequente)
            - 'nome': nome descritivo da regra (opcional)

    Retorna:
        set: conjunto de todos os fatos (iniciais + derivados)
    """
    # Fazemos uma cópia dos fatos para não alterar o original
    fatos = set(fatos)
    rodada = 1

    print("MOTOR DE INFERÊNCIA — Encadeamento para Frente")
    print("=" * 55)
    print(f"Fatos iniciais: {fatos}")
    print()

    # Repetir enquanto novos fatos forem derivados
    houve_mudanca = True
    while houve_mudanca:
        houve_mudanca = False
        print(f"--- Rodada {rodada} ---")

        for regra in regras:
            condicoes = regra["condicoes"]
            conclusao = regra["conclusao"]
            nome = regra.get("nome", "")

            # Verificar se TODAS as condições estão nos fatos
            todas_satisfeitas = all(c in fatos for c in condicoes)

            if todas_satisfeitas and conclusao not in fatos:
                fatos.add(conclusao)
                houve_mudanca = True
                print(f"  Regra '{nome}': {condicoes} → '{conclusao}' ✓ NOVO FATO!")

        if not houve_mudanca:
            print("  Nenhum novo fato derivado. Parando.")

        rodada += 1
        print()

    print(f"Fatos finais: {fatos}")
    return fatos
```

```python
# Exemplo 1: Classificação de animais
#
# Fatos iniciais:
#   - Tom tem pelos
#   - Tom mia
#
# Regras:
#   R1: Se tem pelos → é mamífero
#   R2: Se mia → é gato
#   R3: Se é mamífero E é gato → é animal doméstico
#   R4: Se é animal doméstico → precisa de cuidados veterinários

fatos_iniciais = {"tem_pelos", "mia"}

regras = [
    {
        "nome": "R1",
        "condicoes": ["tem_pelos"],
        "conclusao": "é_mamífero"
    },
    {
        "nome": "R2",
        "condicoes": ["mia"],
        "conclusao": "é_gato"
    },
    {
        "nome": "R3",
        "condicoes": ["é_mamífero", "é_gato"],
        "conclusao": "é_animal_doméstico"
    },
    {
        "nome": "R4",
        "condicoes": ["é_animal_doméstico"],
        "conclusao": "precisa_veterinário"
    },
]

resultado = motor_de_inferencia(fatos_iniciais, regras)
```

```python
# Exemplo 2: Sistema de diagnóstico simplificado
#
# Fatos iniciais: sintomas de um paciente
# Regras: relacionam sintomas a possíveis diagnósticos

fatos_paciente = {"febre", "tosse", "dor_de_garganta"}

regras_diagnostico = [
    {
        "nome": "R1",
        "condicoes": ["febre", "tosse"],
        "conclusao": "possível_gripe"
    },
    {
        "nome": "R2",
        "condicoes": ["febre", "dor_de_garganta"],
        "conclusao": "possível_faringite"
    },
    {
        "nome": "R3",
        "condicoes": ["possível_gripe"],
        "conclusao": "recomendar_repouso"
    },
    {
        "nome": "R4",
        "condicoes": ["possível_faringite"],
        "conclusao": "recomendar_consulta_médica"
    },
    {
        "nome": "R5",
        "condicoes": ["possível_gripe", "possível_faringite"],
        "conclusao": "quadro_complexo_procurar_especialista"
    },
]

resultado = motor_de_inferencia(fatos_paciente, regras_diagnostico)
```

### Entendendo o Motor de Inferência

O motor de inferência acima é um exemplo simplificado, mas demonstra o princípio fundamental usado em muitos sistemas de IA:

1. **Fatos** representam o que sabemos (a base de conhecimento).
2. **Regras** representam o conhecimento do especialista (relações causa-efeito).
3. O **encadeamento para frente** começa dos fatos e "caminha" em direção às conclusões.
4. O processo é **iterativo**: novas conclusões podem ativar novas regras.
5. O processo para quando nenhum novo fato pode ser derivado (**ponto fixo**).

Este é o mesmo princípio usado em **sistemas especialistas** reais, como:
- **MYCIN** (diagnóstico médico)
- **DENDRAL** (análise química)
- **R1/XCON** (configuração de computadores)

## 8. Diferenças entre Lógica Proposicional e Lógica de Primeira Ordem

Agora que estudamos ambas as lógicas, vamos comparar suas características:

| Característica | Lógica Proposicional | Lógica de Primeira Ordem |
|---|---|---|
| **Unidade básica** | Proposições (sentenças completas) | Predicados com argumentos |
| **Variáveis** | Não possui | Sim (x, y, z...) |
| **Quantificadores** | Não possui | ∀ (para todo) e ∃ (existe) |
| **Expressividade** | Limitada | Muito mais expressiva |
| **Exemplo** | "Está chovendo" (p) | Clima(hoje, chuva) |
| **Generalização** | Não consegue | "∀x Humano(x) → Mortal(x)" |
| **Relações** | Não representa | GostaDe(João, Maria) |
| **Complexidade** | Decidível (NP-completo) | Semi-decidível |
| **Uso típico em IA** | Circuitos lógicos, SAT solvers | Bases de conhecimento, Prolog |

### Quando usar cada uma?

- **Lógica Proposicional** é suficiente quando:
  - As informações são simples e não envolvem relações entre objetos.
  - O domínio é pequeno e finito.
  - Exemplo: modelar um sistema de alarme com sensores ligados/desligados.

- **Lógica de Primeira Ordem** é necessária quando:
  - Precisamos falar sobre **objetos** e suas **propriedades**.
  - Precisamos expressar **generalizações** ("todos", "existe").
  - O domínio envolve **relações** entre entidades.
  - Exemplo: modelar relações familiares, bases de conhecimento médico.

### Analogia para lembrar

Pense na Lógica Proposicional como um **interruptor de luz**: está ligado (V) ou desligado (F), ponto final.

Pense na Lógica de Primeira Ordem como uma **planilha**: além de ter valores verdadeiros/falsos, você tem linhas (objetos), colunas (propriedades) e pode fazer perguntas como "todos os valores desta coluna são maiores que 10?".
